Let's walk through this **neural network** layer by layer and show how — **without activation functions** — the output becomes just a **linear combination of the inputs**.

---

###  **Assumptions from the diagram:**

- Inputs:  
  - $( x_1 = 68 )$  
  - $( x_2 = 14 )$

- Weights:  
  - $( w_1, w_2 )$ → from $( x_1 )$  
  - $( w_3, w_4 )$ → from $( x_2 )$  
  - $( w_5, w_6 )$ → from hidden layer to output  

- Biases:  
  - $( b_1, b_2 )$ → for the two hidden layer neurons  
  - $( b_3 )$ → for output neuron  

Let’s denote the hidden neurons as:

- $( h_1 = w_1 x_1 + w_3 x_2 + b_1 )$  
- $( h_2 = w_2 x_1 + w_4 x_2 + b_2 )$


Then, the output neuron becomes:

- $( y_{\text{pred}} = w_5 h_1 + w_6 h_2 + b_3 )$

---

### **Now substitute hidden layer values into output:**

$[
y_{\text{pred}} = w_5 (w_1 x_1 + w_3 x_2 + b_1) + w_6 (w_2 x_1 + w_4 x_2 + b_2) + b_3
]$

Distribute the terms:

$[
y_{\text{pred}} = (w_5 w_1 + w_6 w_2) x_1 + (w_5 w_3 + w_6 w_4) x_2 + (w_5 b_1 + w_6 b_2 + b_3)
]$

---

###  **This is a linear equation:**

$[
y_{\text{pred}} = A x_1 + B x_2 + C
]$

Where:
- $( A = w_5 w_1 + w_6 w_2 )$
- $( B = w_5 w_3 + w_6 w_4 )$
- $( C = w_5 b_1 + w_6 b_2 + b_3 )$

---

###  **Conclusion:**

Even though this is a 2-layer neural network:
- **Without any activation functions**, it simply reduces to a **linear model** — just like linear regression.
- It cannot learn any **nonlinear relationships** between $( x_1 )$ and $( x_2 )$.
