## Image Resizing, Cropping, and Flipping

```python
# Load the image
image = cv2.imread(data_path)  # Replace with your image path
cv2.imshow("Original Image", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# -------------------------
# 1. Resize the image
# -------------------------
resized = cv2.resize(image, (500, 500))  # Resize to 400x300
cv2.imshow("Resized Image (400x300)", resized)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# -------------------------
# 2. Crop the image
# -------------------------



# Get image dimensions
height, width = image.shape[:2]

# Crop the region: image[y1:y2, x1:x2]
# Example: Crop the top-left 100x100 region
cropped = image[100:200, 100:200]

# Display the cropped image
cv2.imshow("Cropped Image", cropped)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python





# -------------------------
# 3. Flip the image
# -------------------------
flip_horizontal = cv2.flip(image, 1)  # 1 = horizontal flip
flip_vertical = cv2.flip(image, 0)    # 0 = vertical flip
flip_both = cv2.flip(image, -1)       # -1 = both axes

cv2.imshow("Flipped Horizontally", flip_horizontal)
cv2.imshow("Flipped Vertically", flip_vertical)
cv2.imshow("Flipped Both", flip_both)

# Wait until a key is pressed and close windows
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## **DRAWING ON IMAGES**

```python
# 1. Import required libraries
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Utility function to display image in Jupyter
def show_img(img, title="Image", cmap_type=None):
    plt.figure(figsize=(6, 4))
    if cmap_type:
        plt.imshow(img, cmap=cmap_type)
    else:
        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        plt.imshow(img_rgb)
    plt.title(title)
    plt.axis("off")
    plt.show()
```

Create a blank image (black canvas)

```python
# 2. Create a blank image (black canvas)
blank_img = np.zeros((400, 600, 3), dtype=np.uint8)
show_img(blank_img, "Blank Canvas")
```

```python
# 3. Draw a green line
cv2.line(blank_img, (50, 50), (550, 50), color=(0, 255, 0), thickness=3)
show_img(blank_img, "Line Drawn")
```

```python
# 4. Draw a blue rectangle
cv2.rectangle(blank_img, (100, 100), (500, 250), color=(255, 0, 0), thickness=2)
show_img(blank_img, "Rectangle Drawn")
```

thickness = 1: draws a thin outline of 1 pixel.

thickness = 5: draws a thicker outline.

thickness = -1: draws a filled shape.

```python
# 5. Draw a red filled circle
cv2.circle(blank_img, center=(300, 175), radius=40, color=(0, 0, 255), thickness=-1)
show_img(blank_img, "Circle Drawn")
```

```python
# 6. Draw an ellipse
cv2.ellipse(blank_img, (300, 300), (100, 50), 0, 0, 180, (255, 255, 0), 2)
show_img(blank_img, "Ellipse Drawn")
```
