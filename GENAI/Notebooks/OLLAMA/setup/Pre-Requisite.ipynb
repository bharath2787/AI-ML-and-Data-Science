### **Installation Guide**  

#### **1. Install Docker**  
Docker is required to run Open WebUI in a containerized environment.  
- **Windows & macOS**: Download and install **Docker Desktop** from [Docker's official website](https://www.docker.com/products/docker-desktop/).  
- **Linux (Debian/Ubuntu-based)**: Run the following commands:  
  ```sh
  sudo apt update
  sudo apt install -y docker.io
  sudo systemctl enable --now docker
  ```
  To verify the installation:  
  ```sh
  docker --version
  ```

#### **2. Install Ollama**

- **macOS (via Homebrew):**

#### **3. Install Open WebUI**  
Once Docker is installed, run the following command to set up Open WebUI:  
```sh
docker run -d -p 3000:8080 --add-host=host.docker.internal:host-gateway \
-v open-webui:/app/backend/data --name open-webui --restart always \
ghcr.io/open-webui/open-webui:main
```
