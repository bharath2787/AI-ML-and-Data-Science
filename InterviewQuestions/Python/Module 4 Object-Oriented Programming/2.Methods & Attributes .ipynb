### **11. What is the difference between an instance method, class method, and static method?**

* **Instance method**: Takes `self` as the first argument and can access instance and class data.
* **Class method**: Uses `@classmethod` and takes `cls` as the first argument, accessing class-level data.
* **Static method**: Uses `@staticmethod`, takes no special first argument, and behaves like a regular function within a class.

---

### **12. When should you use a `@staticmethod`?**

Use `@staticmethod` when the method doesn’t need to access instance or class data. It's purely utility logic grouped within a class context.

---

### **13. How does `@classmethod` work, and when would you use it?**

`@classmethod` receives the class (`cls`) as the first argument. It’s used to access or modify class state and for creating alternative constructors.

---

### **14. What is the significance of the `self` keyword?**

`self` refers to the current object instance. It is used to access instance variables and methods from within class methods.

---

### **15. What happens if you omit `self` in an instance method?**

Python will raise a `TypeError` because the method expects an object instance but receives fewer arguments than defined.

---

### **16. How do you define and use a property in Python?**

You can use the `@property` decorator to define a getter, and `@<property_name>.setter` to define a setter:

```python
@property  
def age(self):  
    return self._age
```

---

### **17. What is the difference between a method and a function in Python?**

A method is a function defined inside a class and takes `self` or `cls` as a parameter. A function is standalone and not bound to a class or instance.

---

### **18. How do you use the `property()` function to create getters and setters?**

```python
def get_x(self): return self._x  
def set_x(self, value): self._x = value  
x = property(get_x, set_x)
```

It binds access control logic to attribute access.

---

### **19. What are dunder (magic) methods? Name a few commonly used ones.**

Dunder methods are special methods with double underscores, like `__init__`, `__str__`, `__repr__`, `__len__`, `__eq__`, `__add__`. They customize object behavior in built-in operations.

---

### **20. How can you override built-in methods like `__str__` or `__eq__`?**

You override them by defining them in your class. For example:

```python
def __str__(self):  
    return f"{self.name}"
```

This controls how the object is represented as a string.
