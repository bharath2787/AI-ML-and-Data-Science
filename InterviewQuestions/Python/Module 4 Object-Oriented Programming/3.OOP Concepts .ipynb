### **21. What are the four pillars of OOP?**

1. **Encapsulation** – Bundling data and methods.
2. **Abstraction** – Hiding internal details.
3. **Inheritance** – Reusing code via parent-child relationships.
4. **Polymorphism** – Using a single interface to represent different types or behaviors.

---

### **22. What is inheritance, and how does Python implement it?**

Inheritance allows a class (child) to acquire attributes and methods from another class (parent). In Python, it’s implemented by passing the parent class in parentheses:

```python
class Child(Parent):
    pass
```

---

### **23. How is method overriding different from method overloading?**

* **Overriding**: Redefining a method in the subclass with the same name and signature as in the parent.
* **Overloading**: Not directly supported in Python. Achieved via default or variable-length arguments.

---

### **24. What is multiple inheritance, and how is it handled in Python?**

Multiple inheritance allows a class to inherit from more than one parent class. Python resolves conflicts using **Method Resolution Order (MRO)**, typically using the C3 linearization algorithm.

---

### **25. Explain polymorphism with a real-world Python example.**

Polymorphism allows different classes to be treated through a common interface:

```python
class Dog:
    def speak(self): return "Bark"

class Cat:
    def speak(self): return "Meow"

def animal_sound(animal):
    print(animal.speak())
```

---

### **26. What is encapsulation? How is it implemented in Python?**

Encapsulation restricts direct access to an object’s data. In Python, it's implemented by defining attributes as protected (`_var`) or private (`__var`) and using methods or properties to access them.

---

### **27. What is the difference between is-a and has-a relationships?**

* **Is-a**: Inheritance (e.g., `Dog is-a Animal`)
* **Has-a**: Composition (e.g., `Car has-a Engine`)

---

### **28. What is duck typing in Python?**

Duck typing is dynamic typing where the object's suitability is determined by presence of methods/attributes rather than the actual type. “If it quacks like a duck, it's a duck.”

---

### **29. Can you explain the SOLID principles?**

* **S**: Single Responsibility
* **O**: Open/Closed Principle
* **L**: Liskov Substitution
* **I**: Interface Segregation
* **D**: Dependency Inversion
  These guide scalable, maintainable OOP design.

---

### **30. How do OOP concepts help in code reuse and maintenance?**

OOP encourages modularity, abstraction, and reusability. Inheritance promotes reuse, while encapsulation and abstraction reduce complexity, making code easier to maintain and extend.
