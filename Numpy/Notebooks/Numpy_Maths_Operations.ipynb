# 10 most commonly used NumPy mathematical operations:

```python
# Import NumPy library
import numpy as np
```

```python

# Example 1: Array creation
print("1. Array Creation:")
a = np.array([1, 2, 3, 4, 5])
print("Array:", a)
```

```python
# Example 2: Element-wise arithmetic operations
print("\n2. Element-wise Arithmetic Operations:")
b = np.array([6, 7, 8, 9, 10])
print("array1: ",a)

print("array b: " ,b)



# Addition
print("Addition:", a + b)

# Subtraction
print("Subtraction:", a - b)

# Multiplication
print("Multiplication:", a * b)

# Division
print("Division:", a / b)

```

```python

# Example 3: Array broadcasting
print("\n3. Array Broadcasting:")
c = np.array([[1, 2, 3], [4, 5, 6]])
print("Original Array:")
print(c)

# Add scalar to array
print("Add Scalar to Array:")
print(c + 10)

# Multiply array by scalar
print("Multiply Array by Scalar:")
print(c * 2)
```

```python

# Example 4: Array aggregation functions
print("\n4. Array Aggregation Functions:")
d = np.array([1, 2, 3, 4, 5])
print("Array:", d)

# Sum
print("Sum:", np.sum(d))

# Mean
print("Mean:", np.mean(d))

# Maximum
print("Maximum:", np.max(d))

# Minimum
print("Minimum:", np.min(d))
```

```python

# Example 5: Array manipulation functions
print("\n5. Array Manipulation Functions:")
e = np.array([[1, 2], [3, 4], [5, 6]])
print("Original Array:")
print(e)

# Transpose
print("Transpose:")
print(np.transpose(e))

# Reshape
print("Reshape:")
print(np.reshape(e, (2, 3)))

# Flatten
print("Flatten:")
print(e.flatten())
```

```python
ar  = np.arange(1,13).reshape(2,3,2)
print(ar)
```

```python
# Reshape
print("Reshape:")
print(np.reshape(ar, (2, 3,2)))
```

```python

# Example 6: Array sorting
print("\n6. Array Sorting:")
f = np.array([5, 2, 9, 1, 7])
print("Original Array:", f)

# Sort array
print("Sorted Array:", np.sort(f))
```

```python

# Example 7: Linear algebra operations
print("\n7. Linear Algebra Operations:")
g = np.array([[1, 2], [3, 4]])
h = np.array([[5, 6], [7, 8]])

# Matrix multiplication
print("Matrix Multiplication:")
print(np.dot(g, h))



```

```python

# Example 8: Random number generation
print("\n8. Random Number Generation:")
# Generate random integers
print("Random Integers:", np.random.randint(1, 10, size=5))

# Generate random floats
print("Random Floats:", np.random.rand(5))
```

```python

# Example 9: Trigonometric functions
print("\n9. Trigonometric Functions:")
i = np.array([0, np.pi/2, np.pi])
print("Array:", i)

# Sine
print("Sine:", np.sin(i))

# Cosine
print("Cosine:", np.cos(i))

# Tangent
print("Tangent:", np.tan(i))
```

```python

# Example 10: Exponential and logarithmic functions
print("\n10. Exponential and Logarithmic Functions:")
j = np.array([1, 2, 3])
print("Array:", j)

# Exponential
print("Exponential:", np.exp(j))

# Natural logarithm
print("Natural Logarithm:", np.log(j))
```
