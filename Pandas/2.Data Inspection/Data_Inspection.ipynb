### df.head()
- df.head()
- The head() method returns the first n rows of the DataFrame.
- n is an optional parameter that defaults to 5.

```python



print("First 3 rows of the DataFrame:")
df.head(1)  # Example 1: Displaying the first 3 rows
```

### df.Info()

- df.info()
- The info() method prints a concise summary of the DataFrame.
- It includes the data types of each column, the number of non-null values, and memory usage.
- Documentation: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.info.html

```python


print("\nInformation about the DataFrame:")
df.info()  # Example 1: Displaying information about the DataFrame
```

#df.describe()

- The describe() method generates descriptive statistics for numerical columns.
- It includes metrics such as count, mean, std (standard deviation), min, 25%, 50%, 75%, and max.
- Documentation: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.describe.html

```python


print("\nDescriptive statistics of the DataFrame:")
df.describe()  # Example 1: Descriptive statistics for all numerical columns
```


# df.dtypes
- The dtypes attribute returns the data types of each column in the DataFrame.

```python


print("\nData types of each column in the DataFrame:")
print(df.dtypes)  # Example 1: Displaying data types of each column
```
