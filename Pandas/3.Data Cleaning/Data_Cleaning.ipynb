### HEAT MAP to show null values in dataset

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample DataFrame
import pandas as pd
import numpy as np

data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [5, np.nan, np.nan, 8, 9],
    'C': ['foo', 'bar', 'baz', 'qux', np.nan],
    'D': [np.nan, 3.5, 6.1, np.nan, 2.9],
    'E': [1, 3.5, 6.1, 1, 2.9]
}
df = pd.DataFrame(data)

# Visualizing missing values
plt.figure(figsize=(10, 6))
heatmap = sns.heatmap(df.isnull(), cbar=True, cmap='viridis', yticklabels=True)

# Add labels for the color bar
cbar = heatmap.collections[0].colorbar
cbar.set_ticks([0, 1])
cbar.set_ticklabels(['Not Null', 'Null'])

# Add title and labels
plt.title('Heatmap of Missing Values')
plt.xlabel('Columns')
plt.ylabel('Rows')

plt.show()
```
