#Aggregate Method using DICTONARY

```python
# agg method
# passing dict
genres.agg(
    {
        'Runtime':'mean',
        'IMDB_Rating':'mean',
        'No_of_Votes':'sum',
        'Gross':'sum',
        'Metascore':'min'
    }
)
```

#Aggregate Using LIST

```python
# passing list
genres.agg(['min','max','sum'])
```

#Aggregation using combination of DICT and LIST

```python

genres.agg(
    {
        'Runtime':['min','mean'],
        'IMDB_Rating':'mean',
        'No_of_Votes':['sum','max'],
        'Gross':'sum',
        'Metascore':'min'
    }
)
```
