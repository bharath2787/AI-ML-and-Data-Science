### Reading JSON from a String
# JSON string example

```python

json_string = '''
[
    {
        "name": "John Doe",
        "age": 30,
        "city": "New York"
    },
    {
        "name": "Jane Smith",
        "age": 25,
        "city": "Los Angeles"
    },
    {
        "name": "Mike Johnson",
        "age": 35,
        "city": "Chicago"
    }
]
'''
```

```python

df_json_string = pd.read_json(json_string)
print("DataFrame from JSON string:")
df_json_string
```


## pd.read_html()

### Reading HTML from a File

```python

# Sample HTML file: data.html

html_file_path = '/content/drive/MyDrive/0.Latest_DS_Course/Pandas/Data/data.html'
dfs_html_file = pd.read_html(html_file_path)
```

```python
type(dfs_html_file)
```

```python
len(dfs_html_file)
```

```python
dfs_html_file[0]
```

```python
print("DataFrames from HTML file:")
for i, df in enumerate(dfs_html_file):
    print(f"DataFrame {i+1}:")
    display(df)
```

```python
type(df['Name'])
```
