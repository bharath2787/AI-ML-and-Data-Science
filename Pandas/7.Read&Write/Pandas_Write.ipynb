# Pandas Export Methods

# This notebook explains how to use various pandas methods to export DataFrames to different file formats.

```python

import pandas as pd
```

# Sample DataFrame


```python

data = {
    'Name': ['John', 'Anna', 'Peter', 'Linda'],
    'Age': [28, 24, 35, 32],
    'City': ['New York', 'Paris', 'Berlin', 'London']
}
df = pd.DataFrame(data)

# Display the DataFrame
df
```

```python
df1 = pd.read_csv("imdb-top-1000.csv")
```


# Export to CSV

```python


df.to_csv('sample.csv', index=False)
```


# Read the CSV file to verify

```python

pd.read_csv('sample.csv')
```

# Export to Excel
 - The index=False parameter in the df.to_excel('sample.xlsx', index=False) method call specifies that the row indices of the DataFrame should not be included in the output file. By default, pandas includes the row indices when exporting a DataFrame to a file. Setting index=False ensures that only the data columns are written to the file, without the row inde

image.png

```python
df.to_excel('sample.xlsx', index=False)
```

# Read the Excel file to verify


```python

pd.read_excel('sample.xlsx')
```

# Export to JSON


```python

df.to_json('sample.json', orient="records", indent=2)
```

# Read the JSON file to verify


```python

pd.read_json('sample.json')
```

# Export to HTML


```python


df.to_html('sample.html', index=False)
```

# Read the HTML file to verify


```python

with open('sample.html', 'r') as file:
    html_content = file.read()
html_content
```
