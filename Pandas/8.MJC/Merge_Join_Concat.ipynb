### Merging, Joining, and Concatenation in Pandas

## **üîπ Merge: All Scenarios (Using Columns and Index)**

### **1Ô∏è‚É£ Merging on a Common Column (SQL-Style Join)**

```python
import pandas as pd

df1 = pd.DataFrame({'ID': [1, 2, 3], 'Name': ['Alice', 'Bob', 'Charlie']})
df2 = pd.DataFrame({'ID': [2, 3, 4], 'Score': [85, 90, 78]})

df_merged = df1.merge(df2, on='ID', how='inner')  # Inner join
df_merged
```

### **2Ô∏è‚É£ Different Join Types in `merge()`**


```python
df_left = df1.merge(df2, on='ID', how='left')   # Left Join
df_right = df1.merge(df2, on='ID', how='right') # Right Join
df_outer = df1.merge(df2, on='ID', how='outer') # Outer Join
df_outer
```

### **3Ô∏è‚É£ Merging on Index Instead of Column**


```python
df1 = pd.DataFrame({'Name': ['Alice', 'Bob', 'Charlie']}, index=[1, 2, 3])
df2 = pd.DataFrame({'Score': [85, 90, 78]}, index=[2, 3, 4])
display(df1)
display(df2)
df_merged = df1.merge(df2, left_index=True, right_index=True, how='outer')
display(df_merged)
```


### **üìå Explanation**
- `left_index=True` ‚Üí Uses `df1`'s **index** as the merge key.
- `right_index=True` ‚Üí Uses `df2`'s **index** as the merge key.
- **`how='outer'`** ‚Üí Includes **all** index values, filling missing ones with `NaN`.

‚úÖ **This method is useful when both DataFrames already have meaningful indexes!**  


### **Final Thoughts**
- **Use `merge(on='column')` when joining on a column.**  
- **Use `merge(left_index=True, right_index=True)` when joining on an index.**  


## **üîπ Join: All Scenarios (Using Columns and Index)**

### **4Ô∏è‚É£ Joining on Index (Default Behavior)**

```python
display(df1)
display(df2)
```

```python
df_joined = df1.join(df2, how='outer')
df_joined
```

### **5Ô∏è‚É£ Joining on a Column Instead of Index**


```python
df1 = pd.DataFrame({'ID': [1, 2, 3], 'Name': ['Alice', 'Bob', 'Charlie']})
df2 = pd.DataFrame({'ID': [1, 2, 4], 'Score': [85, 90, 78]})

df_joined = df1.set_index('ID').join(df2.set_index('ID'), how='left')
print(df_joined)
```

## **üîπ Concatenation: All Scenarios (Horizontal and Vertical)**

### **6Ô∏è‚É£ Vertical Concatenation (`axis=0`)**

```python
df1 = pd.DataFrame({'ID': [1, 2], 'Name': ['Alice', 'Bob']})
df2 = pd.DataFrame({'ID': [3, 4], 'Name': ['Charlie', 'David']})

df_vertical = pd.concat([df1, df2], axis=0, ignore_index=True)
df_vertical
```

### **7Ô∏è‚É£ Horizontal Concatenation (`axis=1`)**


```python
df1 = pd.DataFrame({'A': [1, 2, 3]})
df2 = pd.DataFrame({'B': [4, 5, 6]})

df_horizontal = pd.concat([df1, df2], axis=1)
df_horizontal
```

### **8Ô∏è‚É£ Handling Mismatched Indexes in Concatenation**


```python
df1 = pd.DataFrame({'A': [1, 2]}, index=[0, 1])
df2 = pd.DataFrame({'B': [3, 4]}, index=[1, 2])

df_mismatch = pd.concat([df1, df2], axis=1)
df_mismatch
```

## **üîπ Comparison: `merge()` vs. `join()` vs. `concat()`**

| Feature | `merge()` | `join()` | `concat()` |
|---------|----------|----------|-----------|
| SQL-style joins on column | ‚úÖ Yes | ‚ùå No | ‚ùå No |
| Join on index | ‚ùå No | ‚úÖ Yes | ‚ùå No |
| Stack DataFrames vertically | ‚ùå No | ‚ùå No | ‚úÖ Yes |
| Stack DataFrames horizontally | ‚ùå No | ‚ùå No | ‚úÖ Yes |
| Inner/Outer/Left/Right joins | ‚úÖ Yes | ‚úÖ Yes | ‚ùå No |
| Works with MultiIndex | ‚úÖ Yes | ‚úÖ Yes | ‚ùå No |
| Handles duplicate column names | ‚úÖ Yes | ‚ö†Ô∏è Needs suffix | ‚ùå No |
| Best for structured, relational data | ‚úÖ Yes | ‚ùå No | ‚ùå No |
| Best for quick index-based joins | ‚ùå No | ‚úÖ Yes | ‚ùå No |
| Best for combining multiple DataFrames | ‚ùå No | ‚ùå No | ‚úÖ Yes |

---

## **üîπ Final Recommendation**
- **Use `merge()`** when **you need to join on a specific column (SQL-style joins).**
- **Use `join()`** when **you want to join DataFrames based on the index.**
- **Use `concat()`** when **you need to stack DataFrames (vertically or horizontally).**
