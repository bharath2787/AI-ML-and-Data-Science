```python
import pandas as pd
import numpy as np
```

```python

# Creating a DataFrame
data = {'A': [1, 2, 3, 4], 'B': [5, 6, 7, 8], 'C': [2, 4, 6, 8]}
df = pd.DataFrame(data)
print("Original DataFrame:")
display(df)
```

```python

# Using map() with a function on a Series
df['A_mapped_function'] = df['A'].map(lambda x: x * 2)
print("\nmap() with function:")
display(df)
```

```python
# Using map() with a dictionary mapping
dict_map = {1: 'One', 2: 'Two', 3: 'Three', 4: 'Four'}
df['A_mapped_dict'] = df['A'].map(dict_map)
print("\nmap() with dictionary:")
display(df)
```

```python
# Using map() with a Series
df['A_mapped_series'] = df['A'].map(pd.Series({1: 'First', 2: 'Second', 3: 'Third', 4: 'Fourth'}))
print("\nmap() with Series:")
display(df)
```

```python
# Using apply() with a function on a DataFrame
def sum_columns(row):
    return row['A'] + row['B']
```

```python

df['Sum_Apply_Function'] = df.apply(sum_columns, axis=1)
print("\napply() with function:")
display(df)
```

```python
# Using apply() with a lambda function
df['B_Apply_Lambda'] = df['B'].apply(lambda x: x ** 2)
print("\napply() with lambda:")
display(df)
```

```python
# Using apply() with list comprehension
df['A_Apply_ListComp'] = df['A'].apply(lambda x: [x, x*2, x*3])
print("\napply() with list comprehension:")
display(df)
```

```python
## Using apply() with a lambda function (Column-wise)

df['C_Cubed'] = df['C'].apply(lambda x: x ** 3)
df
```

```python
## Using apply() with a lambda function (Row-wise)
```

```python
df['Row_Product'] = df.apply(lambda row: row['A'] * row['B'] * row['C'], axis=1)
df
```
