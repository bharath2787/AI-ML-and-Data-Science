# String Methods in Python


## 1. Convert to Upper Case
The `upper()` method converts all characters in a string to uppercase.

```
text = "hello world"
print("Original:", text)
print("Uppercase:", text.upper())
```

## 2. Convert to Lower Case
The `lower()` method converts all characters in a string to lowercase.

```
text = "HELLO WORLD"
print("Original:", text)
print("Lowercase:", text.lower())
```

## 3. Remove Whitespace using `strip()`
The `strip()` method removes leading and trailing whitespace from a string.

```
text = "  Python Programming  "
print("Original:", repr(text))
print("Stripped:", repr(text.strip()))
```

## 4. Replace Substring
The `replace()` method replaces occurrences of a specified substring with another substring.

```
text = "I love Java"
print("Original:", text)
print("Replaced:", text.replace("Java", "Python"))
```

## 5. Find Substring Position
The `find()` method returns the index of the first occurrence of a substring, or -1 if not found.

```
text = "Python is fun"
print("Find 'is':", text.find("is"))  # Returns index
print("Find 'Java':", text.find("Java"))  # Returns -1 if not found
```

## 6. Get Index of Substring
The `index()` method returns the index of the first occurrence of a substring, raising an error if not found.

```
text = "Learn Python Programming"
print("Index of 'Python':", text.index("Python"))
# print("Index of 'Java':", text.index("Java"))  # Raises ValueError if not found
```

## 7. Check if String Starts With
The `startswith()` method checks if the string begins with a specified substring.

```
text = "Hello, World!"
print("Starts with 'Hello':", text.startswith("Hello"))
print("Starts with 'World':", text.startswith("World"))
```

## 8. Check if String Ends With
The `endswith()` method checks if the string ends with a specified substring.

```
text = "data.csv"
print("Ends with '.csv':", text.endswith(".csv"))
print("Ends with '.txt':", text.endswith(".txt"))
```

## 9. Split a String
The `split()` method splits a string into a list based on a specified delimiter.

```
text = "apple,banana,cherry"
print("Original:", text)
print("Split:", text.split(","))
```
