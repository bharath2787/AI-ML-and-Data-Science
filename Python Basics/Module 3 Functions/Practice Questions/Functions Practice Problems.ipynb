15 practice problems based on the agenda:

### **Inbuilt vs. User-Defined Functions**
1. **Problem 1:**  
   What is the difference between an inbuilt function and a user-defined function? Provide examples of each.

2. **Problem 2:**  
   Write a Python program that uses the inbuilt `max()` function to find the largest number in a list. Compare it with the output from a user-defined function that performs the same task.

3. **Problem 3:**  
   Use the inbuilt `len()` function to find the length of a string. Then, create a user-defined function that takes a string as input and returns its length.

### **User-Defined Functions**
4. **Problem 4:**  
   Write a user-defined function `greet_user(name)` that takes a user's name as an argument and prints a greeting message ("Hello, {name}").

### **Function Arguments**
8. **Problem 8:**  
   Write a Python function `sum_numbers(a, b)` that takes two arguments and returns their sum. Call the function with positional arguments.

9. **Problem 9:**  
   Create a function `introduce_yourself(name, age)` that takes two arguments, `name` and `age`, and prints a short introduction. Call the function with keyword arguments.

10. **Problem 10:**  
    Define a function `multiply(x, y=5)` where the second argument has a default value of 5. Call the function with both positional and keyword arguments.

### **Types of Function Arguments**
11. **Problem 11:**  
    Write a function `display_info(name, *hobbies)` that accepts a required argument `name` and a variable number of hobby arguments. Print the name and hobbies.

12. **Problem 12:**  
    Create a function `get_person_info(name, **details)` that accepts a required argument `name` and a variable number of keyword arguments for additional details (e.g., age, address). Print all details inside the function.

13. **Problem 13:**  
    Define a function `calculate_total_price(price, discount=0, *extra_fees)` that accepts a price, an optional discount, and any number of additional fees as arguments. The function should return the total price after applying the discount and adding extra fees.

### **Actual Arguments**

15. **Problem 15:**  
    Write a function `make_tea(tea_type, *extras, **preferences)` that takes a required argument `tea_type`, an optional variable number of extras (like sugar, milk), and a variable number of keyword preferences (like strength, temperature). Call the function with various combinations of actual arguments and print the result.
