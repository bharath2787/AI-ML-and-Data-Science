1.	Add two numbers: Create a lambda function that takes two arguments and returns their sum.
2.	Multiply by a constant: Write a lambda function that multiplies its argument by 5.
3.	Calculate the square: Create a lambda function to return the square of a given number.
4.	Check if a number is even: Write a lambda function that returns True if a number is even, and False otherwise.
5.	Find the greater number: Create a lambda function that returns the greater of two numbers.
6.	Add 15 to a number: Write a lambda function that adds 15 to a given number.
7.	Multiply two arguments: Create a lambda function that multiplies argument x with argument y and prints the result.
8.	Check for a specific starting letter: Write a lambda function to check if a given string starts with the letter "C".
9.	Extract the last character: Create a lambda function that takes a string and returns its last character.
10.	Sort by the last character: Write a lambda function that sorts a list of strings by the last character.
11.	Convert Celsius to Fahrenheit: Create a lambda function that converts Celsius to Fahrenheit.
12.	Calculate area of a rectangle: Write a lambda function to calculate the area of a rectangle given its length and width.
13.	Check if a number is divisible by another number: Create a lambda function that checks if one number is divisible by another.
14.	String reversal: Write a lambda function to reverse a string.
15.	Increment a number: Write a lambda function that takes x as parameter and returns x+2.



```python
# 1. Add two numbers
add_numbers = lambda a, b: a + b

# 2. Multiply by a constant (5)
multiply_by_5 = lambda x: x * 5

# 3. Calculate the square
square = lambda x: x ** 2

# 4. Check if a number is even
is_even = lambda x: x % 2 == 0

# 5. Find the greater number
greater_number = lambda a, b: a if a > b else b

# 6. Add 15 to a number
add_15 = lambda x: x + 15

# 7. Multiply two arguments
multiply = lambda x, y: x * y

# 8. Check for a specific starting letter "C"
starts_with_C = lambda s: s.startswith('C')

# 9. Extract the last character
last_character = lambda s: s[-1]

# 10. Sort by the last character
sort_by_last_char = lambda lst: sorted(lst, key=lambda s: s[-1])

# 11. Convert Celsius to Fahrenheit
celsius_to_fahrenheit = lambda c: (c * 9/5) + 32

# 12. Calculate area of a rectangle
rectangle_area = lambda length, width: length * width

# 13. Check if a number is divisible by another number
is_divisible = lambda x, y: x % y == 0

# 14. String reversal
reverse_string = lambda s: s[::-1]

# 15. Increment a number by 2
increment_by_2 = lambda x: x + 2

# Testing the functions
print(add_numbers(3, 7))  # Output: 10
print(multiply_by_5(4))  # Output: 20
print(square(6))  # Output: 36
print(is_even(8))  # Output: True
print(greater_number(10, 15))  # Output: 15
print(add_15(5))  # Output: 20
print(multiply(3, 4))  # Output: 12
print(starts_with_C("ChatGPT"))  # Output: True
print(last_character("Hello"))  # Output: 'o'
print(sort_by_last_char(["apple", "banana", "cherry"]))  # Output: ['banana', 'apple', 'cherry']
print(celsius_to_fahrenheit(0))  # Output: 32.0
print(rectangle_area(5, 10))  # Output: 50
print(is_divisible(10, 2))  # Output: True
print(reverse_string("Python"))  # Output: 'nohtyP'
print(increment_by_2(7))  # Output: 9
```
