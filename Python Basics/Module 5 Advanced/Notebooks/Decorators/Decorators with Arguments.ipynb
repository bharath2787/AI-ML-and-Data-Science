## **Decorators with Arguments**

```python
def repeat_decorator(times):
    def decorator(func):
        def wrapper(*args, **kwargs):
            for _ in range(times):
                func(*args, **kwargs)
        return wrapper
    return decorator

@repeat_decorator(3)
def say_hello():
    print("Hello!")

say_hello()  # Output: Hello! (repeated 3 times)
```

* Explanation: The decorator can accept parameters (like times), allowing customized behavior.
