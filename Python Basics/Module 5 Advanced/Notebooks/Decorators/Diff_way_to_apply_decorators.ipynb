## Different Ways to Apply Decorators

### 1. Using the @ Symbol (Most Common Approach)

* The decorator is applied directly to a function using the @ symbol before the function definition.

```python
@my_decorator
def my_function():
    pass
```

###Manually Passing Functions to Decorators

* A decorator can be applied by passing the function as an argument to the decorator manually.

```python
def my_function():
    pass

my_function = my_decorator(my_function)
```

###Decorating Class Methods

* For class methods, use @staticmethod or @classmethod as decorators.

```python
class MyClass:
    @staticmethod
    def static_method():
        print("This is a static method.")
```

```python
MyClass.static_method()
```

###Decorating Properties

* Use the @property decorator to define getter methods in classes.

```python
class MyClass:
    @property
    def my_property(self):
        return "This is a property."
```

```python
test = MyClass()
test.my_property
```
