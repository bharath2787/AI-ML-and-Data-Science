##Simple Decorator Example:

```python
def greet_decorator(func):
    def wrapper():
        print("Before function execution")
        func()  # Call the original function
        print("After function execution")
    return wrapper

@greet_decorator
def say_hello():
    print("Hello, World!")

say_hello()
```

##Explanation: The greet_decorator adds behavior before and after calling the original say_hello function.
