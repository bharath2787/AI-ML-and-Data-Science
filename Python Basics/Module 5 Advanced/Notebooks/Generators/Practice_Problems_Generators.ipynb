 **10 basic-level** Python generator practice problems:  

1. **Generate Numbers from 1 to n**  
   Write a generator function `count_up(n)` that yields numbers from `1` to `n`.  

2. **Generate Even Numbers**  
   Write a generator `even_numbers(n)` that yields even numbers from `0` to `n`.  

3. **Generate Odd Numbers**  
   Implement `odd_numbers(n)`, a generator that yields all odd numbers up to `n`.  

4. **Generate Multiples of a Given Number**  
   Create `multiples_of(x, n)`, a generator that yields the first `n` multiples of `x`.  

5. **Generate a Countdown**  
   Write a generator `countdown(n)` that yields numbers from `n` down to `1`.  

6. **Generate Squares of Numbers**  
   Implement `squares(n)`, a generator that yields squares of numbers from `0` to `n-1`.  

7. **Generate a Range with Step**  
   Write `custom_range(start, stop, step)`, a generator that works like `range()` but yields values lazily.  

8. **Generate Letters from a Word**  
   Create `letters(word)`, a generator that yields each letter of a given word one at a time.  

9. **Generate the First n Natural Numbers**  
   Implement `natural_numbers(n)`, which yields the first `n` natural numbers.  

10. **Generate Characters from a String in Reverse**  
    Write `reverse_string(s)`, a generator that yields characters from the string `s` in reverse order.
