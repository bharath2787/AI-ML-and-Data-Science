### **1. Create a Custom Iterator for Counting**  
   Implement a class `Counter` that behaves like an iterator, counting from `1` to `n`.  
   ```python
   counter = Counter(5)
   for num in counter:
       print(num)  # Output: 1, 2, 3, 4, 5
   ```
