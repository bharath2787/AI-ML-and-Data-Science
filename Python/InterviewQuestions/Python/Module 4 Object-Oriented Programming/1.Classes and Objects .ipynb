### **1. What is a class in Python?**

A class in Python is a blueprint for creating objects. It defines a set of attributes and methods that the created objects will have. Classes support encapsulation and are fundamental to object-oriented programming.

---

### **2. How do you create and instantiate an object?**

You define a class using the `class` keyword, then instantiate it by calling the class like a function.

```python
class Car:  
    pass  

my_car = Car()
```

### **5. Can you have multiple constructors in Python? How?**

Python does not support method overloading. Instead, you simulate multiple constructors using default arguments or class methods like `@classmethod` as factory constructors.

---

### **6. What is the difference between `__init__` and `__new__`?**

`__new__` creates and returns a new instance of the class. `__init__` initializes the instance. `__new__` is mainly used in custom object creation, especially with immutable types.

---

### **7. What is a class variable vs. an instance variable?**

A class variable is shared by all instances, while an instance variable is unique to each object. Class variables are defined directly inside the class, whereas instance variables are typically set in `__init__`.

---

### **8. How do you access class variables from an instance?**

You can access them using `self.__class__.var_name` or directly via the class name: `ClassName.var_name`.

---

### **9. What is object identity in Python?**

Object identity refers to the memory address of an object, which can be checked using the `id()` function. Two variables referring to the same object will have the same `id`.

---

### **10. How does Python handle memory management for objects?**

Python uses reference counting and a cyclic garbage collector to manage memory. When an object's reference count drops to zero, it is automatically cleaned up.
