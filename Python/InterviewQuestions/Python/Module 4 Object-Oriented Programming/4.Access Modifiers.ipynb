### **31. What are access modifiers in Python?**

Access modifiers control the visibility of class members. Python uses naming conventions:

* Public: `var`
* Protected: `_var`
* Private: `__var`

They guide access rather than enforce strict access control.

---

### **32. How do you define a private variable in Python?**

Prefix the variable with double underscores (`__`). This triggers name mangling, making it harder (but not impossible) to access from outside the class.

---

### **33. What is the difference between `_var`, `__var`, and `__var__`?**

* `_var`: Protected, by convention.
* `__var`: Private, triggers name mangling.
* `__var__`: Special (magic) method/attribute defined by Python (e.g., `__init__`, `__str__`).

---

### **34. How does name mangling work?**

Name mangling rewrites private attributes like `__value` to `_ClassName__value`, preventing accidental access but not absolute protection.

---

### **35. Are private and protected truly enforced in Python?**

No. Python relies on conventions, not strict enforcement. Name mangling discourages direct access but does not prevent it.

---

### **36. Why does Python rely on naming conventions for access control?**

Python follows a philosophy of “we are all consenting adults.” It trusts developers to respect conventions instead of enforcing strict access control.

---

### **37. How can you access a private attribute from outside the class?**

By using the mangled name: `_ClassName__attr`. Example:

```python
obj._MyClass__private_var
```

Though technically possible, it's discouraged.
