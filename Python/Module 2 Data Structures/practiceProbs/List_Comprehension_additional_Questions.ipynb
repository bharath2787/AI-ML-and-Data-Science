##  **Conceptual Questions**

### 10. **What are the advantages of using list comprehension?**

**Answer:**  
- **Readability**: List comprehensions are often more concise and easier to understand than equivalent loops.
- **Efficiency**: In some cases, list comprehensions can be slightly more efficient than loops (though the performance difference is often negligible in modern Python).
- **Pythonic**: List comprehensions are considered a more "Pythonic" way to create lists.

---

### 11. **When should you avoid using list comprehension?**

**Answer:**  
- For very complex logic that spans multiple lines, a traditional loop might be more readable.
- When side effects are involved (modifying other variables within the loop), a regular loop is generally preferred.
- Deeply nested list comprehensions can become difficult to read and maintain.
