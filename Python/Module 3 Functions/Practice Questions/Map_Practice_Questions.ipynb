2.  **Add three lists:** Use `map()` and lambda to add three given lists element-wise.
3.  **Listify strings:** Use `map()` to convert a list of strings into a list of lists, where each inner list contains the characters of the corresponding string.
4.  **Power of numbers:** Create a list containing the power of each number in a list, raised to its index using `map()`.
5.  **Square the elements:** Use `map()` to square each element of a list.
6.  **Convert case and remove duplicates:** Use `map()` to convert all characters in a string to uppercase, then eliminate duplicate letters.
7.  **Add 5 to each item:** Write a `map()` function that adds 5 to each item in the list.
8.  **Squares of items:** Write a `map()` function that returns the squares of the items in the list.
9.  **Greeting:** Write a `map()` function that adds "Hello, " in front of each item in the list.
10. **Round areas:** Use the `round()` and `range()` functions to round the elements of `circle_areas = [6.56773, 9.57668, 4.00914, 56.24241, 9.01344, 32.00013]`.




1. **Triple all numbers in a list:**
```python
numbers = [1, 2, 3, 4]
tripled_numbers = list(map(lambda x: x * 3, numbers))
print(tripled_numbers)  # Output: [3, 6, 9, 12]
```

2. **Add three lists:**
```python
list1 = [1, 2, 3]
list2 = [4, 5, 6]
list3 = [7, 8, 9]
sum_lists = list(map(lambda x, y, z: x + y + z, list1, list2, list3))
print(sum_lists)  # Output: [12, 15, 18]
```

3. **Listify strings:**
```python
strings = ['hello', 'world']
listified_strings = list(map(lambda s: list(s), strings))
print(listified_strings)  # Output: [['h', 'e', 'l', 'l', 'o'], ['w', 'o', 'r', 'l', 'd']]
```

4. **Power of numbers:**
```python
numbers = [2, 3, 4]
powered_numbers = list(map(lambda x, i: x ** i, numbers, range(len(numbers))))
print(powered_numbers)  # Output: [2, 9, 64] (2^0, 3^1, 4^2)
```

5. **Square the elements:**
```python
numbers = [1, 2, 3, 4]
squared_numbers = list(map(lambda x: x ** 2, numbers))
print(squared_numbers)  # Output: [1, 4, 9, 16]
```

6. **Convert case and remove duplicates:**
```python
input_string = "aAbBcCdD"
# Convert to uppercase first, then remove duplicates
unique_upper_case = set(map(lambda x: x.upper(), input_string))
print(unique_upper_case)  # Output might vary: {'A', 'B', 'C', 'D'}
```

7. **Add 5 to each item:**
```python
numbers = [1, 2, 3, 4]
added_five = list(map(lambda x: x + 5, numbers))
print(added_five)  # Output: [6, 7, 8, 9]
```

8. **Squares of items:**
```python
numbers = [1, 2, 3, 4]
squares = list(map(lambda x: x ** 2, numbers))
print(squares)  # Output: [1, 4, 9, 16]
```

9. **Greeting:**
```python
names = ['Alice', 'Bob', 'Charlie']
greeted_names = list(map(lambda name: f"Hello, {name}", names))
print(greeted_names)  # Output: ['Hello, Alice', 'Hello, Bob', 'Hello, Charlie']
```

10. **Round areas:**
```python
circle_areas = [6.56773, 9.57668, 4.00914, 56.24241, 9.01344, 32.00013]
rounded_areas = list(map(lambda x: round(x, 2), circle_areas))
print(rounded_areas)  # Output: [6.57, 9.58, 4.01, 56.24, 9.01, 32.0]
```
