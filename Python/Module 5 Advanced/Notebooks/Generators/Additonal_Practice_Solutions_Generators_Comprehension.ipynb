**10 basic-level practice problems** using **generator comprehension syntax** in Python:

---

### **1. Generate Even Numbers (0 to 10)**
   Create a generator using comprehension that yields even numbers from 0 to 10.  
   ```python
   evens = (x for x in range(11) if x % 2 == 0)
   print(list(evens))  # Output: [0, 2, 4, 6, 8, 10]
   ```

---

### **2. Generate Odd Numbers (1 to 9)**
   Create a generator comprehension for odd numbers up to 9.
   ```python
   odds = (x for x in range(10) if x % 2 != 0)
   print(list(odds))  # Output: [1, 3, 5, 7, 9]
   ```

### **3. Generate Squares of Numbers (1 to 5)**
   Create a generator that yields squares of numbers from 1 to 5.
   ```python
   squares = (x ** 2 for x in range(1, 6))
   print(list(squares))  # Output: [1, 4, 9, 16, 25]
   ```

### **4. Generate Fibonacci Numbers (First 6 Terms)**
   Create a generator comprehension to generate Fibonacci numbers up to the 6th term.
   *(Hint: Use a normal generator function, as comprehension alone wonâ€™t work for Fibonacci.)*
   ```python
   def fibonacci(n):
       a, b = 0, 1
       for _ in range(n):
           yield a
           a, b = b, a + b

   fib_gen = (num for num in fibonacci(6))
   print(list(fib_gen))  # Output: [0, 1, 1, 2, 3, 5]
   ```

   ```python
   multiples_of_3 = (x for x in range(3, 31, 3))
   ```

### **6. Generate Factorial Values (1 to 5)**
   Create a generator for factorial values from `1!` to `5!`.  
   ```python
   import math
   factorials = (math.factorial(x) for x in range(1, 6))
   print(list(factorials))  # Output: [1, 2, 6, 24, 120]
   ```

### **7. Generate Uppercase Letters from a String**
   Convert characters of `"hello"` to uppercase using generator comprehension.
   ```python
   uppercase_letters = (char.upper() for char in "hello")
   print(list(uppercase_letters))  # Output: ['H', 'E', 'L', 'L', 'O']
   ```

### **8. Generate Lengths of Words in a Sentence**
   Create a generator that yields the length of each word in `"Python is fun"`  
   ```python
   sentence = "Python is fun"
   word_lengths = (len(word) for word in sentence.split())
   print(list(word_lengths))  # Output: [6, 2, 3]
   ```

### **9. Generate First 5 Prime Numbers**
   Use a generator comprehension to extract the first 5 prime numbers from a range.
   ```python
   def is_prime(n):
       return n > 1 and all(n % i != 0 for i in range(2, int(n**0.5) + 1))

   primes = (x for x in range(2, 20) if is_prime(x))
   print([next(primes) for _ in range(5)])  # Output: [2, 3, 5, 7, 11]
   ```

### **10. Generate Reverse of Each Word**
   Reverse each word in `"hello world"` using a generator.
   ```python
   reversed_words = (word[::-1] for word in "hello world".split())
   print(list(reversed_words))  # Output: ['olleh', 'dlrow']
   ```
