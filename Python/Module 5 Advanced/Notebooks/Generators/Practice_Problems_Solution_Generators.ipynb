 **solutions** for the **10 basic generator problems**:

---

### **1. Generate Numbers from 1 to n**
```python
def count_up(n):
    for i in range(1, n + 1):
        yield i

gen = count_up(5)
print(list(gen))  # Output: [1, 2, 3, 4, 5]
```

---

### **2. Generate Even Numbers**
```python
def even_numbers(n):
    for i in range(0, n + 1, 2):
        yield i

gen = even_numbers(10)
print(list(gen))  # Output: [0, 2, 4, 6, 8, 10]
```

### **3. Generate Odd Numbers**
```python
def odd_numbers(n):
    for i in range(1, n + 1, 2):
        yield i

gen = odd_numbers(10)
print(list(gen))  # Output: [1, 3, 5, 7, 9]
```

### **4. Generate Multiples of a Given Number**
```python
def multiples_of(x, n):
    for i in range(1, n + 1):
        yield x * i

gen = multiples_of(3, 5)
print(list(gen))  # Output: [3, 6, 9, 12, 15]
```

---

### **5. Generate a Countdown**
```python
def countdown(n):
    for i in range(n, 0, -1):
        yield i

gen = countdown(5)
print(list(gen))  # Output: [5, 4, 3, 2, 1]
```

---

### **6. Generate Squares of Numbers**
```python
def squares(n):
    for i in range(n):
        yield i ** 2

gen = squares(5)
print(list(gen))  # Output: [0, 1, 4, 9, 16]
```

---

### **7. Generate a Range with Step**
```python
def custom_range(start, stop, step):
    while start < stop:
        yield start
        start += step

gen = custom_range(1, 10, 2)
print(list(gen))  # Output: [1, 3, 5, 7, 9]
```

---

### **8. Generate Letters from a Word**
```python
def letters(word):
    for letter in word:
        yield letter

gen = letters("hello")
print(list(gen))  # Output: ['h', 'e', 'l', 'l', 'o']
```

```python
def natural_numbers(n):

```

### **10. Generate Characters from a String in Reverse**
```python
def reverse_string(s):
    for char in reversed(s):
        yield char

gen = reverse_string("hello")
print(list(gen))  # Output: ['o', 'l', 'l', 'e', 'h']
```
