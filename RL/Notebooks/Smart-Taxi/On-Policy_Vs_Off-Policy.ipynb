#On-Policy_Vs_Off-Policy

 **Super Mario Bros** example to **clearly show** how **SARSA (on-policy)** and **Q-learning (off-policy)** behave differently — especially in a **stochastic, risky environment** like Mario's world.

---

##  Super Mario Bros — RL Perspective

Imagine you're training an RL agent to **navigate through a Mario level**:

*  Agent: Mario
*  Goal: Reach the flag at the end of the level
*  Environment:

  * Platforms and gaps (falling = death)
  * Enemies (touching = death)
  * Power-ups (beneficial)
  * Fireballs/enemies move with **some randomness**
* Actions: Move left, move right, jump, crouch, shoot

---

###  Scenario:

You're at a point in the level with **two options**:

1. **Jump across a pit** with **a flying Koopa** enemy above (risky but fast).
2. **Take a longer route** underground, safe but slow.

Now let's compare how **Q-learning** and **SARSA** will behave.

---

##  Q-Learning (Off-Policy)

###  Behavior:

* Updates based on:

  $$
  Q(s,a) \leftarrow r + \gamma \max_{a'} Q(s', a')
  $$
* Assumes **perfect behavior** after current step.
* Ignores the fact that the policy **sometimes takes random actions** (exploration).
* Will eventually learn that:

  * The **fast jump route** gives higher reward if done perfectly.
  * So it updates Q-values as if you **always** make the jump and avoid the Koopa.

###  Result:

* Learns to take the **risky shortcut**.
* **Fails often** during exploration (falls or hits the Koopa).
* **Optimistic** policy: focuses on best-case outcome, not on real risks.

---

##  SARSA (On-Policy)

###  Behavior:

* Updates based on:

  $$
  Q(s,a) \leftarrow r + \gamma Q(s', a')
  $$
* Learns using the **actual action** taken next (ε-greedy).
* Takes into account that:

  * During exploration, the agent might **miss jumps** or **mis-time dodges**.

###  Result:

* Learns that the **risky route is dangerous** under a realistic policy.
* Learns to take the **safer underground route**, even if it’s slower.
* **Safer, more robust** policy that avoids likely pitfalls from exploration.

---

##  Key Difference Illustrated:

| Scenario                   | Q-Learning                     | SARSA                            |
| -------------------------- | ------------------------------ | -------------------------------- |
| Random enemy above pit     | "I’ll jump! I always dodge it" | "I might mess up — better avoid" |
| Exploration risk           | Ignores it (off-policy)        | Factors it in (on-policy)        |
| Learned behavior           | Optimistic but risky           | Conservative but safe            |
| Performance in noisy world | Unstable                       | More stable                      |

---

###  Real Outcome in Training:

* **Q-learning** often results in agents that **fall a lot early on**, but eventually find the fastest path.
* **SARSA** trains agents that may reach the goal **more slowly**, but **more consistently**.

---

###  Takeaway:

> **In risky environments with randomness, SARSA behaves more like a human player — cautious and realistic.**
> **Q-learning is like a speedrunner — only works if everything goes perfectly.**
