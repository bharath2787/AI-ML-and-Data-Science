##Stacked LSTM


```python

tf.keras.backend.clear_session()

# Input layer
inputs = tf.keras.Input(shape=(20,), name="Input_Layer")
```

```python
# Embedding layer
x = tf.keras.layers.Embedding(input_dim=1000, output_dim=50, input_length=20, name="Embedding_Layer")(inputs)
print("After Embedding:", x)
```

```python
# LSTM Layer 1
x = tf.keras.layers.LSTM(200, return_sequences=True, name="LSTM_Layer_1")(x)
print("After LSTM Layer 1 (200 units, return_sequences=True):", x)
```

```python

# LSTM Layer 2
x = tf.keras.layers.LSTM(150, return_sequences=True, name="LSTM_Layer_2")(x)
print("After LSTM Layer 2 (150 units, return_sequences=True):", x)
```

```python

# Define the model
model = tf.keras.Model(inputs=inputs, outputs=x, name="Functional_LSTM_Model")
```

##Bidirectional LSTM

```python
tf.keras.backend.clear_session()

# Input layer
inputs = tf.keras.Input(shape=(20,), name="Input_Layer")
```

```python
# Embedding layer
x = tf.keras.layers.Embedding(input_dim=1000, output_dim=50, input_length=20, name="Embedding_Layer")(inputs)
print("After Embedding:", x)
```

```python
# Bidirectional LSTM 1 with merge_mode='sum'
x = tf.keras.layers.Bidirectional(
    tf.keras.layers.LSTM(200, return_sequences=True),
    merge_mode='sum',
)(x)
print("After BiLSTM Layer 1 (200 units, merge='sum'):", x)
```

```python
# Bidirectional LSTM 2 with merge_mode='concat'
x = tf.keras.layers.Bidirectional(
    tf.keras.layers.LSTM(150, return_sequences=True),
    merge_mode='concat',
)(x)
print("After BiLSTM Layer 2 (150 units, merge='concat'):", x)

```

```python
# Bidirectional LSTM 3 with merge_mode='mul'
x = tf.keras.layers.Bidirectional(
    tf.keras.layers.LSTM(100),
    merge_mode='mul',
)(x)
print("After BiLSTM Layer 3 (100 units, merge='mul'):", x)
```

```python

# Final model
model = tf.keras.Model(inputs=inputs, outputs=x, name="Functional_Bidirectional_LSTM_Model")
```

```python
# Output layer: 1 neuron + sigmoid activation
output = tf.keras.layers.Dense(1, activation='sigmoid', name="Output_Layer")(x)
```

```python
# Final model
model = tf.keras.Model(inputs=inputs, outputs=output, name="Final_BiLSTM_Model")
```
