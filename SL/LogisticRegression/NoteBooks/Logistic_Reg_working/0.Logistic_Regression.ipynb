### **1️⃣ Definition of Logistic Regression**  

* Logistic Regression is a **supervised learning algorithm** used for **binary classification** problems.
* It predicts the probability of an instance belonging to one of two classes using the **sigmoid function**.  

* Instead of predicting exact labels (0 or 1), it models the **log-odds** of an event occurring as a **linear function** of input features and then converts it into a probability.

---



### **2️⃣ Working of Logistic Regression in Simple Steps**  
 1️. Compute the **linear combination**:  

   * $[
   Z = wX + b
   ]$

2️. Convert it to a **probability** using the **sigmoid function**:  
   * $[
   p = \frac{1}{1 + e^{-Z}}
   ]$  

3️. Apply a **threshold** (default = 0.5) to classify:  
   - **If $( p \geq 0.5 )$** → Predict **Class 1**  
   - **If $( p < 0.5 )$** → Predict **Class 0**  

---


### **3️⃣ Example with Visualization**  

image.png



We take a simple **binary classification problem** where we have a decision boundary defined by the equation:  

 - $[
Z = 2X - 4
]$

 - This represents a **straight line**:  
 - $[
2X = 4  \quad \Rightarrow \quad X = 2
]$  

So, the **decision boundary** is the vertical line at **X = 2**.  
- Points on the **right** should be classified as **Class 1 (Blue)**
- Points on the **left** should be classified as **Class 0 (Red)**  

#### **🔹 Step 1: Visualizing the Decision Boundary and Two Points**
 - We consider **two test points**:  
- **Point A**: \( X = 3 \) (Right side, expected Class **1**)  
- **Point B**: \( X = 1 \) (Left side, expected Class **0**)  



Now, let’s calculate **\( Z \), \( p \) (probability), and final classification** for both points.

---

#### **🔹 Step 2: Compute \( Z \), \( p \), and Classification**
Using the decision function:  
$[
Z = 2X - 4
]$

For **Point A (X=3)**:  
- $[
Z = 2(3) - 4 = 6 - 4 = 2
]$


- $[
p = \frac{1}{1 + e^{-2}} \approx 0.881
]$
- ✔ Since \( p > 0.5 \),
- the model predicts **Class 1 (Correct ✅)**  

For **Point B (X=1)**:  
- $[
Z = 2(1) - 4 = 2 - 4 = -2
]$
- $[
p = \frac{1}{1 + e^{2}} \approx 0.119
]$
- ✔ Since \( p < 0.5 \),
- the model predicts **Class 0 (Correct ✅)**  

---

### **4️⃣ Loss Function for These Points**
Logistic Regression uses **Log-Loss (Binary Cross-Entropy):**  

- $
L = -[y \log(p) + (1 - y) \log(1 - p)]
$

For **Point A (Actual = 1, Predicted \( p = 0.881 \))**:  
- $
L = - [1 \cdot \log(0.881) + 0 \cdot \log(1 - 0.881)]
$
- $
= - \log(0.881) \approx 0.126
$
✔ **Low loss since prediction is correct.**  

For **Point B (Actual = 0, Predicted \( p = 0.119 \))**:  
- $
L = - [0 \cdot \log(0.119) + 1 \cdot \log(1 - 0.119)]
$
- $
= - \log(0.881) \approx 0.126
$
✔ **Low loss since prediction is correct.**  

---
- Now, let's see how the **loss function handles a misclassified point** near the decision boundary.

- Here’s a **misclassified point (X=2.2, Purple)** near the decision boundary.  

- Now, let's compute **\( Z \), \( p \), and loss function** for this misclassified case.  

---

#### **🔹 Step 3: Compute \( Z \), \( p \), and Misclassification**
- For **Misclassified Point (X=2.2, Actual = 0, but model predicts 1)**:  

- $[
Z = 2(2.2) - 4 = 4.4 - 4 = 0.4
]$
- $[
p = \frac{1}{1 + e^{-0.4}} \approx 0.598
]$
- ❌ Since \( p > 0.5 \), the model **incorrectly predicts Class 1** instead of **Class 0**.

---

### **5️⃣ Loss Function for Misclassified Point**
- Since the **actual class is 0, but the model predicted \( p = 0.598 \)**, the log-loss is:

- $[
L = - [0 \cdot \log(0.598) + 1 \cdot \log(1 - 0.598)]
]$
- $[
= - \log(0.402) \approx 0.911
]$

- 🚨 **High loss!**  
- ✔ The model is penalized because it confidently predicted the wrong class.  

---

### **🔹 Summary of How Loss Function Works**
| **Point** | **Actual Class** | **Predicted \( p \)** | **Loss \( L \)** | **Correct?** |
|-----------|----------------|------------------|------------|----------|
| **A (X=3, Blue)** | **1** | \( 0.881 \) | **0.126** (Low) | ✅ |
| **B (X=1, Red)** | **0** | \( 0.119 \) | **0.126** (Low) | ✅ |
| **C (X=2.2, Purple, Misclassified)** | **0** | \( 0.598 \) | **0.911** (High) | ❌ |

✔ **Correct predictions get a low loss.**  
✔ **Wrong confident predictions get a high loss, forcing the model to adjust in future updates.**  

---

### **🔹 Final Takeaways**
✅ Logistic Regression **finds a linear decision boundary** and uses sigmoid to classify points.  
✅ **Correctly classified points have low loss**, reinforcing the decision boundary.  
✅ **Misclassified points near the boundary get a high penalty**, encouraging the model to shift the boundary in training.

#**Lostistic Regression uses Gradient Descent to find the optimum values for the coeff but how does it know how many iterations to run or when to stop:**

##Logistic regression determines the number of iterations for coefficient optimization through predefined stopping criteria and hyperparameters. Here's the breakdown:

- **Optimization Stopping Mechanisms:**
Convergence Tolerance (tol)
The algorithm stops when the improvement in the cost function between iterations falls below a specified threshold (e.g., default tol=1e-4 in scikit-learn). Smaller tolerance values require more iterations to achieve precision but increase computational cost.

- **Maximum Iterations (max_iter):**
A hard limit is set to prevent infinite loops (e.g., default max_iter=100). If convergence isn’t achieved by this limit, the algorithm stops and may warn about non-convergence.
