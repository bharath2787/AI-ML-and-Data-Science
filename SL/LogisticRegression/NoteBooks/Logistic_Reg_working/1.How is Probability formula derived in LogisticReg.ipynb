# Deriving probability formula for logistic Regression


### **Why Log-Odds = Linear Combination?**  

1️. **Linear combination** ($( w_1x_1 + w_2x_2 + \dots + b )$) can take any value from **$( -\infty )$ to $( +\infty )$**.

2️. **Probability \( p \)** is always between **0 and 1**, so we can't model it directly using a linear equation.  

3️ Instead, we express probability in terms of **odds**:  
  -  $[
   \text{Odds} = \frac{p}{1 - p}
   ]$
   - Odds range from **0 to $( +\infty )$**.  

4️ To make the range **unrestricted**, we take the **log** of the odds (log-odds or logit function):  
   - $[
   \log \left( \frac{p}{1 - p} \right)
   ]$
   - Log-odds can take values from **$( -\infty )$ to $( +\infty )$**, just like a linear combination!  
5️ So, we **assume** log-odds follow a **linear relationship**:  
   $[
   \log \left( \frac{p}{1 - p} \right) = w_1x_1 + w_2x_2 + \dots + b
   ]$
6️ Rearranging this equation gives us the **sigmoid function**, which maps linear outputs back to probability:  
   $[
   p = \frac{1}{1 + e^{-(w_1x_1 + w_2x_2 + \dots + b)}}
   ]$

### **Key Takeaway**  
-  We use **log-odds** because they range from $( -\infty )$ to $( +\infty )$, just like a **linear combination**.  
- This allows us to fit a **simple linear model**, while ensuring the final output is always a **valid probability**.
