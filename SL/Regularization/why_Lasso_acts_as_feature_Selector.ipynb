- **Lasso (Blue, Dashed Line)** → The derivative is **constant** (\(\pm1\)), except at \( w = 0 \), where it jumps abruptly. This is why small coefficients are forced to **exactly zero**.  
- **Ridge (Red, Solid Line)** → The derivative is **smooth and linear** (\(2w\)), meaning small coefficients shrink slowly and never exactly reach zero.

### **Why is the Lasso Derivative Constant?**  
Lasso (L1 regularization) uses the absolute value of the weights in its penalty term:  
$[
\lambda \sum |w_i|
$]
The key part here is **$(|w_i|)$**. To find the derivative, let's analyze it in three cases:  

#### **1. When $( w > 0 )$:**
$[
\frac{d}{dw} |w| = \frac{d}{dw} w = 1
]$
The function behaves normally like \( w \), so the derivative is **+1**.  

#### **2. When $( w < 0 )$:**
$[
\frac{d}{dw} |w| = \frac{d}{dw} (-w) = -1
]$
Since absolute values make negative numbers positive, we treat \( w \) as \( -w \), so the derivative is **-1**.  

#### **3. When $( w = 0 )$:**
At exactly zero, $(|w|)$ is **not differentiable**. Instead, we use the **subgradient**, which means:  
- The derivative can be **any value between \([-1, 1]\)**.  
- In optimization, this allows the algorithm to **set the coefficient exactly to zero** if needed.  

### **What This Means for Gradient Descent**
- **Lasso always subtracts a fixed amount ($(\lambda)$) from the gradient, no matter how small $( w )$ is.**  
- This is why small coefficients eventually reach zero.  
- In contrast, Ridge ($(2w)$) smoothly shrinks values without hitting zero.
