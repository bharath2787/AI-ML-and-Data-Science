# how data can be easily sepearted by transforming it to higher dimensions

```python
# Step 1: Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Step 2: Generate 1D data
# Generate 20 points from -5 to 5
X_1D = np.linspace(-5, 5, 20)

# Assign classes: first 5 and last 5 points to Class 1, middle 10 points to Class 2
y = np.array([1] * 5 + [0] * 10 + [1] * 5)

# Step 3: Visualize the original 1D data
plt.figure(figsize=(8, 2))
plt.scatter(X_1D, np.zeros_like(X_1D), c=y, cmap='bwr', edgecolors='k')
plt.title("Original 1D Data")
plt.xlabel("X1")
plt.yticks([])
plt.grid(True)
plt.show()

# Step 4: Apply a transformation to map data to 2D
# We'll map the data to a 2D space by adding a second feature, X2 = X1^2
X_2D = np.c_[X_1D, X_1D**2]

# Step 5: Visualize the transformed 2D data
plt.figure(figsize=(8, 6))
plt.scatter(X_2D[:, 0], X_2D[:, 1], c=y, cmap='bwr', edgecolors='k')
plt.title("Transformed 2D Data (Linearly Separable)")
plt.xlabel("X1")
plt.ylabel("X2 (X1^2)")
plt.axhline(y=7, color='k', linestyle='--', label="Linear Decision Boundary")
plt.legend()
plt.grid(True)
plt.show()

# Step 6: Explanation
# In the 1D data, Class 1 points are at the extremes, and Class 2 points are in the middle,
# making it non-linearly separable by a simple threshold. By transforming the data to 2D using X2 = X1^2,
# we create a parabola that makes the data linearly separable. A linear decision boundary in the 2D space
# can now easily separate the two classes.
```
