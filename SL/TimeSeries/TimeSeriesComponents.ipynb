```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create a date range
dates = pd.date_range(start='2020-01-01', periods=100)

# Generate a trend: linearly increasing values with some random noise
trend = np.linspace(start=10, stop=50, num=100)
noise = np.random.normal(loc=0, scale=2, size=100)
data = trend + noise

# Create a time series DataFrame
df = pd.DataFrame({'Date': dates, 'Value': data})
df.set_index('Date', inplace=True)

# Plot the time series
plt.figure(figsize=(10, 5))
plt.plot(df, label='Time Series with Trend')
plt.title('Example of a Trend in Time Series')
plt.xlabel('Date')
plt.ylabel('Value')
plt.grid(True)
plt.legend()
plt.show()
```

##Seasonality:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create date range
dates = pd.date_range(start='2020-01-01', periods=100)

# Simulate seasonality (e.g., sine wave)
seasonality = 10 * np.sin(2 * np.pi * dates.dayofyear / 30)  # Monthly seasonality

# Combine with some noise
noise = np.random.normal(loc=0, scale=1.5, size=100)
data = seasonality + noise

# Create DataFrame
df = pd.DataFrame({'Date': dates, 'Value': data})
df.set_index('Date', inplace=True)

# Plot
plt.figure(figsize=(10, 5))
plt.plot(df, label='Seasonal Pattern')
plt.title('Example of Seasonality in Time Series')
plt.xlabel('Date')
plt.ylabel('Value')
plt.grid(True)
plt.legend()
plt.show()
```

##Cyclic:

```python
# Simulate a slow-changing cyclic component (longer period sine wave)
cycle = 5 * np.sin(2 * np.pi * dates.dayofyear / 90)  # Not as regular as seasonality

# Add trend and noise
trend = np.linspace(20, 40, 100)
noise = np.random.normal(0, 1, 100)
data = trend + cycle + noise

# Create DataFrame
df = pd.DataFrame({'Date': dates, 'Value': data})
df.set_index('Date', inplace=True)

# Plot
plt.figure(figsize=(10, 5))
plt.plot(df, label='Cyclic Pattern + Trend')
plt.title('Example of a Cyclic Pattern in Time Series')
plt.xlabel('Date')
plt.ylabel('Value')
plt.grid(True)
plt.legend()
plt.show()
```
