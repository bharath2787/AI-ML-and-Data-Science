##  What Does INNER JOIN Do?

> An `INNER JOIN` returns **only those records that have matching values in both tables** based on the specified condition.

---

##  Example Use Case:

Letâ€™s say a customer placed an order that includes a product.
You want to **see the product details** along with **the quantity** ordered:

```sql
SELECT
    oi.order_id,
    p.name AS product_name,
    oi.quantity
FROM order_items oi
INNER JOIN products p ON oi.product_id = p.product_id;
```

---

###  1. **Basic INNER JOIN Using `*` (All Columns)**

```sql
SELECT *
FROM orders o
INNER JOIN order_items oi ON o.order_id = oi.order_id;
```

** Output:** Returns all matching records from `orders` and `order_items` based on `order_id`.

---

###  2. **INNER JOIN with Specific Columns**

```sql
SELECT
    o.order_id,
    o.customer_id,
    oi.product_id,
    oi.quantity
FROM orders o
INNER JOIN order_items oi ON o.order_id = oi.order_id;
```

** Output:** Returns only the specified columns from both tables.

---

###  3. **INNER JOIN with Prefixes for Disambiguation**

```sql
SELECT
    orders.order_id,
    order_items.product_id
FROM orders
INNER JOIN order_items ON orders.order_id = order_items.order_id;
```

** Note:** Useful when **not using aliases** or when readability is preferred.

---

###  4. **INNER JOIN with Table Aliases**

```sql
SELECT
    o.order_id,
    oi.product_id,
    oi.quantity
FROM orders AS o
INNER JOIN order_items AS oi ON o.order_id = oi.order_id;
```

** Benefit:** Makes the query concise and readable, especially when using multiple joins.

---

###  5. **INNER JOIN Across Two Databases**

Assume:

* Table `order_items` is in DB `store`
* Table `products` is in DB `inventory`

```sql
SELECT
    oi.product_id,
    p.name AS product_name,
    oi.quantity
FROM store.order_items AS oi
INNER JOIN inventory.products AS p ON oi.product_id = p.product_id;
```

** Output:** Returns product name and quantity for matching product IDs across two DBs.

** Requirement:** Both DBs must be on the same MySQL server, and you must have access to both.
