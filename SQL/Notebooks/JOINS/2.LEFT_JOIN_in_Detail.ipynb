##  SQL **LEFT JOIN**

image.png




###  What is a LEFT JOIN?

A **LEFT JOIN** returns **all records from the left table** (the first one in the query), and the **matched records from the right table** (the second one).
If there is **no match**, the result is `NULL` on the right side.

---

###  Syntax:

```sql
SELECT columns
FROM left_table
LEFT JOIN right_table
ON left_table.common_column = right_table.common_column;
```

Or with aliases:

```sql
SELECT columns
FROM left_table AS l
LEFT JOIN right_table AS r
ON l.common_column = r.common_column;
```

---

###  1. LEFT JOIN Using `*` (All Columns)

```sql
SELECT *
FROM orders o
LEFT JOIN order_items oi
ON o.order_id = oi.order_id;
```

 **Explanation**:
Returns all orders. If an order has no matching item, `order_items` columns will be `NULL`.

---

###  2. LEFT JOIN with Specific Columns

```sql
SELECT
    o.order_id,
    o.customer_id,
    oi.product_id,
    oi.quantity
FROM orders o
LEFT JOIN order_items oi
ON o.order_id = oi.order_id;
```

 **Explanation**:
Shows order details along with their item info, if any.

---

###  3. LEFT JOIN With Aliases and Prefixed Columns

```sql
SELECT
    o.order_id AS order_id,
    o.customer_id,
    oi.product_id,
    oi.quantity
FROM orders AS o
LEFT JOIN order_items AS oi
ON o.order_id = oi.order_id;
```

 **Explanation**:
Same result, but clearer with table aliases. Use `AS` to rename or disambiguate columns.
