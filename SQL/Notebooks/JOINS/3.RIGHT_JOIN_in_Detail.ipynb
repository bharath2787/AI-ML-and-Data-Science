###  What is a RIGHT JOIN?

A **RIGHT JOIN** returns **all records from the right table**, and the **matched records from the left table**.
If there is **no match**, the result is `NULL` on the left side.

---

###  Syntax:

```sql
SELECT columns
FROM left_table
RIGHT JOIN right_table
ON left_table.common_column = right_table.common_column;
```

Or with aliases:

```sql
SELECT columns
FROM left_table AS l
RIGHT JOIN right_table AS r
ON l.common_column = r.common_column;
```

---

###  1. RIGHT JOIN Using `*` (All Columns)

```sql
SELECT *
FROM orders o
RIGHT JOIN order_items oi
ON o.order_id = oi.order_id;
```

 **Explanation**:
Returns all rows from `order_items`. If an item does not belong to any order (hypothetically), it would show `NULL` in order columns.

---

###  2. RIGHT JOIN with Specific Columns

```sql
SELECT
    o.order_id,
    o.customer_id,
    oi.product_id,
    oi.quantity
FROM orders o
RIGHT JOIN order_items oi
ON o.order_id = oi.order_id;
```

 **Explanation**:
Useful when you're primarily interested in `order_items`, but want to include order info if available.
