##  What is a CROSS JOIN?

A **CROSS JOIN** returns the **Cartesian product** of two tables — it combines **every row from the first table** with **every row from the second table**.

A Cartesian Product (also called a Cross Join) occurs when you join every row of one table with every row of another table, without any join condition.

>  If Table A has `m` rows and Table B has `n` rows, the result of a `CROSS JOIN` will have `m * n` rows.

---

##  Syntax:

```sql
SELECT *
FROM table1
CROSS JOIN table2;
```

or (without the keyword):

```sql
SELECT *
FROM table1, table2;
```

Both yield the same result — but `CROSS JOIN` is more explicit and readable.

---

##  Example using our `sql_hr` database:

### Tables:

* `employees` (let’s say it has 5 rows)
* `offices` (let’s say it has 3 rows)

### Query:

```sql
SELECT e.first_name, e.last_name, o.city, o.state
FROM employees e
CROSS JOIN offices o;
```

###  What happens:

Each employee will be paired with **every office**.
So if there are 5 employees and 3 offices, the result will have **5 × 3 = 15 rows**.

---

##  When to Use CROSS JOIN?

* To **generate combinations** of two sets (like pairing all products with all customers)
* Useful in **test scenarios**, **matrix calculations**, or **pivot table-like reports**

---

##  Note:

* **No condition or ON clause** is used in a CROSS JOIN.
* Be cautious on large datasets — the result size grows rapidly!
