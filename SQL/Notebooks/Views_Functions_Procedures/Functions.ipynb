##  What is a Function in SQL?

A **Function** is a stored block of SQL code that:

* Accepts **input parameters**,
* Returns a **single value** (must return),
* Can be used **inside SELECT statements, WHERE clauses, etc.**

---

##  Syntax:

```sql
DELIMITER $$

CREATE FUNCTION function_name(parameter datatype)
RETURNS return_datatype
DETERMINISTIC
BEGIN
    -- Logic
    RETURN value;
END $$

DELIMITER ;
```

>  `DETERMINISTIC` means given the same input, it always returns the same output.

---

##  Examples using `sql_hr` database:

---

###  1. Function to get full name of an employee

```sql
DELIMITER $$

CREATE FUNCTION get_full_name(emp_id INT)
RETURNS VARCHAR(100)
DETERMINISTIC
BEGIN
    DECLARE full_name VARCHAR(100);
    SELECT CONCAT(first_name, ' ', last_name)
    INTO full_name
    FROM employees
    WHERE employee_id = emp_id;
    RETURN full_name;
END $$

DELIMITER ;
```

 **Usage:**

```sql
SELECT get_full_name(33391);
```

---

###  2. Function to calculate yearly salary

```sql
DELIMITER $$

CREATE FUNCTION yearly_salary(monthly_salary INT)
RETURNS INT
DETERMINISTIC
BEGIN
    RETURN monthly_salary * 12;
END $$

DELIMITER ;
```

 **Usage:**

```sql
SELECT first_name, last_name, yearly_salary(salary) AS annual_income
FROM employees;
```

---

###  3. Function to check if an employee is a manager

```sql
DELIMITER $$

CREATE FUNCTION is_manager(emp_id INT)
RETURNS VARCHAR(10)
DETERMINISTIC
BEGIN
    DECLARE count_reports INT;
    SELECT COUNT(*) INTO count_reports
    FROM employees
    WHERE reports_to = emp_id;

    IF count_reports > 0 THEN
        RETURN 'Yes';
    ELSE
        RETURN 'No';
    END IF;
END $$

DELIMITER ;
```

 **Usage:**

```sql
SELECT employee_id, first_name, is_manager(employee_id) AS manager_status
FROM employees;
```
