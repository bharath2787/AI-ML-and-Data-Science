### **Why Do We Need Three Measures of Central Tendency?**  

Each measure—**mean, median, and mode**—serves a unique purpose depending on the nature of the data and its distribution.  

1. **Mean (Average) – Best for Symmetric Data**  
   - Uses all data points, making it the most reliable for **normally distributed** data.  
   - **Example**: Average salary, test scores, or stock market trends.  
   - **Limitation**: Affected by outliers (e.g., a few very high incomes can skew the mean salary).  

2. **Median – Best for Skewed Data & Outliers**  
   - The middle value, making it robust against **outliers and skewed distributions**.  
   - **Example**: Median income (avoids distortion from billionaires), median house prices.  
   - **Limitation**: Ignores specific values, so it may not reflect the full dataset accurately.  

3. **Mode – Best for Categorical or Repeated Values**  
   - Identifies the most frequently occurring value, useful for **nominal data** (categories).  
   - **Example**: Most common exam grade, most sold product size, or favorite ice cream flavor.  
   - **Limitation**: Sometimes, there’s no mode or multiple modes, making interpretation tricky.  

### **Why Not Just Use One?**  
- **Different distributions require different measures** (e.g., for skewed income data, the mean is misleading, but the median is useful).  
- **Together, they provide a fuller picture** of the data, helping in better decision-making.
