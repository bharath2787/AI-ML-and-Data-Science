#STANDARD DEVIATION SIGNIFICANCE

```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#There are two sales person A and B with following sales figures

```python
# Initial sales data for Salesperson A and B
sales_A_initial = [9, 110, 50, 70, 30, 90, 60, 80, 20, 100]
sales_B_initial = [20, 22, 23, 24, 25, 26, 27, 28, 29, 30]
```

```python

# Create a DataFrame for the initial sales data
df_initial = pd.DataFrame({
    'Salesperson A': sales_A_initial,
    'Salesperson B': sales_B_initial
})
```

```python
# Calculate mean and standard deviation for the initial data
mean_A_initial = np.mean(sales_A_initial)
std_A_initial = np.std(sales_A_initial)
mean_B_initial = np.mean(sales_B_initial)
std_B_initial = np.std(sales_B_initial)


```

```python
# Display the initial sales data, mean, and standard deviation
print("Initial Sales Data")
display(df_initial)
print("\nStatistics for Initial Sales Data")
print("MEAN VALUES:")
print(f"Mean Sales - Salesperson A: {mean_A_initial}")
print(f"Min Sales - Salesperson A: {np.min(sales_A_initial)}")
print(f"Max Sales - Salesperson A: {np.max(sales_A_initial)}")
print(f"Mean Sales - Salesperson B: {mean_B_initial}")
print(f"Min Sales - Salesperson B: {np.min(sales_B_initial)}")
print(f"Max Sales - Salesperson B: {np.max(sales_B_initial)}")
print("STANDARD DEVIATION:")
print(f"Standard Deviation - Salesperson A: {std_A_initial}")
print(f"Standard Deviation - Salesperson B: {std_B_initial}")

```

```python
# Plot the initial sales data
plt.figure(figsize=(10, 5))
plt.plot(sales_A_initial, label='Salesperson A', marker='o')
plt.plot(sales_B_initial, label='Salesperson B', marker='o')
plt.title('Initial Sales Data')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
# New sales data for the current month where both performed their worst
sales_A_new = [10, 10, 10, 10, 10, 10, 10, 10, 10, 10]
sales_B_new = [15, 15, 15, 15, 15, 15, 15, 15, 15, 15]

```

```python

# Create a DataFrame for the new sales data
df_new = pd.DataFrame({
    'Salesperson A': sales_A_new,
    'Salesperson B': sales_B_new
})
```

```python
# Calculate mean and standard deviation for the new data
mean_A_new = np.mean(sales_A_new)
std_A_new = np.std(sales_A_new)
mean_B_new = np.mean(sales_B_new)
std_B_new = np.std(sales_B_new)

```

```python
# Display the new sales data, mean, and standard deviation
print("\nNew Sales Data (Worst Performance)")
print(df_new)
print("\nStatistics for New Sales Data")
print("MEAN VALUES:")
print(f"Mean Sales - Salesperson A: {mean_A_new}")

print(f"Mean Sales - Salesperson B: {mean_B_new}")
print("STANDARD DEVIATION:")
print(f"Standard Deviation - Salesperson A: {std_A_new}")
print(f"Standard Deviation - Salesperson B: {std_B_new}")

```

```python
# Combined initial and new sales data
sales_A_combined = sales_A_initial + sales_A_new
sales_B_combined = sales_B_initial + sales_B_new

```

```python

# Calculate mean and standard deviation for the combined data
mean_A_combined = np.mean(sales_A_combined)
std_A_combined = np.std(sales_A_combined)
mean_B_combined = np.mean(sales_B_combined)
std_B_combined = np.std(sales_B_combined)

```

```python
# Display the combined sales data, mean, and standard deviation
print("\nCombined Sales Data")
print("MEAN VALUES:")
print(f"Mean Sales - Salesperson A: {mean_A_combined}")

print(f"Mean Sales - Salesperson B: {mean_B_combined}")
print("STANDARD DEVIATION:")
print(f"Standard Deviation - Salesperson A: {std_A_combined}")
print(f"Standard Deviation - Salesperson B: {std_B_combined}")

```

```python
# Plot the combined sales data
plt.figure(figsize=(10, 5))
plt.plot(sales_A_combined, label='Salesperson A', marker='o')
plt.plot(sales_B_combined, label='Salesperson B', marker='o')
plt.title('Combined Sales Data')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()
plt.show()
```
