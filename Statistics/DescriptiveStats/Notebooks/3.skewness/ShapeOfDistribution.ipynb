Understanding the shape of a distribution is crucial for interpreting data correctly. Skewness and kurtosis are two important measures that describe the shape of a distribution.

### Shape of Distribution

#### Skewness
Skewness measures the asymmetry of the distribution. It tells us whether the data points are skewed to the left or right of the mean.

- **Positive Skew (Right Skew)**:
  - The right tail (larger values) is longer or fatter than the left tail.
  - Most of the data points are concentrated on the left.
  - Mean > Median > Mode.
  - Example: Income distribution, where a small number of people earn significantly more than the majority.


- **Negative Skew (Left Skew)**:
  - The left tail (smaller values) is longer or fatter than the right tail.
  - Most of the data points are concentrated on the right.
  - Mean < Median < Mode.
  - Example: Age at retirement, where most people retire at a common age but some retire significantly earlier.



### Detailed Explanation

#### Skewness Formula
Skewness can be calculated using the formula:

$[ \text{Skewness} = \frac{n}{(n-1)(n-2)} \sum_{i=1}^{n} \left( \frac{x_i - \bar{x}}{s} \right)^3 ]$

where:
- \( n \) is the number of data points.
- $( x_i )$ is each individual data point.
- $( \bar{x} )$ is the mean of the data points.
- \( s \) is the standard deviation of the data points.


### Practical Interpretation

- **Skewness**:
  - A skewness close to 0 indicates a symmetrical distribution.
  - Positive skewness indicates more values are concentrated on the left with longer right tails.
  - Negative skewness indicates more values are concentrated on the right with longer left tails.

- **Kurtosis**:
  - A kurtosis close to 0 (mesokurtic) indicates a normal distribution shape.
  - Positive kurtosis (leptokurtic) indicates a distribution with heavy tails and a sharp peak.
  - Negative kurtosis (platykurtic) indicates a distribution with light tails and a flatter peak.

### Example with a Dataset
Letâ€™s consider a small dataset: $([2, 3, 3, 5, 7, 10])$.

1. **Calculate Mean**:
   $[ \bar{x} = \frac{2 + 3 + 3 + 5 + 7 + 10}{6} = 5 ]$

2. **Calculate Standard Deviation**:
   $[ s = \sqrt{\frac{(2-5)^2 + (3-5)^2 + (3-5)^2 + (5-5)^2 + (7-5)^2 + (10-5)^2}{6-1}} \approx 2.92 ]$

3. **Calculate Skewness**:
   $[ \text{Skewness} \approx \frac{6}{5 \times 4} \left( \frac{(2-5)^3 + (3-5)^3 + (3-5)^3 + (5-5)^3 + (7-5)^3 + (10-5)^3}{2.92^3} \right) \approx 0.77 ]$
   - This indicates a moderately positive skew.

4. **Calculate Kurtosis**:
   $[ \text{Kurtosis} \approx \frac{6 \times 7}{5 \times 4 \times 3} \left( \frac{(2-5)^4 + (3-5)^4 + (3-5)^4 + (5-5)^4 + (7-5)^4 + (10-5)^4}{2.92^4} \right) - \frac{3 \times 4}{2} \approx -0.65 ]$
   - This indicates a platykurtic distribution.

By understanding and calculating skewness and kurtosis, you gain deeper insights into the shape and characteristics of your data distribution, which can help in making more informed decisions and analyses.

# Understanding Skewness with the Seaborn Tips Dataset

## Introduction
### In this notebook, we'll explore the concept of skewness using the 'tips' dataset from the seaborn library. Skewness measures the asymmetry of the distribution of values in a dataset. We'll calculate and visualize the skewness of different numerical columns in the dataset.

### Let's start by importing the necessary libraries.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import skew
```

```python
# Load the 'tips' dataset
tips = sns.load_dataset('tips')
```

```python
# Distribution of the 'total_bill' column
plt.figure(figsize=(10, 6))
sns.histplot(tips['total_bill'], kde=True)
plt.title('Distribution of Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.show()

# Calculate skewness for 'total_bill'
total_bill_skewness = skew(tips['total_bill'])
print(f'Skewness of Total Bill: {total_bill_skewness:.2f}')
```

```python
# Load the tips dataset
tips = sns.load_dataset('tips')

# Create a figure with subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Plot histogram of total_bill
sns.histplot(tips['total_bill'], kde=True, ax=axes[0])
axes[0].set_title('Histogram of Total Bill')
axes[0].set_xlabel('Total Bill')
axes[0].set_ylabel('Frequency')

# Plot box plot of total_bill
sns.boxplot(x=tips['total_bill'], ax=axes[1])
axes[1].set_title('Box Plot of Total Bill')
axes[1].set_xlabel('Total Bill')

# Show plots
plt.tight_layout()
plt.show()
```

```python
# Calculate mean and median of total_bill
mean_total_bill = tips['total_bill'].mean()
median_total_bill = tips['total_bill'].median()

# Create a figure with subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Plot histogram of total_bill
sns.histplot(tips['total_bill'], kde=True, ax=axes[0])
axes[0].axvline(mean_total_bill, color='r', linestyle='--', label=f'Mean: {mean_total_bill:.2f}')
axes[0].axvline(median_total_bill, color='g', linestyle='-', label=f'Median: {median_total_bill:.2f}')
axes[0].set_title('Histogram of Total Bill')
axes[0].set_xlabel('Total Bill')
axes[0].set_ylabel('Frequency')
axes[0].legend()

# Plot box plot of total_bill
sns.boxplot(x=tips['total_bill'], ax=axes[1])
axes[1].axvline(mean_total_bill, color='r', linestyle='--', label=f'Mean: {mean_total_bill:.2f}')
axes[1].axvline(median_total_bill, color='g', linestyle='-', label=f'Median: {median_total_bill:.2f}')
axes[1].set_title('Box Plot of Total Bill')
axes[1].set_xlabel('Total Bill')
axes[1].legend()

# Show plots
plt.tight_layout()
plt.show()
```
