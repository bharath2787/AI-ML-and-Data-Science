```python
# Skewness Analysis with the California Housing Dataset

## Introduction
# In this notebook, we'll explore the concept of skewness using the California Housing dataset. Skewness measures the asymmetry of the distribution of values in a dataset. We'll calculate and visualize the skewness of different numerical columns in the dataset and provide comments on the skewness, mean, and median.

# Let's start by importing the necessary libraries.

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import skew
from sklearn.datasets import fetch_california_housing

# Load the California Housing dataset
housing = fetch_california_housing(as_frame=True)
data = housing.frame

# Display the first few rows of the dataset
data.head()

# Brief summary of the dataset
data.describe()

# Function to plot distribution, boxplot, and calculate skewness
def plot_skewness(data, column):
    fig, axes = plt.subplots(1, 2, figsize=(18, 6))

    # Histogram with KDE
    sns.histplot(data[column], kde=True, ax=axes[0])
    axes[0].axvline(data[column].mean(), color='r', linestyle='--', label=f'Mean: {data[column].mean():.2f}')
    axes[0].axvline(data[column].median(), color='g', linestyle='-', label=f'Median: {data[column].median():.2f}')
    axes[0].set_title(f'Distribution of {column}')
    axes[0].set_xlabel(column)
    axes[0].set_ylabel('Frequency')
    axes[0].legend()

    # Boxplot
    sns.boxplot(x=data[column], ax=axes[1])
    axes[1].set_title(f'Box Plot of {column}')
    axes[1].set_xlabel(column)

    plt.tight_layout()
    plt.show()

    skewness = skew(data[column])
    mean = data[column].mean()
    median = data[column].median()
    print(f'Skewness of {column}: {skewness:.2f}')
    print(f'Mean of {column}: {mean:.2f}')
    print(f'Median of {column}: {median:.2f}')
    if skewness > 0:
        print(f'The {column} data is positively skewed, with a longer right tail.')
    elif skewness < 0:
        print(f'The {column} data is negatively skewed, with a longer left tail.')
    else:
        print(f'The {column} data is approximately symmetric.')
    print("-----------------------------------------------------\n")

# Plot and analyze the skewness for each feature in the dataset
for column in data.columns:
    plot_skewness(data, column)

# Conclusion
# In this notebook, we calculated and visualized the skewness of each feature in the California Housing dataset using histograms and boxplots. Skewness helps us understand the asymmetry of the data distribution. Positive skewness indicates a longer tail on the right side, while negative skewness indicates a longer tail on the left side.
```

```python
# Skewness Analysis with the Titanic Dataset

## Introduction
# In this notebook, we'll explore the concept of skewness using the Titanic dataset. Skewness measures the asymmetry of the distribution of values in a dataset. We'll calculate and visualize the skewness of different numerical columns in the dataset and provide comments on the skewness, mean, and median.

# Let's start by importing the necessary libraries.

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import skew

# Load the Titanic dataset
titanic = sns.load_dataset('titanic')

# Display the first few rows of the dataset
titanic.head()

# Brief summary of the dataset
titanic.describe()

# Function to plot distribution, boxplot, and calculate skewness
def plot_skewness(data, column):
    fig, axes = plt.subplots(1, 2, figsize=(18, 6))

    # Histogram with KDE
    sns.histplot(data[column].dropna(), kde=True, ax=axes[0])
    axes[0].axvline(data[column].mean(), color='r', linestyle='--', label=f'Mean: {data[column].mean():.2f}')
    axes[0].axvline(data[column].median(), color='g', linestyle='-', label=f'Median: {data[column].median():.2f}')
    axes[0].set_title(f'Distribution of {column}')
    axes[0].set_xlabel(column)
    axes[0].set_ylabel('Frequency')
    axes[0].legend()

    # Boxplot
    sns.boxplot(x=data[column].dropna(), ax=axes[1])
    axes[1].set_title(f'Box Plot of {column}')
    axes[1].set_xlabel(column)

    plt.tight_layout()
    plt.show()

    skewness = skew(data[column].dropna())
    mean = data[column].mean()
    median = data[column].median()
    print(f'Skewness of {column}: {skewness:.2f}')
    print(f'Mean of {column}: {mean:.2f}')
    print(f'Median of {column}: {median:.2f}')
    if skewness > 0:
        print(f'The {column} data is positively skewed, with a longer right tail.')
    elif skewness < 0:
        print(f'The {column} data is negatively skewed, with a longer left tail.')
    else:
        print(f'The {column} data is approximately symmetric.')
    print("-----------------------------------------------------\n")

# Select numerical columns
numerical_columns = ['age', 'fare', 'sibsp', 'parch']

# Plot and analyze the skewness for each numerical feature in the dataset
for column in numerical_columns:
    plot_skewness(titanic, column)

# Conclusion
# In this notebook, we calculated and visualized the skewness of each numerical feature in the Titanic dataset using histograms and boxplots. Skewness helps us understand the asymmetry of the data distribution. Positive skewness indicates a longer tail on the right side, while negative skewness indicates a longer tail on the left side.
```
