### **Spearman’s vs. Pearson’s Correlation**  

| Feature         | **Spearman’s Correlation ($(\rho)$)** | **Pearson’s Correlation ($(r)$)** |  
|---------------|------------------|---------------|  
| **Measures**     | Strength & direction of a **monotonic** relationship | Strength & direction of a **linear** relationship |  
| **Data Type**    | Ordinal, skewed, or non-linear data | Continuous, normally distributed data |  
| **Assumptions**  | No normality required | Assumes data is normally distributed |  
| **Ranking-Based?** | Yes, uses ranks | No, uses actual values |  
| **Sensitivity to Outliers** | Less sensitive (ranks reduce outlier impact) | Highly sensitive (outliers can distort results) |  
| **Interpretation Range** | \(-1\) to \(+1\) | \(-1\) to \(+1\) |  

---

### **Key Differences & Use Cases**  

1. **Pearson’s Correlation ($(r)$)**  
   - Measures **linear relationships** (requires a straight-line trend).  
   - Affected by **outliers**.  
   - **Example**: Relationship between **height and weight** (normally distributed data).  

2. **Spearman’s Correlation ($(\rho)$)**  
   - Measures **monotonic relationships** (works even if the trend is curved).  
   - Uses **ranks**, reducing the effect of **outliers**.  
   - **Example**: Relationship between **exam ranks and study hours** (not necessarily linear).  

---

### **Example Comparison**  

#### **Case 1: Linear Relationship (Use Pearson’s)**  
| X (Study Hours) | Y (Exam Score) |  
|---------------|-------------|  
| 2             | 50          |  
| 4             | 60          |  
| 6             | 70          |  
| 8             | 80          |  

- **Pearson’s $( r = 1.0 )$** (Perfect linear correlation)  
- **Spearman’s $( \rho = 1.0 )$** (Ranks also match)  

#### **Case 2: Non-Linear Monotonic Relationship (Use Spearman’s)**  
| X (Ad Spend) | Y (Sales) |  
|-------------|---------|  
| 100         | 10      |  
| 200         | 25      |  
| 300         | 40      |  
| 400         | 50      |  
| 500         | 52      |  

Here, **sales increase but slow down at higher ad spends**, forming a **curved trend**.  
- **Pearson’s $( r )$ will be low** (not linear).  
- **Spearman’s $( \rho )$ will be high** (consistent monotonic trend).  

---

### **When to Use Which?**  
✅ Use **Pearson’s** when the relationship is **linear** and data is **normally distributed**.  
✅ Use **Spearman’s** when the relationship is **monotonic but not linear** or when data contains **outliers or ranks**.
