```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Create a DataFrame to demonstrate correlation unaffected by scale
data = {
    'X': [1, 2, 3, 4, 5],
    'Y': [10, 20, 30, 40, 50],  # Y is simply X multiplied by 10
    'Z': [100, 200, 300, 400, 500] , # Z is simply X multiplied by 100
    'U': [33,-34,0,21,99]
}
```

```python

df = pd.DataFrame(data)

# Display the DataFrame
print("DataFrame:")
print(df)
```

```python

# Calculate the correlation matrix
correlation_matrix = df.corr()

# Display the correlation matrix
print("\nCorrelation Matrix:")
print(correlation_matrix)
```

## Covariance value differes for different scales while correlation doesnt

```python
# Calculate the covariance matrix
covariance_matrix = df.cov()

# Display the covariance matrix
print("\nCovariance Matrix:")
print(covariance_matrix)
```

```python
# Plot the data
plt.figure(figsize=(12, 6))

# Plot X vs Y
plt.subplot(1, 2, 1)
plt.scatter(df['X'], df['Y'])
plt.title('X vs Y')
plt.xlabel('X')
plt.ylabel('Y')
```

```python

# Plot X vs Z
plt.subplot(1, 2, 2)
plt.scatter(df['X'], df['Z'])
plt.title('X vs Z')
plt.xlabel('X')
plt.ylabel('Z')

plt.tight_layout()
plt.show()
```

```python

# Heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(), annot=True, linewidths=0.5)
plt.title('Heatmap of the Correlation Matrix')
plt.show()
```




# Observations:
### The correlation matrix shows that X, Y, and Z have perfect positive correlations (correlation coefficient = 1.0).
### This demonstrates that correlation is not affected by the scale of the variables.
### Despite Y and Z being scaled versions of X, the correlation remains the same.
