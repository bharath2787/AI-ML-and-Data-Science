```python
import pandas as pd
import numpy as np

# Define distances in km and meters
km = [1, 2, 3, 4, 5]
m = [1000, 2000, 3000, 4000, 5000]

# Create a DataFrame
df = pd.DataFrame({"Distance_km": km, "Distance_m": m})
```

```python
df
```

```python

# Compute covariance and correlation
covariance = df.cov().iloc[0, 1]  # Covariance between km and meters
correlation = df.corr().iloc[0, 1]  # Correlation between km and meters

# Display results
print("Covariance:", covariance)
print("Correlation:", correlation)
```
