# Step 5: Multivariate Analysis

````python
# Pairplot for numerical variables
sns.pairplot(data, hue='Product')
plt.show()
````

````python
# Observation:
# - TM798 stands out with customers having higher income and mileage.
````

````python

# Observations:
# - The diagonal of the pairplot displays the KDE plots of individual variables (Age, Income, and Miles) for each product.
#
# - **Age Distribution (KDE plot along the diagonal)**:
#   - **TM195**: The distribution is slightly skewed to the younger age group, with most customers being in their 20s and 30s.
#   - **TM498**: The age distribution is more spread out, showing a broader range of ages, but still with a higher concentration in the 30-50 age range.
#   - **TM798**: This product's customers tend to be older, with the peak of the KDE shifted to the 40-60 age range.
#
# - **Income Distribution (KDE plot along the diagonal)**:
#   - **TM195**: The income distribution is more concentrated in the lower income brackets, with a peak around $40,000-$60,000.
#   - **TM498**: The income distribution is more spread, with a noticeable concentration around $60,000-$80,000.
#   - **TM798**: The income distribution shows a clear peak around the higher income bracket ($80,000+), indicating that TM798 appeals to wealthier customers.
#
# - **Miles Distribution (KDE plot along the diagonal)**:
#   - **TM195**: Customers generally plan to walk/run fewer miles, with a peak at lower values (< 30 miles).
#   - **TM498**: The distribution is more spread out, but still peaks at lower mileages, suggesting moderate use.
#   - **TM798**: Customers plan to use their treadmill significantly more, with a peak at higher mileage values (50+ miles), indicating heavier usage.
#
# - **Summary**:
#   - **TM798** customers generally have higher income, older age, and plan to run more miles.
#   - **TM195** customers are younger, have lower income, and plan to run fewer miles.
#   - **TM498** customers represent a middle ground in terms of income, age, and planned mileage.
````

````python
# Observations:
# - The pairplot shows relationships between the numerical variables: Age, Income, and Miles.
# - **Age vs. Income**: There is a moderate positive correlation, particularly for TM798 users who are older and have higher incomes.
# - **Age vs. Miles**: TM195 customers tend to be younger and have lower planned mileage, while TM798 customers are older and plan to run more miles.
# - **Income vs. Miles**: TM798 users appear to have the highest income and mileage plans, while TM195 customers have the lowest values for both.
# - **TM798 stands out** with customers having both higher income and mileage expectations.
# - **TM195** users are generally younger, with lower income and mileage expectations.
# - **TM498** users show intermediate values for income and mileage, with a more balanced age distribution.
````

 final observations

1. **Product Preference by Age**:
   - **TM195** is more popular among younger customers, as indicated by the lower average age compared to the other products.
   - **TM798** is favored by older customers, with a peak age range around 40-60 years.

2. **Income and Product Preference**:
   - **TM798** customers tend to have higher incomes, with a significant portion of buyers earning more than $80,000.

3. **Mileage Expectations**:
   - **TM798** stands out with customers planning to use their treadmill for significantly higher mileage, indicating more intense usage of the product.
   - **TM195** and **TM498** customers generally expect lower mileage, with TM195 having the lowest mileage expectations.

4. **Education and Product Preference**:
   - **TM798** appeals more to highly educated customers, with a noticeable difference in education level compared to TM195 and TM498.
   - **TM498** and **TM195** attract a more diverse range of education levels, but TM798 tends to skew toward customers with higher education.

5. **Gender and Product Preference**:
   - Gender does not show a significant difference in product preferences, with the distributions being relatively balanced across products.
   - While there is a slight gender imbalance overall, it does not significantly affect the choice of product.

6. **Product Differentiation**:
   - **TM195** is characterized by younger, lower-income customers who plan to use the treadmill moderately.
   - **TM498** has a more balanced customer base in terms of income and mileage expectations.
   - **TM798** is differentiated by its older, wealthier customers who expect the highest mileage and tend to have more education.

7. **Correlation Insights**:
   - Age and income are moderately correlated, with older customers generally having higher incomes.
   - There is no strong correlation between income and planned mileage, suggesting that income may not be the sole driver of higher usage plans.
   - **Miles** does not correlate strongly with age or income, which suggests that customers mileage expectations may be driven by factors other than these two variables, such as fitness goals or lifestyle.
