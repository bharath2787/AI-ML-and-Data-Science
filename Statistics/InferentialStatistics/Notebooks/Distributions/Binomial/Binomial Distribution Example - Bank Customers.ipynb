# Binomial Distribution
## Problem Statement
A bank issues credit card to its customers under the scheme of Master card. Based on past data, it is known that 60% of the customers pay the bill on time. If a sample of 7 accounts is selected at random from current database, construct the binomial  probability distribution of accounts paying on time.

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Libraries:
# numpy: For creating an array of possible outcomes (X).
# scipy.stats.binom: Provides the Binomial PMF function to calculate probabilities.
# matplotlib.pyplot: For visualizing the distribution.
```

```python
# Problem parameters
n = 7  # Number of trials (sample size)
p = 0.6  # Probability of success (paying on time)
```

```python
# Define possible values of X (number of accounts paying on time)
x = np.arange(0, n+1)  # X can take values from 0 to 7 (inclusive)
```

```python
# Calculate the probabilities using the Binomial PMF formula
# P(X = k) = C(n, k) * (p^k) * ((1-p)^(n-k))
pmf = binom.pmf(x, n, p)
```

```python
# Display the results in tabular format
print("Binomial Probability Distribution:")
print("X (accounts paying on time):", x)
print("P(X) (probabilities):", pmf)
```

```python
# Plot the Binomial Distribution
plt.figure(figsize=(8, 5))
plt.bar(x, pmf, color='skyblue', alpha=0.8, edgecolor='black')
plt.title('Binomial Distribution (n=7, p=0.6)', fontsize=16)
plt.xlabel('Number of accounts paying on time', fontsize=12)
plt.ylabel('Probability', fontsize=12)
plt.xticks(x)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### **Explanation of the Output**

The binomial probability distribution gives the likelihood of observing a certain number of accounts paying on time (\(X\)) out of the 7 sampled accounts, assuming the probability of paying on time (\(p\)) is 0.6.

---

### **Tabular Output**

| \( X \) (Accounts Paying on Time) | \( P(X) \) (Probability) |
|-----------------------------------|--------------------------|

| 2                                 | 0.0729                  |

| 4                                 | 0.2722                  |

#### Key Observations:
1. **Low \(X\) Values (0 or 1 Accounts)**:  
   - Very low probabilities (\(P(X=0) = 0.0016\)) because it's unlikely that no accounts or only 1 account pays on time given \(p = 0.6\).

2. **Peak Probability (\(P(X=4)\)):**
   - \(P(X=4) = 0.2722\) is the highest, meaning it’s most likely that **4 accounts** out of 7 will pay on time. This aligns with \(p = 0.6\), as we expect about 60% of the 7 accounts (\(7 \times 0.6 = 4.2\)) to pay on time.

3. **High \(X\) Values (6 or 7 Accounts):**  
   - Lower probabilities for 6 (\(P(X=6) = 0.0961\)) and 7 (\(P(X=7) = 0.0168\)) accounts paying on time because it's rare for nearly all accounts to pay on time, given \(p = 0.6\).

---

### **Graphical Output**

1. **Shape of the Bar Chart**:
   - The distribution is slightly **right-skewed** because the probability of success (\(p = 0.6\)) is higher than 50%.  
   - The highest bar is at \(X = 4\), showing that 4 accounts paying on time is the most likely outcome.

2. **Interpretation of Heights**:
   - Each bar represents the probability of \(X\) accounts paying on time.  
   - The total area (sum of heights) of all bars equals 1, as this is a valid probability distribution.

---

### **Real-World Interpretation**

If a bank selects 7 accounts randomly, here's what the probabilities suggest:
- The bank is **most likely** to find 4 accounts paying on time.  
- There's a **small chance** (1.7%) that only 1 account pays on time.  
- It’s **unlikely** (1.6%) that no accounts will pay on time, which is reassuring for the bank.

This distribution helps the bank plan resources, predict customer behavior, and analyze risks based on historical data.

#Scenario 2: Extentions of above problem:

### **Exploring Cumulative Probabilities**

Cumulative probabilities help us answer questions like:
- **What is the probability of 3 or fewer accounts paying on time?**  
- **What is the probability of more than 5 accounts paying on time?**

We calculate cumulative probabilities using the **Cumulative Distribution Function (CDF)** of the binomial distribution.

---


```python

# Calculate cumulative probabilities (CDF)
cdf = binom.cdf(x, n, p)

```

```python
# Display the results
print("Cumulative Probabilities (CDF):")
print("X (accounts paying on time):", x)
print("Cumulative P(X <= k):", cdf)
```


### **Code Explanation**

1. **`binom.cdf(x, n, p)`**:  
   - Computes the cumulative probability $( P(X \leq k) )$ for each \( k \).  
   - The CDF is useful for determining the probability of events **up to a certain value**.




```python
# Example queries
probability_3_or_less = cdf[3]  # P(X <= 3)  # add all the probs including 3
probability_more_than_5 = 1 - cdf[5]  # P(X > 5) # add all the probs till 5 and subtract from 1

print("\nKey Results:")
print(f"1. Probability of 3 or fewer accounts paying on time: {probability_3_or_less:.4f}")
print(f"2. Probability of more than 5 accounts paying on time: {probability_more_than_5:.4f}")
```


2. **Example Queries**:
   - $( P(X \leq 3) )$: Probability that 3 or fewer accounts pay on time.
   - $( P(X > 5) = 1 - P(X \leq 5) )$: Probability that more than 5 accounts pay on time.

```python



# Visualizing the CDF
plt.figure(figsize=(8, 5))
plt.step(x, cdf, where='mid', color='orange', label='CDF', linewidth=2)
plt.scatter(x, cdf, color='red')  # Highlight points on the CDF curve
plt.title('Cumulative Distribution Function (n=7, p=0.6)', fontsize=16)
plt.xlabel('Number of accounts paying on time', fontsize=12)
plt.ylabel('Cumulative Probability', fontsize=12)
plt.xticks(x)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()
plt.show()

```



### **Expected Output**

| \( X \) (Accounts Paying on Time) | \( P(X \leq k) \) (Cumulative Probability) |
|-----------------------------------|-------------------------------------------|

| 1                                 | 0.0188                                    |
| 2                                 | 0.0918                                    |
| 3                                 | 0.2733                                    |
| 4                                 | 0.5455                                    |
| 5                                 | 0.7717                                    |
| 6                                 | 0.8678                                    |
| 7                                 | 0.8846                                    |

---

#### **Key Results**:
1. **Probability of 3 or fewer accounts paying on time**:  
   $( P(X \leq 3) = 0.2733 )$ (27.33%).  
   - This means there’s about a 27% chance that at most 3 accounts will pay on time.

2. **Probability of more than 5 accounts paying on time**:  
   $( P(X > 5) = 1 - P(X \leq 5) = 0.1283 )$ (12.83%).  
   - This shows it’s relatively rare for more than 5 accounts to pay on time.

---

### **Graph Output**:
- The CDF graph rises step-by-step from $( P(X=0) = 0.0016 )$ to $( P(X=7) = 1 )$.  
- The steepest rise occurs around \( X = 4 \), indicating that outcomes near this value are the most probable.
