
# Problem Statement:
# Fashion Trends Online (FTO) is an e-commerce company that sells women's apparel.
# It is observed that about 10% of their customers return the items purchased by them
# for many reasons (such as size, color, and material mismatch).
# On a particular day, 20 customers purchased items from FTO. Calculate:
#
# 1. Probability that exactly 5 customers will return the items.
# 2. Probability that a maximum of 5 customers will return the items.
# 3. Probability that more than 5 customers will return the items purchased by them.

```python

# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom
```

```python

# Problem parameters
n = 20  # Total number of customers
p = 0.1  # Probability of a customer returning the item
```

```python
# Create an array of possible values for the number of customers returning items (0 to 20)
x = np.arange(0, n+1)
```

```python
# Calculate the binomial probability mass function (PMF) for each value of x
pmf_values = binom.pmf(x, n, p)
```

```python
pmf_values
```

```python

# Plot the binomial PMF
plt.figure(figsize=(10, 6))
plt.bar(x, pmf_values, color='skyblue', edgecolor='black')
plt.title('Binomial Probability Distribution (P(X = k))', fontsize=14)
plt.xlabel('Number of Customers Returning Items', fontsize=12)
plt.ylabel('Probability', fontsize=12)
plt.xticks(x)
plt.grid(True)
plt.show()
```

```python

# Plot the cumulative distribution function (CDF)
cdf_values = binom.cdf(x, n, p)

# Plotting the binomial CDF
plt.figure(figsize=(10, 6))
plt.step(x, cdf_values, where='mid', color='orange', linewidth=2)
plt.title('Cumulative Probability Distribution (P(X <= k))', fontsize=14)
plt.xlabel('Number of Customers Returning Items', fontsize=12)
plt.ylabel('Cumulative Probability', fontsize=12)
plt.xticks(x)
plt.grid(True)
plt.show()
```

```python

# 1. Probability that exactly 5 customers will return the items
prob_exactly_5 = binom.pmf(5, n, p)
print(f"1. Probability that exactly 5 customers will return the items: {prob_exactly_5:.4f}")
```

```python
# 2. Probability that a maximum of 5 customers will return the items (P(X <= 5))
prob_max_5 = binom.cdf(5, n, p)
print(f"2. Probability that a maximum of 5 customers will return the items: {prob_max_5:.4f}")
```

```python

# 3. Probability that more than 5 customers will return the items (P(X > 5))
prob_more_than_5 = 1 - prob_max_5
print(f"3. Probability that more than 5 customers will return the items: {prob_more_than_5:.4f}")
```

### Results for the Problem:

1. **Probability that exactly 5 customers will return the items**:
   - $( P(X = 5) \approx 0.0319 )$ or **3.19%**.

2. **Probability that a maximum of 5 customers will return the items**:
   - $( P(X \leq 5) \approx 0.9887 )$ or **98.87%**.

3. **Probability that more than 5 customers will return the items**:
   - $( P(X > 5) \approx 0.0113 )$ or **1.13%**.

These results show that while the probability of 5 or fewer customers returning items is very high (98.87%), the chance of more than 5 customers returning items is relatively low (1.13%).
