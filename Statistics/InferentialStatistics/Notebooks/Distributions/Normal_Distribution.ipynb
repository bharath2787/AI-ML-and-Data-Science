### Normal Distribution in Statistics

The normal distribution, also known as the Gaussian distribution, is one of the most important probability distributions in statistics. It is characterized by its bell-shaped curve, which is symmetric about the mean. Many natural phenomena and measurements tend to follow this distribution.

#### Key Characteristics of Normal Distribution

1. **Bell-shaped Curve**: The distribution is symmetric about the mean, with the highest frequency occurring at the mean, and frequencies decreasing as you move away from the mean.

2. **Mean, Median, and Mode**: In a perfectly normal distribution, the mean, median, and mode are all equal and located at the center of the distribution.

3. **Symmetry**: The distribution is perfectly symmetric around the mean. This means that the left side of the mean is a mirror image of the right side.

4. **Asymptotic**: The tails of the distribution approach, but never touch, the horizontal axis. This means that extreme values are possible but become increasingly rare.

5. **Defined by Mean and Standard Deviation**: The shape and position of the normal distribution are determined by the mean (μ) and standard deviation (σ). The mean determines the center, and the standard deviation determines the spread.

#### Probability Density Function (PDF)

The probability density function (PDF) of a normal distribution is given by:

$[ f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(x - \mu)^2}{2\sigma^2}} ]$

Where:
- $( \mu )$ is the mean
- $( \sigma )$ is the standard deviation
- $( e )$ is the base of the natural logarithm


```python


# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Generate data following a normal distribution
mean = 0
std_dev = 1
data = np.random.normal(mean, std_dev, 1000)

# Plot the histogram of the data
plt.figure(figsize=(10, 6))
count, bins, ignored = plt.hist(data, bins=30, density=True, alpha=0.6, color='g')

# Plot the normal distribution curve
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, mean, std_dev)
plt.plot(x, p, 'k', linewidth=2)
title = "Normal Distribution with mean = 0 and std deviation = 1"
plt.title(title)
plt.show()

```


### Explanation of the Code

1. **Generate Data**:
   - We generate 1000 data points following a normal distribution with a mean of `0` and a standard deviation of `1` using `np.random.normal()`.

2. **Plot the Histogram**:
   - We create a histogram of the generated data to visualize its distribution.

3. **Plot the Normal Distribution Curve**:
   - We overlay a normal distribution curve using `scipy.stats.norm.pdf()` to show the theoretical normal distribution with the same mean and standard deviation.



### Practical Example: Height of Adults

Let's consider a practical example: the height of adults in a population. Suppose the average height (mean) is 170 cm, and the standard deviation is 10 cm. The height of adults follows a normal distribution. We can visualize this with the following code:


```python


# Parameters for the normal distribution
mean = 170
std_dev = 10

# Generate data
data = np.random.normal(mean, std_dev, 1000)

# Plot the histogram
plt.figure(figsize=(10, 6))
count, bins, ignored = plt.hist(data, bins=30, density=True, alpha=0.6, color='g')

# Plot the normal distribution curve
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, mean, std_dev)
plt.plot(x, p, 'k', linewidth=2)
title = "Normal Distribution of Adult Heights with mean = 170 cm and std deviation = 10 cm"
plt.title(title)
plt.xlabel("Height (cm)")
plt.ylabel("Probability Density")
plt.show()
```



### Explanation of the Code

1. **Set Parameters**:
   - We set the mean to `170` cm and the standard deviation to `10` cm.

2. **Generate Data**:
   - We generate 1000 data points following this normal distribution using `np.random.normal()`.

3. **Plot the Histogram**:
   - We create a histogram to visualize the distribution of heights.

4. **Plot the Normal Distribution Curve**:
   - We overlay a normal distribution curve using `scipy.stats.norm.pdf()` to show the theoretical distribution with the given mean and standard deviation.
