```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Set the style for the plots
sns.set(style="whitegrid")
```

2. Generate Dummy Data
Create a dataset with a normal distribution:

```python
# Parameters for the normal distribution
mean = 50
std_dev = 10
size = 1000

# Generate dummy data
np.random.seed(0)
data = np.random.normal(loc=mean, scale=std_dev, size=size)

# Create a DataFrame
df = pd.DataFrame(data, columns=['Value'])
```

3. Plot the Original Data
Visualize the original normal distribution:

```python
plt.figure(figsize=(12, 6))

# Plot histogram
sns.histplot(df['Value'], kde=True, color='blue', bins=30)
plt.title('Histogram of Original Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

4. Convert to Standard Normal Distribution
Standardize the data to have a mean of 0 and a standard deviation of 1:

```python
# Standardize the data
df['Standardized'] = (df['Value'] - df['Value'].mean()) / df['Value'].std()

```

5. Plot the Standardized Data
Visualize the standardized (standard normal) distribution:

```python
plt.figure(figsize=(12, 6))

# Plot histogram
sns.histplot(df['Standardized'], kde=True, color='green', bins=30)
plt.title('Histogram of Standard Normal Distribution')
plt.xlabel('Standardized Value')
plt.ylabel('Frequency')
plt.show()
```

6. Statistical Summary
Provide a summary of the transformed data:

```python
pd.set_option('display.float_format', lambda x: '%.3f' % x)

# Display summary statistics
original_stats = df['Value'].describe()
standardized_stats = df['Standardized'].describe()

print("Original Data Statistics:")
print(original_stats)
print("\nStandardized Data Statistics:")
print(standardized_stats)
```

7. Z-Score Calculation Example
Include an example of how to calculate the Z-score for a specific value:

```python
# Example value
value = 60

# Calculate Z-score
z_score = (value - mean) / std_dev
print(f"\nZ-score for value {value} is: {z_score}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Set the style for the plots
sns.set(style="whitegrid")

# Parameters for the normal distribution
mean = 50
std_dev = 10
size = 1000

# Generate dummy data
np.random.seed(0)
data = np.random.normal(loc=mean, scale=std_dev, size=size)

# Create a DataFrame
df = pd.DataFrame(data, columns=['Value'])

# Plot histogram of original normal distribution
plt.figure(figsize=(12, 6))
sns.histplot(df['Value'], kde=True, color='blue', bins=30)
plt.title('Histogram of Original Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

# Standardize the data
df['Standardized'] = (df['Value'] - df['Value'].mean()) / df['Value'].std()

# Plot histogram of standardized normal distribution
plt.figure(figsize=(12, 6))
sns.histplot(df['Standardized'], kde=True, color='green', bins=30)
plt.title('Histogram of Standard Normal Distribution')
plt.xlabel('Standardized Value')
plt.ylabel('Frequency')
plt.show()

# Display summary statistics
original_stats = df['Value'].describe()
standardized_stats = df['Standardized'].describe()

print("Original Data Statistics:")
print(original_stats)
print("\nStandardized Data Statistics:")
print(standardized_stats)

# Example value for Z-score calculation
value = 60
z_score = (value - mean) / std_dev
print(f"\nZ-score for value {value} is: {z_score}")
```
