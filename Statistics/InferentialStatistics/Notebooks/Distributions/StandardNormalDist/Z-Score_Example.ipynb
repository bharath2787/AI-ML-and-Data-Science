```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import zscore

# Generating a sample dataset
data = {
    'Feature_A': np.random.randint(10, 100, 20),
    'Feature_B': np.random.randint(200, 500, 20)
}
df = pd.DataFrame(data)

# Displaying the original dataset
print("Original Dataset:")
print(df.head())

# Applying Z-score normalization
df_normalized = df.apply(zscore)

# Displaying the transformed dataset
print("\nDataset After Z-score Normalization:")
print(df_normalized.head())

# Plotting the distributions before and after normalization
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].hist(df['Feature_A'], bins=10, alpha=0.7, label='Feature_A')
axes[0].hist(df['Feature_B'], bins=10, alpha=0.7, label='Feature_B')
axes[0].set_title("Before Z-score Normalization")
axes[0].legend()

axes[1].hist(df_normalized['Feature_A'], bins=10, alpha=0.7, label='Feature_A')
axes[1].hist(df_normalized['Feature_B'], bins=10, alpha=0.7, label='Feature_B')
axes[1].set_title("After Z-score Normalization")
axes[1].legend()

plt.show()

# Explanation
explanation = """
Z-score normalization transforms the data so that it has a mean of 0 and a standard deviation of 1.
This helps in:
1. Making different features comparable.
2. Improving performance in machine learning models that are sensitive to feature scales.
3. Detecting outliers more easily as values farther from 0 indicate extreme values.
"""
print(explanation)
```

 **Z-score normalization removes the original distance (scale) between Feature_A and Feature_B** because each feature is transformed **independently** to have a mean of 0 and a standard deviation of 1.  

### **What Information is Lost?**  
- The original absolute **scale and units** of the features.  
- The **relative distance** between Feature_A and Feature_B in their raw form.  
- If Feature_B had values much larger than Feature_A, that distinction is lost after normalization.

### **Why Does This Happen?**  
Each feature is transformed using its **own mean and standard deviation**, making them **comparable in terms of deviation from their mean**, but losing the direct distance between them.

### **When is this Acceptable?**  
- In **Machine Learning**, especially for algorithms like KNN, SVM, and Gradient Descent-based models, where feature scaling ensures equal importance.
- When features have **different units** (e.g., height in cm vs weight in kg), and we want to remove unit dependence.
- In **PCA**, where scaling ensures that no feature dominates due to its larger magnitude.

### **When is this a Problem?**  
- If the absolute difference between features **carries important meaning** (e.g., raw distances in physics, financial data, or medical measurements).  
- If features need to **retain their original scales for interpretability**.
