# Poisson Distribution Problem Solution

# Problem Statement:
 ## If on an average, 6 customers arrive every two minutes at a bank during the busy hours of working,
### a) what is the probability that exactly four customers arrive in a given minute?
### b) What is the probability that more than three customers will arrive in a given minute?

## Given data:
### The average arrival rate is 6 customers every 2 minutes, so:
### Î» (lambda) = 6/2 = 3 customers per minute

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson
```

```python

# Given values
lambda_rate = 3  # average number of customers per minute
k_values = np.arange(0, 11)  # range of k values for plotting PMF and CDF
```

```python

# Part (a) and (b) calculations
probability_exactly_4 = poisson.pmf(4, lambda_rate)  # PMF for exactly 4 customers
probability_more_than_3 = 1 - poisson.cdf(3, lambda_rate)  # CDF for more than 3 customers
```

```python
# PMF plot
plt.figure(figsize=(12, 6))

# Plotting PMF
plt.subplot(1, 2, 1)
plt.bar(k_values, poisson.pmf(k_values, lambda_rate), color='skyblue', label='PMF')
plt.title('Poisson Distribution - PMF')
plt.xlabel('Number of Customers (k)')
plt.ylabel('Probability')
plt.xticks(k_values)
plt.grid(True)
```

```python
# Plotting CDF
plt.subplot(1, 2, 2)
plt.step(k_values, poisson.cdf(k_values, lambda_rate), color='orange', label='CDF', where='post')
plt.title('Poisson Distribution - CDF')
plt.xlabel('Number of Customers (k)')
plt.ylabel('Cumulative Probability')
plt.xticks(k_values)
plt.grid(True)

# Displaying results
plt.tight_layout()
plt.show()
```

```python
# Output the solution for part (a) and (b)
probability_exactly_4, probability_more_than_3
```



### Summary of Results:
- **Probability of exactly 4 customers arriving in a minute**: $( P(X = 4) \approx 16.8\% )$.
- **Probability of more than 3 customers arriving in a minute**: $( P(X > 3) \approx 35.3\% )$.
