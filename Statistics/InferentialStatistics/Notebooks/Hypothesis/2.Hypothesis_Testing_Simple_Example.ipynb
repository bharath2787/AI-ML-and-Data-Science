

 **Step-by-step breakdown of hypothesis testing:**  

---

## **üîπ Hypothesis Testing Process**  

### **1Ô∏è‚É£ Define the Hypotheses**  
- **Null Hypothesis ($(H_0)$)** ‚Üí Assumes no change, effect, or difference (e.g., "The average bolt length is 5.00 cm").  
- **Alternative Hypothesis ($(H_1)$)** ‚Üí What we are testing for (e.g., "The average bolt length is not 5.00 cm").  

---

### **2Ô∏è‚É£ Collect a Sample & Compute the Sample Statistic**  
- Take a **random sample** from the population.  
- Calculate the **sample mean ($(\bar{X})$)**, standard deviation ($(s)$), and sample size (\(n\)).  

---

### **3Ô∏è‚É£ Convert the Test Statistic to a Probability (Find p-value)**  
- Compute a **test statistic** (e.g., Z-score or t-score):  

  $[
  Z = \frac{\bar{X} - \mu}{\sigma / \sqrt{n}}
  ]$

- Use a **Z-table** (for large samples) or a **t-table** (for small samples) to find the **p-value** (the probability of getting this sample if $(H_0)$ is true).  

---

### **4Ô∏è‚É£ Compare p-value with $(\alpha)$ (Significance Level)**  
- If **$( p < \alpha )$** ‚Üí **Reject $(H_0)$** (strong evidence against $(H_0)$).  
- If **$( p > \alpha )$** ‚Üí **Fail to reject $(H_0)$** (not enough evidence to reject it).  

üîπ **Common $(\alpha)$ values:**  
‚úî **0.05 (5%)** ‚Üí We accept a 5% risk of wrongly rejecting $(H_0)$.  
‚úî **0.01 (1%)** ‚Üí Stricter test, only 1% risk allowed.  

---

### **5Ô∏è‚É£ Make a Conclusion**  
- **If $( p < \alpha )$** ‚Üí **Reject $(H_0)$** ‚Üí There is significant evidence that $(H_1)$ is true.  
- **If $( p > \alpha )$** ‚Üí **Fail to reject $(H_0)$** ‚Üí We don‚Äôt have enough evidence to support $(H_1)$.  

---

## **üîç Example: Testing Bolt Length**
üöÄ A factory claims the **average bolt length is 5.00 cm**. We suspect otherwise and test this claim.

‚úÖ **Step 1: Define Hypotheses**  
- **$(H_0)$**: The true mean bolt length is **5.00 cm**.  
- **$(H_1)$**: The true mean bolt length **is not** 5.00 cm.  

‚úÖ **Step 2: Collect Sample & Compute Test Statistic**  
- Sample of **30 bolts** ‚Üí Sample mean **$( \bar{X} = 5.05 )$ cm**.  
- Standard deviation **\( s = 0.05 \) cm**.  

‚úÖ **Step 3: Convert to Probability (p-value)**  
- Compute **Z-score**:  

  $[
  Z = \frac{5.05 - 5.00}{0.05 / \sqrt{30}} = 2.74
  ]$

- Look up **Z = 2.74** in the normal table ‚Üí **p-value = 0.006**.  

‚úÖ **Step 4: Compare p-value with $(\alpha)$**  
- If **$( \alpha = 0.05 )$**, then **$( p = 0.006 < 0.05 )$** ‚Üí **Reject $(H_0)$**.  

‚úÖ **Step 5: Conclusion**  
- The p-value is very small, meaning the sample mean (5.05 cm) is unlikely under $(H_0)$.  
- **We reject $(H_0)$** and conclude that the factory bolts are likely **not 5.00 cm on average**.  

---

### **üöÄ Final Takeaways**
‚úî **p-value tells us how extreme the sample is under $(H_0)$.**  
‚úî **If $( p < \alpha )$, we reject $(H_0)$; if $( p > \alpha )$, we fail to reject $(H_0)$.**  
‚úî **Hypothesis testing helps make data-driven decisions instead of guessing.**
