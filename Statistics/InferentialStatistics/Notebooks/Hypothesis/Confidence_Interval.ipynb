### **Confidence Interval (CI) - Explained**  

A **confidence interval (CI)** is a range of values, derived from sample data, that is **likely to contain the true population parameter** with a certain level of confidence.  

---

### **Key Components of a Confidence Interval**
1. **Point Estimate**: The sample statistic (e.g., sample mean \( \bar{x} \) or sample proportion \( \hat{p} \)) that serves as the best estimate of the population parameter.
2. **Margin of Error (ME)**: The range around the point estimate, influenced by **sample variability** and **confidence level**.
3. **Confidence Level** (\( 1 - \alpha \)): The probability that the confidence interval contains the true population parameter. Common confidence levels:
   - **90% CI** (\( \alpha = 0.10 \))
   - **95% CI** (\( \alpha = 0.05 \)) ‚Üí Most commonly used
   - **99% CI** (\( \alpha = 0.01 \))

---

### **Formula for Confidence Interval**
#### **For a Population Mean (\(\mu\))**  
When the **population standard deviation (\(\sigma\)) is known**, we use the **Z-distribution**:

$[
CI = \bar{x} \pm Z_{\alpha/2} \times \frac{\sigma}{\sqrt{n}}
]$

When **\(\sigma\) is unknown**, we use the **T-distribution**:

$[
CI = \bar{x} \pm t_{\alpha/2, df} \times \frac{s}{\sqrt{n}}
]$

Where:  
- $( \bar{x} )$ = Sample mean  
- $( Z_{\alpha/2} )$ or $( t_{\alpha/2, df} )$ = Critical value from Z- or T-table  
- $( \sigma )$ or $( s )$ = Population or sample standard deviation  
- $( n )$ = Sample size  

---

#### **For a Population Proportion (\( p \))**
$[
CI = \hat{p} \pm Z_{\alpha/2} \times \sqrt{\frac{\hat{p}(1 - \hat{p})}{n}}
]$

Where:  
- $( \hat{p} )$ = Sample proportion  
- $( Z_{\alpha/2} )$ = Z-critical value  
- $( n )$ = Sample size  

---

### **Example: Confidence Interval for a Mean**
A researcher surveys **100 students** about their weekly study hours. The sample mean is **15 hours**, and the standard deviation is **3 hours**. Find a **95% confidence interval** for the true mean study hours.

#### **Step 1: Identify Values**
- $( \bar{x} = 15 )$, $( s = 3 )$, $( n = 100 )$
- **95% confidence level** ‚Üí **\( Z_{0.025} = 1.96 \)** (from Z-table)

#### **Step 2: Compute Margin of Error (ME)**
$[
ME = 1.96 \times \frac{3}{\sqrt{100}} = 1.96 \times 0.3 = 0.588
]$

#### **Step 3: Compute Confidence Interval**
$[
CI = 15 \pm 0.588
]$

$[
(14.41, 15.59)
]$

üîπ **Interpretation**: We are **95% confident** that the true mean study hours lie between **14.41 and 15.59 hours**.

---

### **Interpreting Confidence Intervals**
- **Correct**: "We are 95% confident that the population mean lies within the interval."
- **Incorrect**: "There is a 95% probability that the population mean is in this interval." ‚ùå (The true mean is fixed; the interval varies.)

---

### **Real-World Applications of Confidence Intervals**
1. **Business Analytics**: Estimating **customer satisfaction scores**.
2. **Medical Research**: Determining **average blood pressure** in a population.
3. **Elections**: Estimating **polling results** (e.g., "Candidate A: 52% ¬± 3% at 95% confidence").
