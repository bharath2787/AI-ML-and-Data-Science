In hypothesis testing, Type I and Type II errors are two possible errors that can occur when making decisions based on statistical tests:

1. **Type I Error (False Positive):**
   - **Definition:** Occurs when the null hypothesis (\(H_0\)) is incorrectly rejected when it is actually true.
   - **Consequences:** This means you conclude there is an effect or difference when there really isn't one. For example, a medical test might indicate a patient has a disease when they do not.
   - **Probability:** The probability of making a Type I error is denoted by \(\alpha\) (alpha) and is also known as the significance level of the test. Common values for \(\alpha\) are 0.05 or 0.01.

2. **Type II Error (False Negative):**
   - **Definition:** Occurs when the null hypothesis (\(H_0\)) is incorrectly accepted when the alternative hypothesis (\(H_A\)) is actually true.
   - **Consequences:** This means you fail to detect an effect or difference that is actually present. For instance, a test might fail to detect a disease that is actually there.
   - **Probability:** The probability of making a Type II error is denoted by \(\beta\) (beta). The power of a test, which is \(1 - \beta\), measures the testâ€™s ability to correctly detect an effect when there is one.

In summary:
- **Type I Error:** Incorrectly rejecting a true null hypothesis.
- **Type II Error:** Incorrectly failing to reject a false null hypothesis.

Balancing these errors involves trade-offs: reducing the risk of a Type I error typically increases the risk of a Type II error, and vice versa.
