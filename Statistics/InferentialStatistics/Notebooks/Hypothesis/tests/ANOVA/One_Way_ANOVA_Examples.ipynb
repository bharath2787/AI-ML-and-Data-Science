Here are two example problems for **One-Way ANOVA**:

---

### **Example 1: Examining the Effect of Different Study Methods on Exam Scores**  
A teacher wants to determine whether different study methods have an impact on student exam scores. She randomly assigns 15 students to three different study methods:  

1. **Self-study**  
2. **Group study**  
3. **Tutoring sessions**  

The exam scores (out of 100) for each group are as follows:  

| **Self-Study** | **Group Study** | **Tutoring** |  
|--------------|--------------|------------|  
| 75          | 82           | 90         |  
| 80          | 85           | 95         |  
| 78          | 87           | 92         |  
| 72          | 83           | 88         |  
| 74          | 81           | 94         |  

#### **Step 1: Hypotheses**  
- **H₀ (Null Hypothesis):** The mean exam scores are the same across all study methods.  
- **H₁ (Alternative Hypothesis):** At least one study method leads to a different mean exam score.  

#### **Step 2: Compute F-Value**  
1. Calculate the **group means**.  
2. Compute the **Between-Group Variance** and **Within-Group Variance**.  
3. Use the **F-ratio formula**:  
   $[
   F = \frac{\text{Between-group variance}}{\text{Within-group variance}}
   ]$  
4. Compare the **F-value** with the critical value from the F-distribution table or calculate the **p-value**.  

#### **Step 3: Conclusion**  
- If **p < 0.05**, we reject H₀ and conclude that study methods significantly impact exam scores.  
- Otherwise, we fail to reject H₀, meaning no significant difference in scores.  

---


```python
import numpy as np
import scipy.stats as stats
import pandas as pd
import matplotlib.pyplot as plt

```

```python
# Example 1: Exam Scores for Different Study Methods
self_study = [75, 80, 78, 72, 74]
group_study = [82, 85, 87, 83, 81]
tutoring = [90, 95, 92, 88, 94]

```

```python
# Performing One-Way ANOVA
test_statistic, p_value = stats.f_oneway(self_study, group_study, tutoring)

print("Example 1: Study Methods vs Exam Scores")
print(f"F-value: {test_statistic:.4f}")
print(f"p-value: {p_value:.4f}")
if p_value < 0.05:
    print("Reject the null hypothesis: Study methods significantly impact exam scores.")
else:
    print("Fail to reject the null hypothesis: No significant difference in exam scores.")

```

```python
# Visualization: Boxplots
plt.figure(figsize=(12, 5))

# Exam Scores
plt.subplot(1, 2, 1)
plt.boxplot([self_study, group_study, tutoring], labels=["Self-Study", "Group Study", "Tutoring"])
plt.title("Study Methods vs Exam Scores")
plt.ylabel("Exam Score")
```


### **Example 2: Analyzing Customer Satisfaction Across Three Coffee Brands**  
A market researcher wants to determine if there is a significant difference in customer satisfaction among three coffee brands: **Brand A, Brand B, and Brand C**. A survey is conducted, and customers rate their satisfaction on a scale of 1 to 10.  

| **Brand A** | **Brand B** | **Brand C** |  
|------------|------------|------------|  
| 7          | 8          | 6          |  
| 6          | 9          | 7          |  
| 8          | 7          | 5          |  
| 7          | 6          | 6          |  
| 9          | 8          | 7          |  

#### **Step 1: Hypotheses**  
- **H₀:** There is no difference in mean customer satisfaction across the three coffee brands.  
- **H₁:** At least one brand has a different mean satisfaction level.  

#### **Step 2: Compute F-Value**  
Follow the same steps as in Example 1:  
1. Compute the **mean satisfaction score** for each brand.  
2. Calculate the **Between-Group** and **Within-Group Variances**.  
3. Use the **F-ratio formula** to determine the F-value and p-value.  

#### **Step 3: Conclusion**  
- If **p < 0.05**, we reject H₀ and conclude that at least one coffee brand has a significantly different customer satisfaction level.  
- If **p > 0.05**, we fail to reject H₀, meaning no significant difference between brands.  

---


```python
# Example 2: Customer Satisfaction for Coffee Brands
brand_A = [7, 6, 8, 7, 9]
brand_B = [8, 9, 7, 6, 8]
brand_C = [6, 7, 5, 6, 7]

```

```python
# Performing One-Way ANOVA
test_statistic, p_value = stats.f_oneway(brand_A, brand_B, brand_C)

print("\nExample 2: Coffee Brand vs Customer Satisfaction")
print(f"F-value: {test_statistic:.4f}")
print(f"p-value: {p_value:.4f}")
if p_value < 0.05:
    print("Reject the null hypothesis: At least one coffee brand has different customer satisfaction.")
else:
    print("Fail to reject the null hypothesis: No significant difference in customer satisfaction.")
```

```python
# Visualization: Boxplots
plt.figure(figsize=(12, 5))
```

```python

# Customer Satisfaction
plt.subplot(1, 2, 2)
plt.boxplot([brand_A, brand_B, brand_C], labels=["Brand A", "Brand B", "Brand C"])
plt.title("Coffee Brand vs Customer Satisfaction")
plt.ylabel("Satisfaction Score")

plt.tight_layout()
plt.show()
```
