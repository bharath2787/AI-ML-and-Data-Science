

### Problem Statement

A teacher wants to know if there is a relationship between students' preferred study methods and their exam performance. The study methods are categorized as "Group Study," "Self-Study," and "Online Resources," and the exam performance is categorized as "Pass" or "Fail." The data collected from a sample of 120 students is as follows:

| Study Method      | Pass | Fail | Total |
|-------------------|------|------|-------|
| Group Study       | 20   | 10   | 30    |
| Self-Study        | 50   | 20   | 70    |
| Online Resources  | 10   | 10   | 20    |
| **Total**         | 80   | 40   | 120   |


```python
import numpy as np
import scipy.stats as stats

# Observed frequencies
observed = np.array([[20, 10], [50, 20], [10, 10]])

# Perform the Chi-Square Test of Independence
chi2, p, dof, expected = stats.chi2_contingency(observed)

# Output the results
print(f"Chi-Square Statistic: {chi2}")
print(f"P-Value: {p}")
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies:")
print(expected)

# Interpretation
alpha = 0.05
if p < alpha:
    print("Reject the null hypothesis: There is a significant association between study method and exam performance.")
else:
    print("Fail to reject the null hypothesis: There is no significant association between study method and exam performance.")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Degrees of freedom
dof = 2

# Chi-Square statistic from the example
chi2_stat = 3.209

# Generate a range of values for the Chi-Square distribution
x = np.linspace(0, 10, 1000)
y = stats.chi2.pdf(x, df=dof)

# Plotting the Chi-Square distribution
plt.figure(figsize=(10, 6))
plt.plot(x, y, label=f'Chi-Square Distribution df={dof}', color='blue')

# Shade the area under the curve where the chi2_stat falls
plt.fill_between(x, 0, y, where=(x >= chi2_stat), color='red', alpha=0.5, label=f'χ² = {chi2_stat:.3f}')

# Mark the calculated Chi-Square statistic
plt.axvline(chi2_stat, color='red', linestyle='--', label=f'Calculated χ² = {chi2_stat:.3f}')

# Critical value for 95% confidence level (alpha = 0.05)
critical_value = stats.chi2.ppf(0.95, dof)
plt.axvline(critical_value, color='green', linestyle='--', label=f'Critical Value = {critical_value:.3f}')

# Labels and title
plt.title('Chi-Square Distribution with df=2')
plt.xlabel('Chi-Square Value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```
