#What is T-Distribution

* The **t-distribution** (Student's t-distribution) is a probability distribution that arises in situations where the sample size is small and the population standard deviation is unknown.

* It is similar to the normal distribution but with heavier tails, meaning it is more prone to producing values far from the mean.

### Key Concepts

1. **When to use the t-distribution**:
   - You have a small sample size (typically less than 30).
   - The population standard deviation ($( \sigma )$) is unknown, and you are estimating it with the sample standard deviation (\( s \)).
   - The data is approximately normally distributed.

2. **Shape of the t-distribution**:
   - Symmetrical and bell-shaped like the normal distribution.
   - The t-distribution has fatter tails compared to the normal distribution, meaning it has more extreme values. This accounts for the extra variability in small samples.
   - As the sample size increases, the t-distribution approaches the normal distribution.

3. **Degrees of Freedom (df)**:
   - The shape of the t-distribution depends on the **degrees of freedom (df)**, which is related to the sample size. For a one-sample t-test, \( df = n - 1 \), where \( n \) is the sample size.
   - Smaller degrees of freedom (small sample size) result in a distribution with heavier tails.
   - As degrees of freedom increase (larger sample size), the t-distribution becomes closer to a normal distribution.

4. **Applications**:
   - **t-tests**: Used to compare the means of two groups (one-sample, two-sample, paired sample t-tests).
   - **Confidence Intervals**: Used to calculate confidence intervals for a population mean when the sample size is small.

### Formula for t-Statistic

For a one-sample t-test, the t-statistic is calculated as:

$[
t = \frac{\bar{x} - \mu}{\frac{s}{\sqrt{n}}}
]$

Where:
- $( \bar{x} )$ is the sample mean.
- $( \mu )$ is the population mean (the null hypothesis value).
- $( s )$ is the sample standard deviation.
- $( n )$ is the sample size.

This t-value can then be compared to critical values from the t-distribution to determine whether the result is statistically significant.

### Example of Use

If you wanted to test whether the average height of a group of 10 people is different from a known value (say, 170 cm), you would:
1. Calculate the mean height of your sample.
2. Use the sample standard deviation as an estimate of the population standard deviation.
3. Compute the t-statistic using the formula above.
4. Compare this t-statistic to a critical value from the t-distribution table (based on your desired significance level and degrees of freedom).

### Summary:
- The t-distribution is a bell-shaped, symmetric distribution used when sample sizes are small and the population standard deviation is unknown.
- It's wider and has heavier tails than the normal distribution, but becomes similar to the normal distribution as the sample size increases.
- It's applied in t-tests and confidence interval calculations when estimating a population mean based on a sample.
