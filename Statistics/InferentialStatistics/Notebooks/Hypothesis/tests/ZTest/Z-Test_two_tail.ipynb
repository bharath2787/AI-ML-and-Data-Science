##Problem Statement:

- The average weight of all residents in town XYZ is known to be 168 lbs. A nutritionist believes that the true average weight of the residents is different from this value. To test this hypothesis, she collects a random sample of 36 individuals from the town and measures their weights. She finds that the sample has a mean weight of 169.5 lbs with a standard deviation of 3.9 lbs.

since the sample size is 36, which is greater than 30, we can use a z-test instead of a t-test. Let's go through the steps for the z-test.

### **Available stats:**

- **Population mean ($(\mu)$):** 168 lbs
- **Sample size ($(n)$):** 36
- **Sample mean ($(\bar{x})$):** 169.5 lbs
- **Sample standard deviation ($(s)$):** 3.9 lbs
- **Significance level ($(\alpha)$):** 0.05

### **Null and Alternative Hypotheses:**

- **Null Hypothesis (\(H_0\))**: The average weight of all residents in town XYZ is 168 lbs.
  
  $[
  H_0: \mu = 168
  ]$

- **Alternative Hypothesis ($(H_1)$)**: The average weight of all residents in town XYZ is different from 168 lbs.
  
  $[
  H_1: \mu \neq 168
  ]$

### **Z-test Steps:**

#### **Step 1: Calculate the test statistic**

The z-test statistic is given by:

$[
z = \frac{\bar{x} - \mu}{\frac{\sigma}{\sqrt{n}}}
]$

Since the population standard deviation (\(\sigma\)) is not provided, we'll use the sample standard deviation as an estimate:

$[
z = \frac{169.5 - 168}{\frac{3.9}{\sqrt{36}}} = \frac{1.5}{\frac{3.9}{6}} = \frac{1.5}{0.65} \approx 2.31
]$

#### **Step 2: Find the p-value**

Using a standard normal distribution (z-distribution), we look up the z-value of 2.31.

For a two-tailed test:

- The p-value corresponds to the area under the curve beyond $( |z| = 2.31 )$.
- The p-value for $( z = 2.31 )$ is approximately 0.021.

#### **Step 3: Compare the p-value with the significance level**

- **Significance level ($(\alpha)$)**: 0.05
- **P-value**: 0.021

Since the p-value (0.021) is less than the significance level (0.05), we reject the null hypothesis.

### **Conclusion:**

At the 95% confidence level, there is enough evidence to reject the null hypothesis. This means the average weight of all residents in town XYZ is significantly different from 168 lbs.

```python
import numpy as np
from scipy.stats import norm

# Given data
population_mean = 168
sample_mean = 169.5
sample_std = 3.9
sample_size = 36

# Step 1: Calculate the z-test statistic
z = (sample_mean - population_mean) / (sample_std / np.sqrt(sample_size))

# Step 2: Calculate the p-value for a two-tailed test
p_value = 2 * (1 - norm.cdf(np.abs(z)))

# Step 3: Print the results
print(f"Z-test statistic: {z:.2f}")
print(f"P-value: {p_value:.4f}")

# Step 4: Conclusion based on significance level (alpha = 0.05)
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is sufficient evidence to suggest that the average weight is different from 168 lbs.")
else:
    print("Fail to reject the null hypothesis: There is not enough evidence to suggest that the average weight is different from 168 lbs.")
```



### **Explanation:**

1. **Calculating the z-test statistic**:
   - The formula for the z-test statistic is \( z = \frac{\bar{x} - \mu}{\frac{s}{\sqrt{n}}} \).
   
2. **Calculating the p-value**:
   - Since this is a two-tailed test, we calculate the p-value as \( 2 \times (1 - \text{CDF}(|z|)) \), where CDF is the cumulative distribution function of the standard normal distribution.
   
3. **Conclusion**:
   - The code checks if the p-value is less than the significance level (0.05). If it is, we reject the null hypothesis; otherwise, we fail to reject it.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Given data
z_value = 2.31  # from previous calculation
alpha = 0.05

# Define the x range for the plot
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1)  # Standard normal distribution

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Normal Distribution', color='blue')

# Shade the rejection region on the left
plt.fill_between(x, y, where=(x <= -norm.ppf(1 - alpha/2)), color='red', alpha=0.3, label='Rejection Region (left)')

# Shade the rejection region on the right
plt.fill_between(x, y, where=(x >= norm.ppf(1 - alpha/2)), color='red', alpha=0.3, label='Rejection Region (right)')

# Plot the z-value
plt.axvline(z_value, color='green', linestyle='--', label=f'z-value = {z_value:.2f}')

# Add labels and title
plt.title('Normal Distribution with Rejection Regions')
plt.xlabel('Z')
plt.ylabel('Probability Density')
plt.legend()

# Show the plot
plt.show()
```

### **2. Explanation:**

1. **Normal Distribution Curve:**

2. **Rejection Regions:**
   - The areas shaded in red on both tails of the distribution represent the rejection regions for a two-tailed test at a 5% significance level (alpha = 0.05).
   - For a 95% confidence level, the critical z-values that define these regions are approximately 1.96.

3. **Z-value:**
   - The green dashed line represents the z-value we calculated, which is approximately 2.31.
   - This z-value lies in the right tail of the distribution, within the rejection region.

4. **P-value:**
   - The p-value is the probability of observing a test statistic at least as extreme as the one calculated, under the assumption that the null hypothesis is true.
   - Since our z-value falls within the rejection region, the p-value is smaller than our significance level (0.05). Specifically, the p-value is approximately 0.021.

5. **Decision to Reject the Null Hypothesis:**
   - Since the p-value (0.021) is less than the significance level (0.05), we reject the null hypothesis.
   - This indicates that there is statistically significant evidence to suggest that the average weight of residents in town XYZ is different from 168 lbs.

### **Why We Reject the Null Hypothesis:**

- **Critical Region (Rejection Region):**
  - The rejection regions are the areas of the distribution where, if the z-value falls within, the null hypothesis is rejected.
  - Since our z-value of 2.31 lies in one of these regions, it implies that the observed sample mean is significantly different from the population mean.

- **P-value Interpretation:**
  - The p-value represents the probability of obtaining a test statistic at least as extreme as the observed one, assuming the null hypothesis is true.
  - A low p-value (below the significance level) suggests that such an extreme result is unlikely under the null hypothesis, leading us to reject it.
