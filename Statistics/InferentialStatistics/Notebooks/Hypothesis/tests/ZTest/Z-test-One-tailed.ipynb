
## **Problem Statement:**

- A factory manufactures cars with a warranty of 5 years on the engine and transmission. An engineer believes that the engine or transmission will malfunction in less than 5 years. He tests a sample of 40 cars and finds that the average time to malfunction is 4.8 years with a standard deviation of 0.50 years.At a 2% significance level , determine if there is enough evidence to support the idea that the warranty should be revised.

**Tasks:**

1. **State the Hypotheses:**
   - Null Hypothesis ($(H_0)$): The mean malfunction time is 5 years. (\($mu = 5)$)
   - Alternative Hypothesis ($(H_1)$): The mean malfunction time is less than 5 years. ($(\mu < 5)$)

2. **Perform a One-Tailed Z-test:**
   - At a 2% significance level (alpha = 0.02), determine if there is enough evidence to support the idea that the warranty should be revised.

### **Step 1: Calculating the Z-test Statistic**

The formula for the z-test statistic is:

$[
z = \frac{\bar{x} - \mu}{\frac{s}{\sqrt{n}}}
]$

Where:
- $(\bar{x})$ = Sample mean = 4.8 years
- $(\mu)$ = Population mean (hypothesized mean) = 5 years
- $(s)$ = Sample standard deviation = 0.50 years
- $(n)$ = Sample size = 40

### **Step 2: Calculating the P-value**

For a one-tailed test, the p-value is calculated as:

$[
\text{p-value} = \text{CDF}(z)
]$

Where CDF is the cumulative distribution function of the standard normal distribution.

### **Step 3: Conclusion**

If the p-value is less than the significance level (alpha = 0.02), we reject the null hypothesis. Otherwise, we fail to reject it.


```python

import numpy as np
from scipy.stats import norm

# Given data
population_mean = 5
sample_mean = 4.8
sample_std = 0.50
sample_size = 40

# Step 1: Calculate the z-test statistic
z = (sample_mean - population_mean) / (sample_std / np.sqrt(sample_size))

# Step 2: Calculate the p-value for a one-tailed test
p_value = norm.cdf(z)

# Step 3: Print the results
print(f"Z-test statistic: {z:.2f}")
print(f"P-value: {p_value:.4f}")

# Step 4: Conclusion based on significance level (alpha = 0.02)
alpha = 0.02
if p_value < alpha:
    print("Reject the null hypothesis: There is sufficient evidence to suggest that the average malfunction time is less than 5 years.")
else:
    print("Fail to reject the null hypothesis: There is not enough evidence to suggest that the average malfunction time is less than 5 years.")

```


### **Explanation:**

1. **Z-test Statistic:**
   - The z-test statistic tells us how many standard deviations the sample mean is away from the population mean.

2. **P-value:**
   - The p-value indicates the probability of observing a test statistic at least as extreme as the one calculated, assuming the null hypothesis is true.

3. **Decision:**
   - If the p-value is less than the significance level (0.02), we reject the null hypothesis, suggesting that the average malfunction time is less than 5 years.

### **Running the Code:**

After running the above code, it will output the z-test statistic, the p-value, and the conclusion based on the 2% significance level.

This approach will allow us to determine whether there is sufficient evidence to support the idea that the warranty period should be revised.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Given data
population_mean = 5
sample_mean = 4.8
sample_std = 0.50
sample_size = 40

# Step 1: Calculate the z-test statistic
z = (sample_mean - population_mean) / (sample_std / np.sqrt(sample_size))

# Step 2: Calculate the p-value for a one-tailed test
p_value = norm.cdf(z)

# Print the results
print(f"Z-test statistic: {z:.2f}")
print(f"P-value: {p_value:.4f}")

# Step 4: Plot the normal distribution and highlight the z-value
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Standard Normal Distribution', color='blue')

# Shade the rejection region
alpha = 0.02
critical_value = norm.ppf(alpha)
print(f"critical area: {critical_value}")
plt.fill_between(x, y, where=(x <= critical_value), color='red', alpha=0.3, label=f'Rejection Region (alpha = {alpha})')

# Plot the z-value
plt.axvline(z, color='green', linestyle='--', label=f'z-value = {z:.2f}')

# Labels and title
plt.title('Normal Distribution with Rejection Region')
plt.xlabel('Z')
plt.ylabel('Probability Density')
plt.legend()

# Show the plot
plt.show()

# Decision based on significance level (alpha = 0.02)
if p_value < alpha:
    print("Reject the null hypothesis: There is sufficient evidence to suggest that the average malfunction time is less than 5 years.")
else:
    print("Fail to reject the null hypothesis: There is not enough evidence to suggest that the average malfunction time is less than 5 years.")
```
