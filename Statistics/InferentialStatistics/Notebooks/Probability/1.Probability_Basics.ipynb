#**PROBABILITY**

### What is Probability?

 - Probability is a measure of the likelihood that a particular event will occur. It quantifies uncertainty and is expressed as a number between 0 and 1, where 0 indicates impossibility and 1 indicates certainty.

 - Example:
If you flip a fair coin, the probability of getting heads is 0.5 or 50%, because there are two equally likely outcomes (heads and tails).









## Terminologies in Probability

## 1.Random Experiment
 - A random experiment is a process or action that leads to one of several possible outcomes, where the outcome is not predictable in advance.

- **Example:**
Rolling a die is a random experiment. You can't predict which number (1-6) will appear.


## 2.Trial
- A trial is a single performance or observation of a random experiment.

- **Example:**
Each individual roll of a die is a trial.

## 3.Outcome
- An outcome is a possible result of a random experiment.

- **Example:**
In the random experiment of rolling a die, possible outcomes are 1, 2, 3, 4, 5, and 6.



## 4.Sample Space
 - The sample space is the set of all possible outcomes of a random experiment.

 - **Example:**
For rolling a die, the sample space is {1, 2, 3, 4, 5, 6}.


##  5.Event
- An event is a subset of the sample space. It consists of one or more outcomes.

- **Example:**
In rolling a die, an event might be getting an even number. The event includes the outcomes {2, 4, 6}.



### Examples in Context

1. **Random Experiment:** Flipping a coin.
   - **Trial:** Flipping the coin once.
   - **Outcome:** Heads or Tails.
   - **Sample Space:** {Heads, Tails}.
   - **Event:** Getting Heads.

2. **Random Experiment:** Drawing a card from a standard deck of 52 cards.
   - **Trial:** Drawing one card.
   - **Outcome:** Any one of the 52 cards.
   - **Sample Space:** {Ace of Spades, 2 of Spades, ..., King of Hearts}.
   - **Event:** Drawing a King. The event includes the outcomes {King of Spades, King of Clubs, King of Diamonds, King of Hearts}.

Understanding these basic terms helps in analyzing and solving problems in probability theory.
