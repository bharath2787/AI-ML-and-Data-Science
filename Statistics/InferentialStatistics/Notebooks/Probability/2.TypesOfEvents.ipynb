#**Types of Events in Probabiltiy**

### only read : independent , dependent and mutually exclusive events

In probability, events are categorized based on their characteristics and relationships. Here are the main types of events along with two examples for each:

## 1. Independent Events*****
Independent events are those whose outcomes do not affect one another.

- **Example 1**: Flipping a coin and rolling a die are independent events; the outcome of one does not influence the other.
  
- **Example 2**: Drawing a card from a deck and then flipping a coin are independent events, as the result of the coin flip does not affect the card drawn.

## 2. Dependent Events
Dependent events are those where the outcome of one event affects the outcome of another.

- **Example 1**: Drawing two cards from a deck without replacement is a dependent event; the probability of the second card drawn depends on the first card drawn.
  
- **Example 2**: If you have a bag of marbles and you draw one marble, the probability of drawing a specific color on the second draw depends on the color of the first marble drawn.

## 3. Mutually Exclusive Events  ****
Mutually exclusive events cannot occur at the same time.

- **Example 1**: When rolling a die, getting a 3 and getting a 5 are mutually exclusive events because both cannot happen simultaneously.
  
- **Example 2**: In a single coin toss, getting heads and tails are mutually exclusive events since you can only get one outcome at a time.
