# **Empirical vs. Theoretical Probability**

## Theoretical Probability

- **Definition:**
 - Theoretical probability is based on the assumption that all outcomes in the sample space are equally likely. It is calculated using known mathematical formulas and logical reasoning without the need for actual experiments.

- **Formula:**
$[ P(A) = \frac{\text{Number of favorable outcomes}}{\text{Total number of possible outcomes}} ]$

- **Characteristics:**
  - Relies on logical analysis and known mathematical models.
  - Does not require experimental data.
  - Assumes all outcomes are equally likely.
  - Useful in ideal or controlled scenarios.

- **Example:**
  - **Random Experiment:** Rolling a six-sided die.
  - **Theoretical Probability of rolling a 3:**
  $[ P(3) = \frac{1}{6} ]$

- **Advantages:**
  - Provides exact probabilities.
  - Easy to calculate for simple events with equally likely outcomes.

- **Disadvantages:**
  - May not accurately reflect real-world situations where outcomes are not equally likely.
  - Assumes ideal conditions that may not exist in practice.

## Empirical Probability

- **Definition:**
Empirical probability, also known as experimental or observed probability, is based on actual data from experiments or historical observations. It is calculated as the ratio of the number of times an event occurs to the total number of trials.

- **Formula:**
$[ P(A) = \frac{\text{Number of times event A occurs}}{\text{Total number of trials}} ]$

- **Characteristics:**
  - Relies on real-world data and experimentation.
  - Requires conducting experiments or collecting historical data.
  - Reflects the observed frequency of events.
  - Useful in practical scenarios where theoretical models are not applicable.

- **Example:**
  - **Experiment:** Flipping a coin 100 times.
  - **Observed Data:** Heads appear 48 times.
  - **Empirical Probability of heads:**
  $[ P(\text{Heads}) = \frac{48}{100} = 0.48 ]$

- **Advantages:**
  - Reflects real-world situations and outcomes.
  - Can be applied to complex events where theoretical calculations are difficult or impossible.

- **Disadvantages:**
  - Requires a large number of trials to be accurate.
  - Results can be influenced by experimental error or bias.
  - May not provide exact probabilities, only estimates based on observed data.

## Comparison

| Feature               | Theoretical Probability                     | Empirical Probability                      |
|-----------------------|---------------------------------------------|-------------------------------------------|
| Basis                 | Mathematical reasoning                      | Experimental data                         |
| Calculation           | Known formulas                              | Observed frequencies                      |
| Accuracy              | Exact for ideal conditions                  | Estimates based on real data              |
| Requirement           | Assumes equally likely outcomes             | Requires conducting experiments           |
| Application           | Ideal or controlled scenarios               | Practical, real-world scenarios           |
| Example Calculation   | Rolling a die: $( \frac{1}{6} )$             | Flipping a coin 100 times: $( \frac{48}{100} )$  |

### Summary

- **Theoretical Probability**: Ideal and exact, useful for simple and controlled experiments where outcomes are equally likely.
- **Empirical Probability**: Practical and data-driven, useful for real-world scenarios where outcomes may not be equally likely and where experimental or historical data is available.

Understanding the distinction between these two types of probability helps in selecting the appropriate method for different situations, whether dealing with theoretical models or analyzing real-world data.

#**Exmple**

### Example: Flipping a Coin

Let's use the example of flipping a coin to explain the difference between theoretical and empirical probability.

## Theoretical Probability

- **Definition:**
Theoretical probability is calculated based on the assumption that all outcomes are equally likely. For a fair coin, there are two possible outcomes: heads and tails. Each outcome is equally likely.

- **Calculation:**
  - **Total number of possible outcomes:** 2 (heads, tails)
  - **Number of favorable outcomes (e.g., getting heads):** 1

- **Formula:**
$[ P(\text{Heads}) = \frac{\text{Number of favorable outcomes}}{\text{Total number of possible outcomes}} ]$

- **Calculation:**
$[ P(\text{Heads}) = \frac{1}{2} = 0.5 ]$

 - So, the theoretical probability of getting heads when flipping a fair coin is 0.5 or 50%.

## Empirical Probability

- **Definition:**
Empirical probability is calculated based on actual data from experiments or observations. It reflects the observed frequency of an event occurring.

- **Experiment:**
  - Flip a coin 100 times and record the outcomes.

- **Observed Data:**
  - Heads appeared 45 times.
  - Tails appeared 55 times.

- **Calculation:**
  - **Total number of trials:** 100
  - **Number of times heads appeared:** 45

- **Formula:**
$[ P(\text{Heads}) = \frac{\text{Number of times event A occurs}}{\text{Total number of trials}} ]$

- **Calculation:**
$[ P(\text{Heads}) = \frac{45}{100} = 0.45 ]$

- So, the empirical probability of getting heads based on 100 coin flips is 0.45 or 45%.

## Comparison

| Aspect                | Theoretical Probability                     | Empirical Probability                      |
|-----------------------|---------------------------------------------|-------------------------------------------|
| Basis                 | Assumes equally likely outcomes             | Based on observed data from experiments   |
| Calculation           | Uses known formulas                         | Uses observed frequencies                 |
| Example Probability   | \( \frac{1}{2} = 0.5 \)                     | \( \frac{45}{100} = 0.45 \)               |
| Accuracy              | Exact for ideal conditions                  | Reflects real-world observations          |
| Requirement           | No need for actual experiments              | Requires conducting experiments           |
| Application           | Ideal and controlled scenarios              | Practical and real-world scenarios        |

## Summary

- **Theoretical Probability** assumes a fair coin with equally likely outcomes, giving a probability of 0.5 for heads.
- **Empirical Probability** is derived from actual data, in this case, 100 coin flips resulting in 45 heads, giving a probability of 0.45.

The difference lies in how each type of probability is determined: theoretical probability is based on known mathematical principles and assumes perfect conditions, while empirical probability relies on observed data from actual experiments, reflecting real-world variations and imperfections.
