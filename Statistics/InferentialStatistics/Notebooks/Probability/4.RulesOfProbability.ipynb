

# Probability Rules!

- **A probability is a number between 0 and 1, inclusive.**
  - For any event A, $( 0 \leq P(A) \leq 1)$

- **The probability of the set of all possible outcomes must be 1.**
  - $( P(\text{Sample Space}) = P(S) = 1 )$

- **The complement of \( P(A) \) is the probability that A does NOT occur.**
  - Complement of A: $( P(A^c) = 1 - P(A) )$
  - Using algebra we also get
    - $( P(A) + P(A^c) = 1 )$
    - $( P(A) = 1 - P(A^c) )$



- **Notation**

  - $( P(A^c) = P(A') = P(\sim A) = P(\text{not A}) )$

Probability is a branch of mathematics that deals with quantifying uncertainty and analyzing random events. Here are the fundamental rules of probability:

## Fundamental Rules

### 1. Complement Rule
The probability of an event A not occurring (the complement of A) is given by:
$$
P(\text{not } A) = 1 - P(A)
$$
For instance, if the probability of rain tomorrow is 0.3, then the probability of no rain is:
$$
P(\text{not rain}) = 1 - 0.3 = 0.7
$$

### 2. Addition Rule
For two mutually exclusive events A and B (events that cannot occur simultaneously), the probability of either A or B occurring is:
$$
P(A \cup B) = P(A) + P(B)
$$
$$
P(A or B) = P(A) + P(B)
$$
If A and B are not mutually exclusive, the formula adjusts to account for their intersection:
$$
P(A \cup B) = P(A) + P(B) - P(A \cap B)
$$
$$
 P(A or B) = P(A) + P(B) - P(A and B)
 $$
### 3. Multiplication Rule
For two independent events A and B (the occurrence of one does not affect the other), the probability of both A and B occurring is:
$$
P(A \cap B) = P(A) \times P(B)
$$
If A and B are dependent events, the formula becomes:
$$
P(A \cap B) = P(A) \times P(B | A)
$$
where $$ P(B | A) $$ is the conditional probability of B given A.

### 4. Conditional Probability
The probability of event A occurring given that event B has occurred is defined as:
$$
P(A | B) = \frac{P(A \cap B)}{P(B)}
$$

## Practical Examples

- **Example of Complement Rule**: If the probability of a student passing an exam is 0.85, then the probability of failing is:
$$
P(\text{fail}) = 1 - 0.85 = 0.15
$$

- **Example of Addition Rule**: If the probability of rolling a 2 (P(A) = 1/6) and rolling a 3 (P(B) = 1/6) on a die, then:
$$
P(A \cup B) = P(A) + P(B) = \frac{1}{6} + \frac{1}{6} = \frac{2}{6} = \frac{1}{3}
$$

- **Example of Multiplication Rule**: If the probability of drawing a red card from a standard deck is 26/52 and then drawing another red card without replacement is 25/51, then:
$$
P(\text{two red cards}) = P(\text{first red}) \times P(\text{second red | first red}) = \frac{26}{52} \times \frac{25}{51}
$$
