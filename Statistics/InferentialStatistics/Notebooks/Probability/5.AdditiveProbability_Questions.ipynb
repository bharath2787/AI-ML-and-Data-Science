Here are five examples illustrating the additive rule of probability, along with their solutions:

### Example 1: Rolling a Die
**Problem**: What is the probability of rolling a 2 or a 5 on a six-sided die?

**Solution**:
- The probability of rolling a 2, $$ P(2) = \frac{1}{6} $$
- The probability of rolling a 5, $$ P(5) = \frac{1}{6} $$

Since rolling a 2 and rolling a 5 are mutually exclusive events:

$$
P(2 \text{ or } 5) = P(2) + P(5) = \frac{1}{6} + \frac{1}{6} = \frac{2}{6} = \frac{1}{3}
$$

### Example 2: Spinner
**Problem**: A spinner is divided into four equal sections: red, blue, green, and yellow. What is the probability of landing on red or blue?

**Solution**:
- The probability of landing on red, $$ P(\text{red}) = \frac{1}{4} $$
- The probability of landing on blue, $$ P(\text{blue}) = \frac{1}{4} $$

These events are mutually exclusive:

$$
P(\text{red or blue}) = P(\text{red}) + P(\text{blue}) = \frac{1}{4} + \frac{1}{4} = \frac{2}{4} = \frac{1}{2}
$$

### Example 3: Drawing Marbles
**Problem**: A jar contains 1 red, 3 green, and 4 yellow marbles. What is the probability of drawing a red or a green marble?

**Solution**:
- Total marbles = 1 + 3 + 4 = 8
- The probability of drawing a red marble, $$ P(\text{red}) = \frac{1}{8} $$
- The probability of drawing a green marble, $$ P(\text{green}) = \frac{3}{8} $$

These events are mutually exclusive:

$$
P(\text{red or green}) = P(\text{red}) + P(\text{green}) = \frac{1}{8} + \frac{3}{8} = \frac{4}{8} = \frac{1}{2}
$$

### Example 4: Students in a Class
**Problem**: In a class of 30 students, 12 study Math, 15 study Science, and 5 study both subjects. What is the probability that a student studies either Math or Science?

**Solution**:
- Let $ A $ be the event of studying Math and $ B $ be the event of studying Science.
- $ P(A) = \frac{12}{30} $, $ P(B) = \frac{15}{30} $, and $ P(A \cap B) = \frac{5}{30} $.

Using the general addition rule for non-mutually exclusive events:

$$
P(A \cup B) = P(A) + P(B) - P(A \cap B) = \frac{12}{30} + \frac{15}{30} - \frac{5}{30} = \frac{22}{30} = \frac{11}{15}
$$

### Example 5: Weather Events
**Problem**: The probability of rain on Saturday is 0.4, and the probability of rain on Sunday is 0.5. If the probability of rain on both days is 0.2, what is the probability of rain on either Saturday or Sunday?

**Solution**:
- Let $ A $ be the event of rain on Saturday and $ B $ be the event of rain on Sunday.
- $ P(A) = 0.4 $, $ P(B) = 0.5 $, and $ P(A \cap B) = 0.2 $.

Using the addition rule:

$$
P(A \cup B) = P(A) + P(B) - P(A \cap B) = 0.4 + 0.5 - 0.2 = 0.7
$$

### Example 7: Non-Mutually Exclusive Events

**Question**: What is the probability of drawing either a King or a red card from a standard deck of 52 cards?

**Solution**:
- There are 4 Kings in a deck, so $( P(\text{King}) = \frac{4}{52} = \frac{1}{13} )$.
- There are 26 red cards in a deck, so $( P(\text{Red card}) = \frac{26}{52} = \frac{1}{2} )$.
- There are 2 red Kings in a deck, so $( P(\text{Red King}) = \frac{2}{52} = \frac{1}{26} )$.

Since drawing a King and drawing a red card are not mutually exclusive:
$[ P(\text{King or Red card}) = P(\text{King}) + P(\text{Red card}) - P(\text{Red King}) ]$
$[ P(\text{King or Red card}) = \frac{1}{13} + \frac{1}{2} - \frac{1}{26} ]$
$[ P(\text{King or Red card}) = \frac{2}{26} + \frac{13}{26} - \frac{1}{26} = \frac{14}{26} = \frac{7}{13} ]$

### Example 8: Mutually Exclusive Events

**Question**: What is the probability of rolling a 2 or a 5 on a six-sided die?

### Example 9: Non-Mutually Exclusive Events

**Question**: What is the probability of drawing either a spade or a face card (Jack, Queen, King) from a standard deck of 52 cards?

**Solution**:
- There are 13 spades in a deck, so $( P(\text{Spade}) = \frac{13}{52} = \frac{1}{4} )$.
- There are 12 face cards in a deck, so $( P(\text{Face card}) = \frac{12}{52} = \frac{3}{13} )$.
- There are 3 spade face cards (Jack, Queen, King of spades), so $( P(\text{Spade face card}) = \frac{3}{52} )$.

Since drawing a spade and drawing a face card are not mutually exclusive:
$[ P(\text{Spade or Face card}) = P(\text{Spade}) + P(\text{Face card}) - P(\text{Spade face card}) ]$
$[ P(\text{Spade or Face card}) = \frac{1}{4} + \frac{3}{13} - \frac{3}{52}]$
$[ P(\text{Spade or Face card}) = \frac{13}{52} + \frac{12}{52} - \frac{3}{52} = \frac{22}{52} = \frac{11}{26} ]$

### Example 10: Mutually Exclusive Events

**Question**: What is the probability of drawing either a 7 or a 9 from a deck of cards numbered from 1 to 10?

**Solution**:
- There is 1 card numbered 7, so $( P(\text{7}) = \frac{1}{10} )$.
- There is 1 card numbered 9, so $( P(\text{9}) = \frac{1}{10} )$.

Since drawing a 7 and drawing a 9 are mutually exclusive events:
$[ P(\text{7 or 9}) = P(\text{7}) + P(\text{9}) = \frac{1}{10} + \frac{1}{10} = \frac{2}{10} = \frac{1}{5} ]$

These examples illustrate the application of the additive rule of probability for both mutually exclusive and non-mutually exclusive events.

Here are five additional examples illustrating the additive rule of probability, complete with solutions:

### Example 11: Drawing Cards
**Problem**: A standard deck of cards has 52 cards. What is the probability of drawing either a heart or a queen?

**Solution**:
- There are 13 hearts and 4 queens in a deck. However, one of the queens is also a heart (the queen of hearts).
- Let $$ A $$ be the event of drawing a heart and $$ B $$ be the event of drawing a queen.
- $$ P(A) = \frac{13}{52} $$ and $$ P(B) = \frac{4}{52} $$.
- Since the queen of hearts is counted in both events, we need to subtract it once:

$$
P(A \cup B) = P(A) + P(B) - P(A \cap B) = \frac{13}{52} + \frac{4}{52} - \frac{1}{52} = \frac{16}{52} = \frac{4}{13}
$$

### Example 12: Tossing Coins
**Problem**: If you toss two coins, what is the probability of getting either two heads or one head?

**Solution**:
- The sample space when tossing two coins is: {HH, HT, TH, TT}.
- Let $$ A $$ be the event of getting two heads (HH) and $$ B $$ be the event of getting one head (HT or TH).
- $$ P(A) = \frac{1}{4} $$ and $$ P(B) = \frac{2}{4} = \frac{1}{2} $$.
- The events are mutually exclusive:

$$
P(A \cup B) = P(A) + P(B) = \frac{1}{4} + \frac{1}{2} = \frac{1}{4} + \frac{2}{4} = \frac{3}{4}
$$

### Example 13: Rolling Two Dice
**Problem**: What is the probability of rolling a total of 2 or a total of 12 with two six-sided dice?

**Solution**:
- The only way to roll a total of 2 is to roll a (1,1), so $$ P(2) = \frac{1}{36} $$.
- The only way to roll a total of 12 is to roll a (6,6), so $$ P(12) = \frac{1}{36} $$.
- Since these events are mutually exclusive:

$$
P(2 \text{ or } 12) = P(2) + P(12) = \frac{1}{36} + \frac{1}{36} = \frac{2}{36} = \frac{1}{18}
$$

### Example 14: Picking Fruits
**Problem**: A basket contains 5 apples, 3 bananas, and 2 oranges. What is the probability of picking either an apple or a banana?

**Solution**:
- Total fruits = 5 + 3 + 2 = 10.
- The probability of picking an apple, $$ P(\text{apple}) = \frac{5}{10} = \frac{1}{2} $$.
- The probability of picking a banana, $$ P(\text{banana}) = \frac{3}{10} $$.
- These events are mutually exclusive:

$$
P(\text{apple or banana}) = P(\text{apple}) + P(\text{banana}) = \frac{1}{2} + \frac{3}{10} = \frac{5}{10} + \frac{3}{10} = \frac{8}{10} = \frac{4}{5}
$$

### Example 15: Sports Teams
**Problem**: In a tournament, the probability of Team A winning is 0.6, and the probability of Team B winning is 0.4. If both teams cannot win at the same time, what is the probability that either Team A or Team B wins?

**Solution**:
- Since Team A and Team B are mutually exclusive events:

$$
P(A \text{ or } B) = P(A) + P(B) = 0.6 + 0.4 = 1.0
$$

This means that one of the two teams will definitely win.
