
### Example 1: Coin Flips
**Problem**: What is the probability of flipping two heads in two consecutive coin flips?

**Solution**:
- The probability of getting heads on one flip, $$ P(H) = \frac{1}{2} $$.
- Since the flips are independent, the probability of getting heads on both flips is:
  
$$
P(H \cap H) = P(H) \times P(H) = \frac{1}{2} \times \frac{1}{2} = \frac{1}{4}.
$$

**Explanation**: Each flip does not affect the other, so we multiply their individual probabilities.

### Example 2: Rolling Dice
**Problem**: What is the probability of rolling a 3 on a six-sided die followed by rolling a 5 on another die?

**Solution**:
- The probability of rolling a 3, $$ P(3) = \frac{1}{6} $$.
- The probability of rolling a 5, $$ P(5) = \frac{1}{6} $$.
  
$$
P(3 \cap 5) = P(3) \times P(5) = \frac{1}{6} \times \frac{1}{6} = \frac{1}{36}.
$$

**Explanation**: The rolls are independent events, so we multiply the probabilities of each event occurring.

### Example 3: Drawing Cards
**Problem**: What is the probability of drawing an Ace from a standard deck of cards and then drawing a King without replacement?

**Solution**:
- The probability of drawing an Ace, $$ P(A) = \frac{4}{52} $$.
- After drawing an Ace, there are now 51 cards left, with 4 Kings remaining, so $$ P(K | A) = \frac{4}{51} $$.

Using the general multiplication rule:

$$
P(A \cap K) = P(A) \times P(K | A) = \frac{4}{52} \times \frac{4}{51} = \frac{16}{2652} = \frac{4}{663}.
$$

**Explanation**: The second event's probability changes based on the outcome of the first event, indicating that these events are dependent.

### Example 4: Marbles in a Bag
**Problem**: A bag contains 3 red and 2 blue marbles. What is the probability of drawing two red marbles in succession without replacement?

**Solution**:
- The probability of drawing the first red marble, $$ P(R_1) = \frac{3}{5} $$.
- After drawing one red marble, there are now 2 red marbles left out of 4 total marbles, so $$ P(R_2 | R_1) = \frac{2}{4} = \frac{1}{2} $$.

Using the general multiplication rule:

$$
P(R_1 \cap R_2) = P(R_1) \times P(R_2 | R_1) = \frac{3}{5} \times \frac{1}{2} = \frac{3}{10}.
$$

**Explanation**: The second event's probability is affected by the first event, showing they are dependent.

### Example 5: Weather Events
**Problem**: The probability of raining on a given day is 0.3. What is the probability that it rains for two consecutive days?

**Solution**:
- Let $$ P(\text{Rain}) = 0.3 $$.
- Since the days are independent:

$$
P(\text{Rain on Day 1} \cap \text{Rain on Day 2}) = P(\text{Rain}) \times P(\text{Rain}) = 0.3 \times 0.3 = 0.09.
$$

**Explanation**: The days are independent events, so we multiply the probabilities directly.

### Example 7: Picking Fruits
**Problem**: A basket contains 4 apples and 6 oranges. What is the probability of picking an apple followed by an orange without replacement?

- After picking an apple, there are now 9 fruits left, so $$ P(O | A) = \frac{6}{9} = \frac{2}{3} $$.

$$
P(A \cap O) = P(A) \times P(O | A) = \frac{2}{5} \times \frac{2}{3} = \frac{4}{15}.
$$

### Example 8: Sports Outcomes
**Problem**: The probability that Team A wins a match is 0.6, and the probability that Team B wins the next match is 0.5. What is the probability that both teams win their respective matches?

**Solution**:
- Assuming the matches are independent:
  
$$
P(A \cap B) = P(A) \times P(B) = 0.6 \times 0.5 = 0.3.
$$

**Explanation**: The outcomes of the matches are independent, allowing us to multiply the probabilities directly.
