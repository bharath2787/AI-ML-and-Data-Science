
**Support**, **Confidence**, and **Lift** are the **3 core concepts** that truly capture the **essence of how the Apriori algorithm works**.
---

###  1. **Support**
**How often does an item or itemset appear in the transactions?**

*  *"Out of 100 customers, how many bought Bread & Butter?"*

 **Formula:**  
$[
\text{Support}(A \rightarrow B) = \frac{\text{Transactions containing both A and B}}{\text{Total transactions}}
]$

---

###  2. **Confidence**
**When A is bought, how often is B also bought?**  
It shows the **strength of implication**.

>  *"Out of those who bought Bread, how many also bought Butter?"*

 **Formula:**  
$[
\text{Confidence}(A \rightarrow B) = \frac{\text{Support}(A \cap B)}{\text{Support}(A)}
]$

---

###  3. **Lift**
**How much more likely is B to be bought when A is bought, compared to random chance?**

>  *"Is Butter really associated with Bread, or is it just coincidence?"*

 **Formula:**  
$[
\text{Lift}(A \rightarrow B) = \frac{\text{Confidence}(A \rightarrow B)}{\text{Support}(B)}
]$

- **Lift > 1** → A and B are **positively associated** (more likely to be bought together).
- **Lift = 1** → A and B are **independent** (no special relationship).
- **Lift < 1** → A and B are **negatively associated** (less likely to occur together).

---

###  Summary Table

| Metric     | Meaning                                      | Good Value      |
|------------|----------------------------------------------|------------------|
| **Support**    | How frequent an itemset is                   | Depends on data |
| **Confidence** | How often B is bought when A is bought       | High (> 0.6)    |
| **Lift**       | Strength of association beyond random chance | > 1 is good     |

### **confidence is a chance and lift is a proof**


---

###  **Confidence is a chance**  
> “Given that item A was bought, what’s the **chance** that item B is also bought?”

It tells us:
- How frequently B follows A
- But **not whether A and B are related**  
- Can be **misleading** if B is just very common

---

###  **Lift is a proof**  
> “Let me **prove** that buying A actually affects buying B — not just by chance.”

It checks:
- If the confidence is **better than random**
- If A and B are truly **associated**, not just coincidentally co-occurring

---

### Simple Analogy:

- **Confidence**: “80% of people who drink coffee also eat donuts.”
- **Lift**: “Eating donuts is **2x more likely** if someone drinks coffee — that's not random.”
