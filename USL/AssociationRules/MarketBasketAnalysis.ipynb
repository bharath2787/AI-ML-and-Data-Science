###  **Association Rules**

These are **if-then statements** that help capture relationships between items.

Each rule has 3 key metrics:

1. **Support**  
   - How frequently the items appear together in the dataset.  
   - Example: 20% of all transactions include both bread and butter.

2. **Confidence**  
   - How often the rule has been found to be true.  
   - Example: In 80% of cases where bread is bought, butter is also bought.

3. **Lift**  
   - How much more likely the two items are to be bought together compared to if they were independent.  
   - Lift > 1 means a **positive association**.
