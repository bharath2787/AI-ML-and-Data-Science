```python
# Import necessary libraries
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
```

```python

# Step 1: Load the Iris dataset
iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target
```

```python
iris_df
```

```python


# Step 2: Standardize the features
scaler = StandardScaler()
scaled_data = scaler.fit_transform(iris_df.drop('species', axis=1))
```

```python
scaled_data
```

```python

# Step 3: Apply PCA
pca = PCA(3)  # We want to visualize in 3D later
pca_result = pca.fit_transform(scaled_data)
```

```python
scaled_data
```

```python
pca_result
```

```python
# Create a DataFrame for PCA results
pca_df = pd.DataFrame(data=pca_result, columns=['PC1', 'PC2', 'PC3'])
pca_df['species'] = iris_df['species']
```

```python
pca_df
```

```python
# Step 4: Visualizations

# 1D Plot: Plot the first principal component
plt.figure(figsize=(10, 6))
sns.boxplot(x=pca_df['PC1'], y=pca_df['species'], palette='Set1')
plt.title('1D PCA Visualization: PC1')
plt.xlabel('Principal Component 1 (PC1)')
plt.ylabel('Species')
plt.show()
```

```python
# 2D Plot: Plot the first two principal components
plt.figure(figsize=(10, 6))
sns.scatterplot(x='PC1', y='PC2', hue='species', palette='Set1', data=pca_df)
plt.title('2D PCA Visualization: PC1 vs PC2')
plt.xlabel('Principal Component 1 (PC1)')
plt.ylabel('Principal Component 2 (PC2)')
plt.legend(title='Species')
plt.show()
```

```python






# 3D Plot: Plot the first three principal components
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(pca_df['PC1'], pca_df['PC2'], pca_df['PC3'], c=pca_df['species'], cmap='Set1')
ax.set_title('3D PCA Visualization: PC1, PC2, PC3')
ax.set_xlabel('Principal Component 1 (PC1)')
ax.set_ylabel('Principal Component 2 (PC2)')
ax.set_zlabel('Principal Component 3 (PC3)')
plt.legend(*scatter.legend_elements(), title='Species')
plt.show()
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Iris dataset
iris = load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target

# Visualize the original data with pairplot
plt.figure(figsize=(12, 6))
sns.pairplot(iris_df, hue='species', palette='Set1')
plt.title('Original Iris Dataset Visualization')
plt.show()

# Standardize the features
scaler = StandardScaler()
scaled_data = scaler.fit_transform(iris_df.drop('species', axis=1))

# Apply PCA to reduce to 2 components
pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_data)

# Create a DataFrame for PCA results
pca_df = pd.DataFrame(data=pca_result, columns=['PC1', 'PC2'])
pca_df['species'] = iris_df['species']

# Visualize PCA result
plt.figure(figsize=(10, 6))
sns.scatterplot(x='PC1', y='PC2', hue='species', palette='Set1', data=pca_df)
plt.title('PCA Visualization of Iris Dataset')
plt.xlabel('Principal Component 1 (PC1)')
plt.ylabel('Principal Component 2 (PC2)')
plt.legend(title='Species')
plt.show()

# Side-by-side comparison using subplots
fig, axs = plt.subplots(1, 2, figsize=(16, 6))

# Original Data Visualization
sns.scatterplot(data=iris_df, x=iris_df.columns[0], y=iris_df.columns[1], hue='species', ax=axs[0], palette='Set1')
axs[0].set_title('Original Iris Dataset')
axs[0].set_xlabel(iris_df.columns[0])
axs[0].set_ylabel(iris_df.columns[1])

# PCA Data Visualization
sns.scatterplot(x='PC1', y='PC2', hue='species', data=pca_df, ax=axs[1], palette='Set1')
axs[1].set_title('PCA Visualization of Iris Dataset')
axs[1].set_xlabel('Principal Component 1 (PC1)')
axs[1].set_ylabel('Principal Component 2 (PC2)')

plt.legend(title='Species')
plt.tight_layout()
plt.show()
```



### Key Features of the Visualizations
- **Original Data Visualization**: This uses a pair plot to show the relationships between different features of the Iris dataset, colored by species. It gives a good idea of how the original data points are distributed.
- **PCA Visualization**: A scatter plot displays the reduced data in the space of the first two principal components, providing insights into how the data clusters after PCA.
- **Side-by-Side Comparison**: This allows direct comparison of the original data against the PCA results, illustrating how dimensionality reduction affects the visualization of species separability.

### Expected Outcome
- You should see how the different species of the Iris dataset are distributed in the original feature space and how they are represented in the PCA-reduced space.
- The PCA visualization may show clearer separation between species, demonstrating the effectiveness of PCA in capturing variance and simplifying the data structure.
