# Introduction to Matplotlib
"""
# Visualization with Matplotlib
Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. In this notebook, we will explore various types of plots and how to create them using Matplotlib.
"""


# Import necessary libraries

```python

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

```


# Basic Line Plot
"""
## Basic Line Plot
Let's start with a simple line plot. We will plot a sine wave.
"""


```python

# Create data for the plot
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Sine Wave')
plt.title('Basic Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()
```


# Scatter Plot
"""
## Scatter Plot
A scatter plot displays points representing values for typically two variables. Here, we will create a scatter plot using random data.
"""


```python

# Create data for the scatter plot
x = np.random.rand(50)
y = np.random.rand(50)

# Plot the data
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='r', label='Random Points')
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()
```


# Bar Plot
"""
## Bar Plot
Bar plots are used to represent categorical data with rectangular bars. Here, we will create a bar plot showing the count of different categories.
"""


```python

# Create data for the bar plot
categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 7, 3, 8, 4]

# Plot the data
plt.figure(figsize=(10, 6))
plt.bar(categories, values, color='b')
plt.title('Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.grid(True)
plt.show()


```


# Histogram
"""
## Histogram
Histograms are used to represent the distribution of a dataset. Here, we will create a histogram using random data.
"""


```python

# Create data for the histogram
data = np.random.randn(1000)

# Plot the data
plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, color='g', edgecolor='k')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

```

# Pie Chart
"""
## Pie Chart
Pie charts are used to represent the proportion of different categories. Here, we will create a pie chart showing the distribution of different categories.
"""


```python

# Create data for the pie chart
sizes = [15, 30, 45, 10]
labels = ['Category A', 'Category B', 'Category C', 'Category D']
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']

# Plot the data
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart')
plt.show()
```


# Box Plot
"""
## Box Plot
Box plots are used to represent the distribution of a dataset through their quartiles. Here, we will create a box plot using random data.
"""


```python

# Create data for the box plot
data = [np.random.rand(50) for _ in range(4)]

# Plot the data
plt.figure(figsize=(10, 6))
plt.boxplot(data, patch_artist=True)
plt.title('Box Plot')
plt.xlabel('Category')
plt.ylabel('Values')
plt.grid(True)
plt.show()
```
