#**HISTOGRAM**

 - A histogram is a way to visualize the distribution of a dataset by grouping the data into bins and plotting the frequency of data points in each bin.

 - A histogram is a graph showing frequency distributions.

 - It is a graph showing the number of observations within each given interval.

#**When to Use Histograms**
1.	**Understanding Data Distribution:**
 - 	Histograms are ideal for visualizing the distribution of a dataset. They help you understand the frequency of different ranges of values.
2.	**Identifying Skewness:**
 - 	They can help you identify if your data is skewed to the left or right. This is useful for understanding the nature of your data.
3.	**Spotting Outliers:**
 - 	Histograms can highlight outliers or unusual values that don't fit the overall pattern of the data.
4.	**Comparing Data Distributions:**
 - 	By creating multiple histograms side-by-side, you can compare the distributions of different datasets or different subsets of a single dataset.



#**What Kind of Data to Use with Histograms**
1.	**Continuous Data:**
 - 	Histograms are most effective with continuous data. This type of data can take any value within a given range, such as heights, weights, temperatures, or time durations.
2.	**Large Datasets:**
 - 	They work well with large datasets where summarizing the frequency of ranges of values provides meaningful insights.
3.	**Interval Data:**
 - 	Interval data, where the difference between values is meaningful, is also suitable for histograms. This includes test scores, financial figures, and measurement data.


#**Example Data Types for Histograms**
*	Height of people in a population
*	Scores on a test
*	Daily temperatures over a month
*	Age distribution in a community
*	Time taken to complete tasks
*	Sales figures over a period

#**Examples of When to Use Histograms**
 -	Analyzing Exam Scores:
  -	If you have scores from a large exam, a histogram can show you how the scores are distributed, if most students scored in the middle range, or if there are many high or low outliers.
 -	Financial Data Analysis:
  -	Histograms can be used to understand the distribution of sales over a year, helping to identify peak sales periods and seasonal trends.
 -	Quality Control in Manufacturing:
  - 	In manufacturing, histograms can help analyze the distribution of product measurements to ensure they meet quality standards.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
data.shape
```

```python
data.head()
```

```python
data['petal length (cm)'].min(), data['petal length (cm)'].max()
```

```python
plt.hist(data['petal length (cm)'])
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.title('Histogram of Petal Lengths for Iris Species')
```

#bins

```python
plt.hist(data['petal length (cm)'], bins=3)
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.title('Histogram of Petal Lengths for Iris Species')
```

```python
plt.hist(data['petal length (cm)'], bins=13)
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.title('Histogram of Petal Lengths for Iris Species')
```

#Relativewidth

rwidth=0.95:

This parameter controls the relative width of the bars in the histogram. A value of 0.95 means the bars will take up 95% of the available space for each bin, leaving a small gap between bars. If rwidth=1.0, the bars would touch each other with no gaps.

```python
plt.hist(data['petal length (cm)'],bins=3,rwidth=0.44)
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.title('Histogram of Petal Lengths for Iris Species')
```

```python
plt.hist(data['petal length (cm)'],bins=[0,2,5,7],rwidth=0.9, color='#95f414')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.title('Histogram of Petal Lengths for Iris Species')
```

```python
# histtype{'bar', 'barstacked', 'step', 'stepfilled'}, default: 'bar'
```

```python
plt.hist(data['petal length (cm)'],bins=[0,2,5,7],rwidth=0.9, color='r', histtype='step')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.title('Histogram of Petal Lengths for Iris Species')
```
