### 2D Plots

 - A 2D plot in matplotlib refers to a two-dimensional graphical representation of data, which is used to visualize the relationship between two variables. The matplotlib.pyplot module in matplotlib provides various functions to create 2D plots. Common types of 2D plots include line plots, scatter plots, bar plots, and histograms.

###When to Use a 2D Plot
 - 2D plots are used when you want to:

 - Visualize the relationship between two continuous variables.
 - Compare different data sets.
 - Show trends or patterns in data over time.
 - Display the distribution of data.

###Types of Data Used in 2D Plots
 - Numerical Data: Continuous data like height, weight, temperature, etc.
 - Categorical Data: Discrete data like product categories, types of fruits, etc.
 - Time Series Data: Data points collected or recorded at specific time intervals.
 - By using 2D plots in matplotlib, you can effectively visualize and interpret the relationships, trends, and distributions in your data.

 ### Types of 2D Plots and Their Usage
  1.**Line Plot (plt.plot())**:

  - Usage: Used to show trends over time or a sequence.
  - Example Data: Time series data, stock prices over time.

  2.**Scatter Plot (plt.scatter()):**

  - Usage: Used to show the relationship between two variables.
  - Example Data: Height vs. weight, age vs. income.

  3.**Bar Plot (plt.bar())**:

  - Usage: Used to compare different categories.
  - Example Data: Sales of different products, population of different countries.

  4.**Histogram (plt.hist())**:

  - Usage: Used to show the distribution of a single variable.
  - Example Data: Distribution of test scores, frequency of data points.

#LineStyles in Matplotlib
 - Simple linestyles can be defined using the strings "solid", "dotted", "dashed" or "dashdot".

 - More refined control can be achieved by providing a dash tuple
   - (offset, (on_off_seq)).
   - For example, (0, (3, 10, 1, 15)) means (3pt line, 10pt space, 1pt line, 15pt space) with no offset,
   - while (5, (10, 3)), means (10pt line, 3pt space), but skip the first 5pt line.

```python


linestyle_str = [
     ('solid', 'solid'),      # Same as (0, ()) or '-'
     ('dotted', 'dotted'),    # Same as (0, (1, 1)) or ':'
     ('dashed', 'dashed'),    # Same as '--'
     ('dashdot', 'dashdot')]  # Same as '-.'

linestyle_tuple = [
     ('loosely dotted',        (0, (1, 10))),
     ('dotted',                (0, (1, 1))),
     ('densely dotted',        (0, (1, 1))),
     ('long dash with offset', (5, (10, 3))),
     ('loosely dashed',        (0, (5, 10))),
     ('dashed',                (0, (5, 5))),
     ('densely dashed',        (0, (5, 1))),

     ('loosely dashdotted',    (0, (3, 10, 1, 10))),
     ('dashdotted',            (0, (3, 5, 1, 5))),
     ('densely dashdotted',    (0, (3, 1, 1, 1))),

     ('dashdotdotted',         (0, (3, 5, 1, 5, 1, 5))),
     ('loosely dashdotdotted', (0, (3, 10, 1, 10, 1, 10))),
     ('densely dashdotdotted', (0, (3, 1, 1, 1, 1, 1)))]


def plot_linestyles(ax, linestyles, title):
    X, Y = np.linspace(0, 100, 10), np.zeros(10)
    yticklabels = []

    for i, (name, linestyle) in enumerate(linestyles):
        ax.plot(X, Y+i, linestyle=linestyle, linewidth=1.5, color='black')
        yticklabels.append(name)

    ax.set_title(title)
    ax.set(ylim=(-0.5, len(linestyles)-0.5),
           yticks=np.arange(len(linestyles)),
           yticklabels=yticklabels)
    ax.tick_params(left=False, bottom=False, labelbottom=False)
    ax.spines[:].set_visible(False)

    # For each line style, add a text annotation with a small offset from
    # the reference point (0 in Axes coords, y tick value in Data coords).
    for i, (name, linestyle) in enumerate(linestyles):
        ax.annotate(repr(linestyle),
                    xy=(0.0, i), xycoords=ax.get_yaxis_transform(),
                    xytext=(-6, -12), textcoords='offset points',
                    color="blue", fontsize=8, ha="right", family="monospace")


fig, (ax0, ax1) = plt.subplots(2, 1, figsize=(10, 8), height_ratios=[1, 3])

plot_linestyles(ax0, linestyle_str[::-1], title='Named linestyles')
plot_linestyles(ax1, linestyle_tuple[::-1], title='Parametrized linestyles')

plt.tight_layout()
plt.show()
```
