
# Using multiple figures and subplots



- plt.tight_layout(): This function automatically adjusts the spacing between subplots and elements like titles, axis labels, and tick labels to make the figure look better and avoid overlaps.

 - pad=4: The pad argument specifies the amount of padding (in points) between the figure’s edges and the subplots. The default value is pad=1.08, so setting pad=4 would increase the padding around the subplots, creating more space between them and the edges of the figure.

```python
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(8, 12)) # https://matplotlib.org/stable/gallery/subplots_axes_and_figures/figure_size_units.html
ax1.bar(cricket_data['PLAYER'],cricket_data[feature1])
ax1.set_title(feature1)
ax1.set_xlabel('Players')
ax1.set_ylabel('50s')

ax2.bar(cricket_data['PLAYER'],cricket_data[feature2])
ax2.set_title(feature2)

ax3.bar(cricket_data['PLAYER'],cricket_data[feature3])
ax3.set_title(feature3)

plt.tight_layout(pad=2)
plt.show()
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(4, 8),sharex=True) # https://matplotlib.org/stable/gallery/subplots_axes_and_figures/figure_size_units.html
ax1.bar(cricket_data['PLAYER'],cricket_data[feature1])
ax1.set_title(feature1)
ax1.set_xlabel('Players')
ax1.set_ylabel('50s')
ax2.bar(cricket_data['PLAYER'],cricket_data[feature2])
ax2.set_title(feature2)
ax3.bar(cricket_data['PLAYER'],cricket_data[feature3])
ax3.set_title(feature3)
plt.tight_layout(pad=1)

plt.show()
```



1. **`plt.subplots(3, 1)`**: This function creates a grid of subplots. The arguments `(3, 1)` indicate that you want 3 rows and 1 column of subplots, resulting in 3 vertical subplots stacked on top of each other.

2. **`fig`**: This is the **Figure** object, which represents the overall container for all the subplots. It holds the layout and any title or other global elements for the figure.

3. **`(ax1, ax2, ax3)`**: These are **Axes** objects, which represent the individual subplots. Each one corresponds to a different section of the figure. `ax1` is the top plot, `ax2` is the middle plot, and `ax3` is the bottom plot.

4. **`figsize=(10, 18)`**: This argument sets the size of the figure in inches. The width is 10 inches, and the height is 18 inches, which makes the figure tall enough to accommodate the three stacked subplots.

5. **`sharex=True`**: This argument makes all subplots share the same x-axis, so they will have a synchronized horizontal scale. This is useful when you want to compare the data across subplots with the same x-axis values.

In summary, this line of code creates a figure with three vertically stacked subplots, sharing the same x-axis, and sizes the overall figure to 10 inches wide and 18 inches tall.

#Logrithmic Scale

###Explanation of plt.subplot(1, 2, 1):
- 1 (nrows): This is the number of rows in the grid of subplots. In this case, 1 means that you want a single row of subplots.

- 2 (ncols): This is the number of columns in the grid of subplots. Here, 2 means you want two subplots in that row (side by side).

- 1 (index): This specifies which subplot you are currently working on. Subplots are numbered from left to right and top to bottom. So 1 refers to the first subplot in this layout.

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Generate some example data
x = np.linspace(1, 1000, 100)
y_linear = x
y_quadratic = x**2
y_exponential = np.exp(x / 100)

# Plot the data using a linear scale
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(x, y_linear, label='Linear')
plt.plot(x, y_quadratic, label='Quadratic')
plt.plot(x, y_exponential, label='Exponential')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Linear Scale')
plt.legend()

# Plot the same data using a logarithmic scale
plt.subplot(1, 2, 2)
plt.plot(x, y_linear, label='Linear')
plt.plot(x, y_quadratic, label='Quadratic')
plt.plot(x, y_exponential, label='Exponential')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('X-axis (log scale)')
plt.ylabel('Y-axis (log scale)')
plt.title('Logarithmic Scale')
plt.legend()

plt.tight_layout()
plt.show()
```
