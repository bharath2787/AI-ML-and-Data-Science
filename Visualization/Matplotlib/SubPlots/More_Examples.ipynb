```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the tips dataset
df = sns.load_dataset("tips")
```

```python


# Create a figure with subplots (1 row, 2 columns)
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Scatter plot with regression line on first subplot
sns.regplot(data=df, x="total_bill", y="tip", ax=axes[0])
axes[0].set_title("Total Bill vs Tip")

# Histogram of total_bill on second subplot
sns.histplot(data=df, x="total_bill", kde=True, ax=axes[1])
axes[1].set_title("Distribution of Total Bill")

# Adjust layout
plt.tight_layout()
plt.show()
```

```python


# Create a 2x2 subplot layout
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# First subplot: Scatter plot
sns.scatterplot(data=df, x="total_bill", y="tip", hue="sex", ax=axes[0, 0])
axes[0, 0].set_title("Scatter: Total Bill vs Tip")

# Second subplot: Box plot
sns.boxplot(data=df, x="day", y="total_bill", ax=axes[0, 1])
axes[0, 1].set_title("Box Plot: Total Bill per Day")

# Third subplot: Violin plot
sns.violinplot(data=df, x="day", y="tip", hue="sex", split=True, ax=axes[1, 0])
axes[1, 0].set_title("Violin Plot: Tips by Day")

# Fourth subplot: Bar plot
sns.barplot(data=df, x="day", y="total_bill", hue="sex", ax=axes[1, 1])
axes[1, 1].set_title("Bar Plot: Total Bill per Day by Gender")

# Adjust layout
plt.tight_layout()
plt.show()
```
