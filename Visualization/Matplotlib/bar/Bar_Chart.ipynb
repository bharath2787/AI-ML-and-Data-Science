#BAR CHART
 - Bar charts are a versatile and widely used visualization tool, ideal for representing categorical data
 - x-axis category

 - aggregate analysis of group
 - univariate and bivariate analysis




#BAR CHARTS TYPES:

**Vertical Bar Chart**: This is the most common type of bar chart

where bars are plotted vertically along the x-axis. Each bar represents a category and its height represents the value.

**Horizontal Bar Chart:** In this chart, bars are plotted horizontally along the y-axis. It's useful when you have long category names or want to emphasize the differences in values more effectively.

**Stacked Bar Chart:** This type of chart shows multiple bars stacked on top of each other for each category. It's used to show the total and the breakdown of each category's value.

**Grouped Bar Chart:** In a grouped bar chart, multiple bars for each category are grouped together side by side. It's useful for comparing values across categories.

#Here are some specific scenarios when bar charts are particularly useful:

 - Comparing Categories:
   - Use bar charts to compare the values of different categories side by side. This helps highlight differences and similarities across categories.
 - Visualizing Frequencies:

  - Bar charts are excellent for showing the frequency of occurrence of categories, such as the number of sales per product, the count of different types of incidents, or survey responses.
 - Displaying Trends Over Time:

  - If you have categorical data over time (e.g., sales per quarter, monthly expenses), bar charts can effectively show trends and changes over time.
 - Emphasizing Individual Values:

   - When you want to focus on individual values rather than trends or distributions, bar charts are more effective than line charts or scatter plots.
 - Part-to-Whole Relationships:

  - Stacked bar charts can illustrate how individual parts contribute to the whole, making them useful for showing composition.
 - Comparing Subcategories:

  - Grouped or clustered bar charts are useful when you need to compare subcategories within each main category, such as sales by product within different regions.
 - Ranking Data:

  - Bar charts are helpful for ranking data from highest to lowest or vice versa, such as ranking countries by population or ranking products by revenue.

#**Staked Bar Graph:**
 - The stacked bar chart (aka stacked bar graph) extends the standard bar chart from looking at numeric values across one categorical variable to two.

```python
# Showing the composition of 4s and 6s for top 5 players
plt.figure(figsize=(10, 6))
bottom = top_players['4s']
plt.bar(top_players['PLAYER'], top_players['4s'], color='lightblue', label='4s')
# plt.bar(top_players['PLAYER'], top_players['6s'], bottom=bottom, color='lightcoral', label='6s')
plt.xlabel('Player')
plt.ylabel('Number of 4s and 6s')
plt.title('Composition of 4s and 6s for Top 5 Players')
plt.legend()
plt.show()
```
