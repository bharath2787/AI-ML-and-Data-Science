#**PIE CHART**

A pie chart is a circular statistical graphic divided into slices to illustrate numerical proportions. Each slice of the pie represents a category's contribution to the whole. The size of each slice is proportional to the quantity it represents.

```python
import pandas as pd

# Load the CSV file
cricket_df = pd.read_csv('/content/drive/MyDrive/0.Latest_DS_Course/Visualization/Data/cricketStats.csv')
# Display the first few rows of the dataframe to understand its structure
cricket_df.head()
```

```python
import matplotlib.pyplot as plt

# Extract the top 5 players and their runs
top_players = cricket_df.head(5)
players = top_players['PLAYER']
runs = top_players['Runs']

# Plot the pie chart
plt.figure(figsize=(10, 7))
plt.pie(runs, labels=players, autopct='%1.1f%%', startangle=30)
plt.title('Distribution of Total Runs Scored by Top 5 Players')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()


# runs: Data for the pie chart (total runs).
# labels=players: Labels for each slice (player names).
# autopct='%1.1f%%': Display the percentage value on each slice.
# startangle=140: Starting angle for the pie chart.
# plt.axis('equal'): Ensures the pie chart is drawn as a circle.
```

The pie chart shows the distribution of total runs scored by the top 5 players in the dataset. Each slice of the pie represents the proportion of runs scored by each player, with the percentage of the total runs displayed on each slice.



```python
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'violet']
explode = (0.8, .2, .2, .2, .2)  # "Explode" the first slice (i.e., 'David Warner')

plt.figure(figsize=(10, 7))
plt.pie(runs, labels=players, colors=colors, autopct='%1.1f%%', startangle=120, explode=explode, shadow=True)
plt.title('Distribution of Total Runs Scored by Top 5 Players')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()


# Colors: You can specify colors for each slice using the colors parameter.
# Exploding Slices: Highlight a particular slice by "exploding" it using the explode parameter.
# Shadow: Add a shadow to the pie chart using the shadow parameter.
```
