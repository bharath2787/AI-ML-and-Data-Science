# Scatter Plot in Matplotlib

## Introduction
- A scatter plot is a type of plot that displays values for typically two variables for a set of data. If the points are color-coded, one additional variable can be displayed. Scatter plots are useful to identify potential relationships, detect outliers, show distribution, and reveal clusters in the data.

# Why Use Scatter Plots?
- Visualize Relationships: Identify potential relationships or correlations between two variables.
- Detect Outliers: Easily spot outliers or anomalies in the data.
- Show Distribution: Show how data points are distributed across the axes.
- Display Clusters: Reveal clusters or groupings in the data.

# When to Use Scatter Plots?
- When you have two continuous variables.
- When you want to explore the relationship between these two variables.
- When you want to see how one variable is affected by another.
- When you want to identify clusters, patterns, or trends within the data.

```python


## Import Libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# iris dataset link : https://archive.ics.uci.edu/dataset/53/iris
```

```python

# https://en.wikipedia.org/wiki/Iris_flower_data_set
## Load Dataset
# We'll use the Iris dataset, which contains measurements of iris flowers from three different species.
iris = sns.load_dataset('iris')
iris.head()
```

```python

## Basic Scatter Plot
# Create a scatter plot to visualize the relationship between sepal length and sepal width.
plt.scatter(iris['sepal_length'], iris['sepal_width'], c='blue')

# Add titles and labels
plt.title('Sepal Length vs Sepal Width')  # Title of the plot
plt.xlabel('Sepal Length (cm)')  # Label for the x-axis
plt.ylabel('Sepal Width (cm)')  # Label for the y-axis

# Show the plot
plt.show()
```


## Explanation
### In this basic scatter plot:
 - `plt.scatter(iris['sepal_length'], iris['sepal_width'])`: Creates the scatter plot with sepal length on the x-axis and sepal width on the y-axis.
 - `c='blue'`: Sets the color of the markers to blue.
 - `alpha=0.5`: Sets the transparency level of the markers to 0.5 (50% transparency), making it easier to see overlapping points.


```python
iris['species']
```

```python
iris['species'].astype('category').cat.codes
```

```python

## Enhanced Scatter Plot
# Enhance the scatter plot by color-coding points based on species and scaling marker sizes by petal length.
plt.figure(figsize=(12, 6))  # Set the figure size to make the plot larger
plt.scatter(iris['sepal_length'], iris['sepal_width'],
            c=iris['species'].astype('category').cat.codes,  # Color by species
            cmap='viridis',  # Use the 'viridis' colormap
            s=iris['petal_length']*20,  # Scale marker size by petal length
            alpha=0.7,  # Set transparency level to 0.7 (30% transparency)
            edgecolors='w',  # Set edge color to white
            linewidth=0.5)  # Set edge line width to 0.5

# Add titles and labels
plt.title('Sepal Length vs Sepal Width (Colored by Species, Sized by Petal Length)')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.colorbar(ticks=[0, 1, 2], label='Species')  # Add color bar with species labels

# Show the plot
plt.show()


```

# Explanation
In this enhanced scatter plot:
- `c=iris['species'].astype('category').cat.codes`: Color-codes the points based on the species of the iris flower. `astype('category').cat.codes` converts species names to numerical codes.
- `cmap='viridis'`: Sets the colormap to 'viridis', which is a color map that transitions from yellow to green to blue.
- `s=iris['petal_length']*20`: Sets the size of the markers to be 20 times the petal length, making larger petal lengths appear as larger markers.
- `alpha=0.7`: Sets the transparency level of the markers to 0.7 (30% transparency), making it easier to see overlapping points.
- `edgecolors='w'`: Sets the edge color of the markers to white, making the markers stand out more.
- `linewidth=0.5`: Sets the width of the edge lines to 0.5.

```python

## Customizing Scatter Plots
# You can further customize scatter plots by adding trend lines, different marker styles, and more.
import numpy as np

# Fit a trend line
z = np.polyfit(iris['sepal_length'], iris['sepal_width'], 1)  # Fit a polynomial of degree 1 (linear fit)
p = np.poly1d(z)  # Create a polynomial object

# Plot scatter plot with trend line
plt.figure(figsize=(12, 6))  # Set the figure size to make the plot larger
plt.scatter(iris['sepal_length'], iris['sepal_width'], c='blue', alpha=0.5)
plt.plot(iris['sepal_length'], p(iris['sepal_length']), "r--")  # Plot the trend line in red dashed style

# Add titles and labels
plt.title('Sepal Length vs Sepal Width with Trend Line')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')

# Show the plot
plt.show()


```

# Explanation
In this scatter plot with a trend line:
- `z = np.polyfit(iris['sepal_length'], iris['sepal_width'], 1)`: Fits a linear trend line to the data.
- `p = np.poly1d(z)`: Creates a polynomial object from the fit.
- `plt.plot(iris['sepal_length'], p(iris['sepal_length']), "r--")`: Plots the trend line in red dashed style.



## Conclusion
- Scatter plots are a versatile tool for visualizing relationships between two continuous variables. By enhancing them with color, size, and trend lines, you can uncover deeper insights and patterns in your data.


```python


# Map species names to numerical codes
species_to_code = iris['species'].astype('category').cat.categories
print(species_to_code)

# Enhanced Scatter Plot with Species Names
plt.figure(figsize=(12, 6))  # Set the figure size to make the plot larger
scatter = plt.scatter(iris['sepal_length'], iris['sepal_width'],
                      c=iris['species'].astype('category').cat.codes,  # Color by species
                      cmap='viridis',  # Use the 'viridis' colormap
                      s=iris['petal_length']*20,  # Scale marker size by petal length
                      alpha=0.7,  # Set transparency level to 0.7 (30% transparency)
                      edgecolors='w',  # Set edge color to white
                      linewidth=0.5)  # Set edge line width to 0.5

# Add titles and labels
plt.title('Sepal Length vs Sepal Width (Colored by Species, Sized by Petal Length)')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')

# Create a color bar with species labels
cbar = plt.colorbar(scatter, ticks=[0, 1, 2])
cbar.ax.set_yticklabels(species_to_code)
cbar.set_label('Species')

# Show the plot
plt.show()
```
