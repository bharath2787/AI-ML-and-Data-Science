```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
data.columns
```

```python
plt.hist(data['petal length (cm)'])
plt.title("histogram for petal length")
plt.xlabel("Petal Length")
plt.ylabel("Frequency")
```

```python
plt.hist(data['petal length (cm)'], bins=30)
plt.title("histogram for petal length")
plt.xlabel("Petal Length")
plt.ylabel("Frequency")
```

```python
data['petal length (cm)'].min(), data['petal length (cm)'].max()
```

```python
plt.hist(data['petal length (cm)'], bins=4, rwidth=1, color='r', histtype='step')
plt.title("histogram for petal length")
plt.xlabel("Petal Length")
plt.ylabel("Frequency")
```
