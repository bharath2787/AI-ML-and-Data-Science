- **Bar Plot:** Displays the mean (or another summary statistic) of a continuous variable for each category with error bars.
-**Count Plot:** Displays the count of observations for each category of a categorical variable.
-	**Point Plot:** Similar to a bar plot but uses points and lines to show the mean (or another summary statistic) of a continuous variable for each category.

###These plots are useful for different purposes:
-	Use a bar plot when you want to compare the average value of a continuous variable across categories.
-	Use a count plot when you want to visualize the frequency distribution of a categorical variable.
-	Use a point plot when you want to compare the average value of a continuous variable across categories and see the trend or change across categories.



#Bar Plot
 - A bar plot shows the relationship between a categorical variable and a continuous variable.
 - It displays the mean (or another summary statistic) of the continuous variable for each category, with error bars representing the confidence interval or standard deviation.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

####simple example for a bar plot

```python

# Step 1: Create the dataset
data = {
    'Student': [f'Student{i}' for i in range(1, 21)],
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Male', 'Female',
               'Female', 'Male', 'Male', 'Female', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male']
}
df = pd.DataFrame(data)
print("gneder count: ",df['Gender'].value_counts())
```

```python
df
```

```python
df.info()
```

```python
# Step 2: Plot the bar plot
plt.figure(figsize=(8, 6))
sns.countplot(x='Gender', data=df, palette='pastel')
plt.title('Gender Distribution in Class')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

#Count Plot
 - A count plot is a type of bar plot that shows the number of observations for each category of a categorical variable.
 - It is useful for visualizing the frequency distribution of categorical data.
 -**this plot tells you how is ur dataset divided along a categorical column(i.e. count of records per category in a given categorical column)**

```python
tips.info()
```

```python
# Create a count plot
plt.figure(figsize=(10, 6))
sns.countplot(x='day', data=tips, palette='muted')
plt.title('Count Plot Example')
plt.show()

# Explanation:

# x='day': Categorical variable for the x-axis.
# The y-axis shows the count of observations for each category.
```

```python
tips['day'].value_counts(normalize=True)
```

#Point Plot
 - A point plot is similar to a bar plot but uses points instead of bars to show the mean (or another summary statistic) of a continuous variable for each category.
 - It also connects the points with lines, making it easier to compare changes across categories.

```python
# Create a point plot
plt.figure(figsize=(10, 6))
sns.pointplot(x='day', y='total_bill', data=tips, estimator='mean')
plt.title('Point Plot Example')
plt.show()


# Explanation:

# x='day': Categorical variable for the x-axis.
# y='total_bill': Continuous variable for the y-axis.
# capsize=.2: Adds caps to the error bars, which represent the confidence interval by default.
```

```python
tips['time'].unique()
```

```python
# facet plots
sns.catplot(data=tips, x="day", y="total_bill", kind="point", col="time", row='smoker')
```

```python

# Create the catplot
g = sns.catplot(data=tips, x="day", y="total_bill", kind="bar", col="time", row='smoker')

# Set a global title
g.fig.suptitle("Total Bill by Day, Time, and Smoking Status", y=1.1)  # Adjust y for spacing
```

```python

# Create the catplot
g = sns.catplot(data=tips, x="day", y="total_bill", kind="bar", col="time", row='smoker')

# Set a title for each subplot
g.set_titles("{col_name} {row_name}")
```
