#Box Plot
-
A box plot (or box-and-whisker plot) is a graphical representation of the distribution of a dataset that displays the dataset's:
 -  minimum,
 - first quartile (Q1),
 - median,
 - third quartile (Q3),  
 - maximum values.

 It may also highlight outliers.

 The box represents the interquartile range (IQR), which is the range between Q1 and Q3, and the line inside the box shows the median.

 Whiskers extend from the box to the dataset's minimum and maximum values within 1.5 times the IQR from the quartiles.

 Outliers are typically shown as individual points beyond the whiskers.

##Types of Data for Box Plots
Box plots are used with continuous numerical data. Here are some specific scenarios:
1.	**Single Continuous Variable**: To show the distribution of one continuous variable.
**Example**: Distribution of total bills in a restaurant dataset.
2.	**Continuous Variable Across Categories**: To compare the distribution of a continuous variable across different categories or groups.
**Example**: Comparing total bills across different days of the week.



#When to Use a Box Plot
####A box plot is typically used when you need to:
1.	Visualize the Distribution of Data: To understand the spread, central tendency, and variability within your dataset.
2.	Compare Distributions Across Groups: To compare the distributions of the same variable across different categories or groups.
3.	Identify Outliers: To detect unusual data points that fall outside the expected range.
4.	Summarize Data: To provide a quick summary of the data's main statistical properties.



```python
import seaborn as sns
import matplotlib.pyplot as plt

# Example dataset
data = sns.load_dataset('tips')

```

```python
data.info()
```


# Box plot for a single variable


```python
data['total_bill'].max()
```

```python
# plt.figure(figsize=(10, 16))
sns.boxplot(x='total_bill', data=data)
plt.title('Box Plot of total bill')
plt.show()
```

```python
sns.histplot(x='total_bill', data=data)
```

```python
#hue
sns.boxplot(y='total_bill', data=data, hue='day')
```

```python
# Example dataset
tips = sns.load_dataset('tips')

# Box plot to compare groups
plt.figure(figsize=(10, 6))
sns.boxplot(x='day', y='total_bill', data=tips, palette='Dark2')
plt.title('Box Plot of Total Bill by Day')
plt.show()
```

```python
# sns.catplot
sns.catplot(x='day', y='total_bill', data=tips, palette='Set2', kind='box')
```

###Violin plots provide a more detailed view of the distribution of a single variable by combining a box plot and a kernel density estimate (KDE).

```python
sns.violinplot(data=tips, x='day', y='total_bill')
```

```python
#  palette='tab10'
sns.violinplot(data=tips, x='day', y='total_bill',palette='tab10')
```
