##Scatter plot:
 - Bivariate analysis
 - tells the relationship between two variables

Strip Plot and Swarm Plot are specialized scatter plots for categorical data, showing individual data points and their distribution within each category.

####Strip plot - Displays individual data points for one categorical variable against one numerical variable.

####Swarm plot - Similar to a strip plot but adjusts the data points to prevent overlap more effectively.

##Distribution Plot
- Univariate analysis

 - BoxPlot
 - Violin plot = BoxPlot + KDEPlot

####A box plot is typically used when you need to:
1.	Visualize the Distribution of Data: To understand the spread, central tendency, and variability within your dataset.
2.	Compare Distributions Across Groups: To compare the distributions of the same variable across different categories or groups.
3.	Identify Outliers: To detect unusual data points that fall outside the expected range.
4.	Summarize Data: To provide a quick summary of the data's main statistical properties.

###Estimates/Stats:

####Bar Plot
 - Bar plots are used to display the relationship between a categorical variable and a continuous variable. They provide a visual comparison of the size of different categories.

 - **When to Use**:
  - 	Comparing Categories: To compare the mean (or sum) of a continuous variable across different categories.
  - Highlighting Differences: To highlight differences between categories.

####Point Plot
 - Point plots are similar to bar plots but use points instead of bars to show the central tendency of the data. They are useful for comparing changes across categories.
 - **When to Use**:
  -	Trends Over Categories: To compare changes in central tendency (mean, median, etc.) across categories.
  -	Highlighting Variability: To highlight variability or differences between categories using error bars.



####Count Plot
 - Count plots are used to show the frequency of occurrence of each category in a categorical variable. They are essentially bar plots for categorical data.
 - **When to Use**:
  -	Frequency Distribution: To display the count of observations in each category.
  -	Understanding Category Distribution: To understand the distribution of data points across categories.
