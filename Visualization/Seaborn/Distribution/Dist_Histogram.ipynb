#Distribution Plots

#**HISTOGRAM**
 - used for univariate analysis
 - A histogram is a way to visualize the distribution of a dataset by grouping the data into bins and plotting the frequency of data points in each bin.

 - A histogram is a graph showing frequency distributions.

 - It is a graph showing the number of observations within each given interval.

#**Example Data Types for Histograms**
*	Height of people in a population
*	Scores on a test
*	Daily temperatures over a month
*	Age distribution in a community
*	Time taken to complete tasks
*	Sales figures over a period

#**Examples of When to Use Histograms**
 -	Analyzing Exam Scores:
  -	If you have scores from a large exam, a histogram can show you how the scores are distributed, if most students scored in the middle range, or if there are many high or low outliers.
 -	Financial Data Analysis:
  -	Histograms can be used to understand the distribution of sales over a year, helping to identify peak sales periods and seasonal trends.
 -	Quality Control in Manufacturing:
  - 	In manufacturing, histograms can help analyze the distribution of product measurements to ensure they meet quality standards.





```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load example dataset
tips = sns.load_dataset("tips")
```

```python
# Display the first few rows of the dataset
display(tips.head())
```

```python
tips.info()
```

```python
tips.columns
```

```python
tips['total_bill'].min(), tips['total_bill'].max()
```

```python
sns.histplot(data=tips, x='total_bill')
```

```python
sns.displot(data=tips, x='total_bill', kind='hist')
```

```python
tips.smoker.unique()
```

```python
#bins
sns.histplot(data=tips, x='total_bill', hue = 'smoker',bins=15)
```

```python
tips['day'].nunique(),tips['day'].unique()
```

```python
#hue
# sns.histplot(data=tips, x='total_bill', bins=8,hue='sex')

sns.kdeplot(data=tips, x='total_bill',hue='sex')
```

```python
sns.displot(data=tips, x='total_bill', bins=8,col='sex')
```

```python
tips.day.unique()
```

```python
#col
sns.displot(data=tips, x='total_bill',row='day',col='sex', kind='kde')

```

```python
sns.displot(data=tips, x='total_bill',row='day',col='sex', kind='hist')
```

```python
#element = step

sns.histplot(data=tips, x='total_bill', bins=8,hue='sex', element='step')
```

```python
tips['smoker'].nunique(), tips['day'].nunique()
```

```python
#facet plot

sns.displot(data=tips, x="total_bill", col='smoker', row='day', kde=True)
```
