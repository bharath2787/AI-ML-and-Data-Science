#Summary of Distribution Plots in Seaborn
##Why They Are Used
###Distribution plots in Seaborn are used to:

- Visualize the distribution of a dataset.
-	Understand the underlying pattern of data.
-	Identify the shape of the data distribution (e.g., normal, skewed).
-	Detect outliers, skewness, and other statistical characteristics.
-	Compare the distribution of multiple datasets.

##When They Are Used
###Distribution plots are used when:
-	You need to explore the distribution of a single variable.
-	You want to compare the distributions of different variables.
-	You are performing exploratory data analysis (EDA).
-	You need to visualize statistical properties of the data such as mean, median, and standard deviation.
-	You want to check the fit of a data distribution to a theoretical distribution.

##With What Type of Data They Are Used
###Distribution plots are used with:

-	Continuous data (e.g., age, income, height).
-	Discrete data (e.g., count data).
-	Data that can be categorized and binned for histogram analysis.

###Types of Distribution Plots in Seaborn
1.	Histogram (sns.histplot)
-	Displays the distribution of a single variable by dividing the data into bins and counting the number of observations in each bin.
-	Suitable for both continuous and discrete data.
2.	KDE Plot (sns.kdeplot)
-	Estimates the probability density function of a continuous variable.
-	Smooths the data using a kernel density estimation (KDE) technique.
