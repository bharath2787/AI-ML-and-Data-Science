#Kernal Density Estimation

```python
# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset
tips = sns.load_dataset("tips")

# Display the first few rows of the dataset
tips.head()
```

## KDE Plot in Seaborn

Kernel Density Estimate (KDE) plots are used to visualize the probability density function of a continuous variable. Let's explore how to create KDE plots using Seaborn's `sns.kdeplot` and explain each parameter.


```python
# Basic KDE Plot
sns.kdeplot(data=tips, x="total_bill")
```

### Parameters in `sns.kdeplot`

1. **data**: The dataset containing the variable to plot.
2. **x**: The variable for which to plot the KDE.
3. **fill**: If True, fills the area under the KDE curve.
4. **color**: The color of the KDE plot.
5. **hue**: Variable that defines subsets to draw multiple KDE plots.



## KDE Plot using `sns.displot`

Seaborn's `sns.displot` function is a more general function for creating distribution plots, and it also supports KDE plots. Here's how you can create a KDE plot using `sns.displot`.


```python
# KDE Plot using sns.displot
sns.displot(data=tips, x="total_bill", kind="kde")
```

```python
sns.displot(data=tips, x="total_bill", kind="kde", hue='smoker', fill=True)
```

```python
sns.displot(data=tips, x="total_bill", kind="kde", col='smoker', fill=False)
```

## Conclusion

KDE plots in Seaborn are a powerful tool for visualizing the distribution of continuous variables. By understanding and using the various parameters in `sns.kdeplot` and `sns.displot`, you can customize KDE plots to suit your data visualization needs.
