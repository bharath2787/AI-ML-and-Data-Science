##1. What is Seaborn?

####Seaborn is a Python visualization library based on Matplotlib that provides a high-level interface for drawing attractive and informative statistical graphics. It simplifies the process of creating complex visualizations with fewer lines of code and offers more aesthetically pleasing default styles.

##2. How does Seaborn differ from Matplotlib?
####Answer:Seaborn is built on top of Matplotlib and provides a higher-level interface for creating statistical plots. While Matplotlib is more flexible and allows for detailed customizations, Seaborn simplifies the process of creating complex visualizations and provides a more aesthetically pleasing default style.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset
tips = sns.load_dataset("tips")

# Create a FacetGrid
g = sns.FacetGrid(tips, col="day", row="sex", margin_titles=True)

# Map a bar plot to each subplot in the grid
g.map(sns.barplot, "time", "total_bill", ci=None)

# Add titles
g.set_axis_labels("Time of Day", "Average Total Bill")
g.set_titles(col_template="{col_name}", row_template="{row_name}")

# Adjust the plot to prevent overlapping titles
plt.subplots_adjust(top=0.9)
g.fig.suptitle('Average Total Bill by Day and Gender')

# Show the plot
plt.show()
```
