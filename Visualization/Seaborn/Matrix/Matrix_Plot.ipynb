#HEATMAP
 - Plot rectangular data as a color-encoded matrix

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the flights dataset
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python

# Pivot the dataset to get the matrix format
flights_pivot = flights.pivot(index='month', columns='year', values='passengers')

```

```python
flights_pivot
```

```python
# Create the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(flights_pivot)
plt.title("Heatmap of Flight Passengers Over the Years")
plt.show()
```

```python
# , cmap='YlGnBu'

# Create the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(flights_pivot,fmt='d', cmap='binary')
plt.title("Heatmap of Flight Passengers Over the Years")
plt.show()
```

ColorMaps:  https://matplotlib.org/stable/users/explain/colors/colormaps.html

```python
#Annotation

# Create the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(flights_pivot, annot=True,fmt='d', cmap='YlOrRd')
plt.title("Heatmap of Flight Passengers Over the Years")
plt.show()
```

```python
#fmt

# Create the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(flights_pivot, annot=True, fmt='.2f', cmap='YlGnBu')
plt.title("Heatmap of Flight Passengers Over the Years")
plt.show()
```

```python
# linewidth=.5
plt.figure(figsize=(12, 8))
sns.heatmap(flights_pivot, annot=True, fmt='.2f', cmap='YlGnBu',linewidth=2)
plt.savefig("heatmap.png")
```
