# Line Plots in Seaborn using Plotly Gapminder Dataset

 ## Introduction to Line Plots in Seaborn
 - Line plots are useful for visualizing time series data or trends over a continuous variable.
 - Seaborn provides a convenient interface for creating line plots, along with many customization options.


```python

import seaborn as sns
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
```

## Loading the Gapminder Dataset
- We will use the Gapminder dataset available in the Plotly library.




```python
# Load the dataset
gapminder = px.data.gapminder()
gapminder.head()
```

## Basic Line Plot with Seaborn
- Let's create a basic line plot showing the life expectancy over the years for a specific country.



```python

# Filter the data for a specific country
country_data = gapminder[gapminder['country'] == 'Canada']

```

```python
country_data.shape
```

```python
# Create the line plot
sns.lineplot(data=country_data, x='year', y='lifeExp')
plt.title('Life Expectancy in Canada Over Years')
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
plt.show()
```

## Explanation of Parameters in `seaborn.lineplot`

- 1. `data`: The data frame containing the data.
 - This parameter takes a pandas DataFrame that holds the data to be plotted.
 -  Example: `data=gapminder`

- 2. `x` and `y`: The names of the variables to be plotted on the x-axis and y-axis.
 -  Example: `x='year'`, `y='lifeExp'`



```python
sns.lineplot(data=country_data, x='year', y='lifeExp')
plt.title('Life Expectancy in Canada Over Years')
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
plt.show()
```


### 3. `hue`: Grouping variable that will produce lines with different colors.
 - Example: `hue='continent'`


```python
gapminder.continent.nunique(),gapminder.continent.unique(),
```

```python
gapminder['country'].unique()
```

```python
sns.lineplot(data=gapminder, x='year', y='lifeExp', hue='continent')
plt.title('Life Expectancy Over Years by Continent')
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
plt.show()
```

### 4. `style`: Grouping variable that will produce lines with different styles.
 -  Example: `style='continent'`


```python
three_countries_data.columns
```

```python
#We have data from two continents in this sub dataset
three_countries_data[['continent', 'country']].drop_duplicates()
```

```python
sns.lineplot(data=three_countries_data, x='year', y='lifeExp', hue='country', style='continent')
plt.title('Life Expectancy Over Years by Continent (Different Styles)')
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
plt.grid()
plt.show()
```

### 5. `size`: Grouping variable that will produce lines with different widths.
 -  Example: `size='pop'` (Note: Normally, this parameter is used with numerical variables.)



```python
three_countries_data.columns
```

```python
sns.lineplot(data=three_countries_data, x='year', y='lifeExp', hue='country', size='continent')
plt.title('Life Expectancy Over Years by Continent (Different Sizes)')
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
plt.show()
```

### 6. `markers`: Boolean or list to draw markers on the lines.
-  Example: `markers=True`



```python
sns.lineplot(data=gapminder, x='year', y='lifeExp', hue='continent', style='continent', markers=True)
plt.title('Life Expectancy Over Years by Continent (With Markers)')
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
plt.grid()
plt.show()
```

### 7. `dashes`: Boolean or list to draw dashed lines.
 - Example: `dashes=True`



```python
sns.lineplot(data=gapminder, x='year', y='lifeExp', hue='continent', style='continent', dashes=True)
plt.title('Life Expectancy Over Years by Continent (With Dashes)')
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
plt.show()
```

### 8. `legend`: Controls the legend display.
 -  Example: `legend='full'`


```python
sns.lineplot(data=gapminder, x='year', y='lifeExp', hue='continent', style='continent', markers=True, dashes=False, legend='full')
plt.title('Life Expectancy Over Years by Continent (With Legend)')
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
plt.legend(title='Continent', loc=2)
plt.show()
```

## Line Plot with `seaborn.relplot`
- `seaborn.relplot` is a higher-level interface for creating relational plots, including line plots.



```python
# Basic relplot line plot

sns.relplot(data=country_data, x='year', y='lifeExp', kind='line')
plt.title('Life Expectancy in Canada Over Years (relplot)')
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
plt.legend()

plt.show()
```

```python
# Basic relplot line plot

sns.lineplot(data=country_data, x='year', y='lifeExp')
plt.title('Life Expectancy in Canada Over Years (relplot)')
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
plt.show()
```

```python
# relplot with multiple parameters
sns.relplot(data=three_countries_data, x='year', y='lifeExp', kind='line',hue='country',style='continent')
plt.title('Life Expectancy Over Years by Continent (relplot)')
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
plt.show()
```
