#**Relational Plots**

Relational plots in Seaborn are used to visualize relationships between variables in a dataset.

They are particularly useful for exploring how two or more variables interact with each other.

Seaborn provides several functions to create relational plots, each suited for different types of relationships and data structures.

**Types of Relational Plots in Seaborn**
1.	scatterplot()
2.	lineplot()
3.	relplot()

**Customizing Relational Plots:**
Seaborn's relational plots can be extensively customized using various parameters:
*	**hue**: Assign different colors to different categories.
*	**style**: Assign different marker styles or line styles to different categories.
*	**size**: Assign different sizes to different categories.
*	**palette**: Specify the color palette to use for different categories.
*	**markers**: Customize the markers used in scatter plots.

### 1. `sns.scatterplot`
This function creates a scatter plot where individual data points are displayed on the graph according to the values of the specified columns.

### 2. `data=tips`
- The `data` parameter specifies the DataFrame being used. In this case, it’s `tips`, a dataset containing restaurant tips (from Seaborn’s example datasets).

### 3. `x="total_bill"`
- This defines the data to be plotted along the x-axis. Here, `total_bill` represents the total amount spent at the restaurant.

### 4. `y="tip"`
- This defines the data to be plotted along the y-axis. Here, `tip` represents the tip amount left by the customer.

### 5. `hue="day"`
- The `hue` parameter categorizes the data by color based on different values of the `day` column (e.g., days of the week). Each `day` will have a different color.

### 6. `style="time"`
- The `style` parameter categorizes the data by marker shape based on the values of the `time` column (e.g., `Lunch` or `Dinner`). Different values of `time` will have different marker styles.

### 7. `size="size"`
- The `size` parameter changes the size of the markers based on the values in the `size` column (e.g., party size). Larger groups will be represented by bigger markers.

### 8. `palette="Set2"`
- The `palette` parameter specifies a color palette used for the plot. In this case, `"Set2"` is a predefined color palette in Seaborn, which gives visually distinct colors.

### 9. `markers=["1", "+"]`
- The `markers` parameter customizes the specific shapes used for different categories in `style`. Here:
  - `"1"`: A certain marker (small triangle).
  - `"+"`: A plus sign.


These shapes are applied to different groups of the `time` variable.

### 10. `plt.title("Scatter Plot of Tips")`
- Adds a title to the plot: "Scatter Plot of Tips."

### 11. `plt.show()`
- This displays the plot.

### Visualization Summary:
- **X-axis**: `total_bill` (how much was spent)
- **Y-axis**: `tip` (how much tip was given)
- **Color (hue)**: Based on `day` (days of the week)
- **Marker shape**: Based on `time` (Lunch or Dinner)
- **Marker size**: Based on `size` (party size)
