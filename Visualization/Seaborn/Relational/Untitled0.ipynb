```python
def find_unique(col):
  print("-------------------")
  print(col.name)
  print(col.unique())
  print(col.nunique())
  print("-------------------")
```

```python
df = tips[['sex', 'smoker', 'day', 'time', 'size']]
```

```python
df.columns
```

```python
for eachcol in df.columns:
  find_unique(df[eachcol])
```

```python
# Customizing Relational Plots: Seaborn's relational plots can be extensively customized using various parameters:

# hue: Assign different colors to different categories.
# style: Assign different marker styles or line styles to different categories.
# size: Assign different sizes to different categories.
# palette: Specify the color palette to use for different categories.
# markers: Customize the markers used in scatter plots.
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip")
```

```python
tips.columns
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue='day')
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue='day', style='time')
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue='day', style='time', size='size', palette="Set1", markers= ["s","P"])
plt.legend(loc=2)
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", hue='day', style='time', size='size', palette="Set1", markers= ["s","P"])
# plt.legend(loc=2)
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", col = "smoker")
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", col = "smoker", row='sex')
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", col = "smoker", row='day')
```

```python

import seaborn as sns
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
# Load the dataset
gapminder = px.data.gapminder()
gapminder.head()
```

```python
canada =  gapminder[gapminder['country']=='Canada']
```

```python
canada.info()
```

```python
# sns.lineplot(data=df, x=?, y=?)
```

```python
sns.lineplot(data=canada, x='year', y='lifeExp')
```

```python
countries = ['Canada', 'United States', 'India','Albania']
three_countries_data = gapminder[gapminder['country'].isin(countries)]
```

```python
sns.lineplot(data=three_countries_data, x='year', y='lifeExp')
```

```python
three_countries_data
```

```python
three_countries_data['continent'].unique()
```

```python
sns.lineplot(data=three_countries_data, x='year', y='lifeExp', hue='country')
```

```python
sns.lineplot(data=three_countries_data, x='year', y='lifeExp', hue='continent')
```

```python
sns.lineplot(data=three_countries_data, x='year', y='lifeExp', hue='country', style='continent')
```
